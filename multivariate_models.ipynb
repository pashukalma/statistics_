### Multivariate Models

#### **Random Vectors**

Multivariate models involve a random vector $\mathbf{X}$ of the form $\mathbf{X} = (X_1, \dots, X_n)^T$. The mean of a random vector $\mathbf{X}$ is defined by $\boldsymbol{\mu} = [\mu_1, \dots, \mu_n]^T = [E[X_1], \dots, E[X_n]]^T$.

The covariance matrix $\boldsymbol{\Sigma}$ is defined to be $\boldsymbol{\Sigma} = V(\mathbf{X})$. This is the variance matrix or the variance-covariance matrix. More explicitly, its $(i, j)$-th element is given by $\sigma_{ij} = \text{Cov}(X_i, X_j) = E[(X_i - E[X_i])(X_j - E[X_j])]$. When $i=j$, $\sigma_{ii} = \text{Var}(X_i)$.

$\boldsymbol{\Sigma} = \begin{pmatrix}
V(X_1) & \text{Cov}(X_1, X_2) & \cdots & \text{Cov}(X_1, X_k) \\
\text{Cov}(X_2, X_1) & V(X_2) & \cdots & \text{Cov}(X_2, X_k) \\
\vdots & \vdots & \ddots & \vdots \\
\text{Cov}(X_k, X_1) & \text{Cov}(X_k, X_2) & \cdots & V(X_k)
\end{pmatrix}$

Let $\mathbf{a}$ be a vector of length same as that of random vector $\mathbf{X}$ with mean $\boldsymbol{\mu}$ and variance $\boldsymbol{\Sigma}$. Then $E(\mathbf{a}^T \mathbf{X}) = \mathbf{a}^T \boldsymbol{\mu}$ and $V(\mathbf{a}^T \mathbf{X}) = \mathbf{a}^T \boldsymbol{\Sigma} \mathbf{a}$. If $\mathbf{A}$ is a matrix with $k$ columns then $E(\mathbf{A}\mathbf{X}) = \mathbf{A}\boldsymbol{\mu}$ and $V(\mathbf{A}\mathbf{X}) = \mathbf{A}\boldsymbol{\Sigma}\mathbf{A}^T$.

Now suppose we have a random sample of $N$ vectors. Let the sample data be represented by a matrix $\mathbf{X}_{data}$ where each row corresponds to an observation and each column to a component of the random vector.

The sample mean $\bar{\mathbf{X}}$ is a vector of $(\bar{X}_1, \dots, \bar{X}_n)^T$, where $\bar{X}_i = \frac{1}{N} \sum_{j=1}^{N} X_{ij}$. We define the sample variance matrix $\mathbf{S}$ where $s_{ab} = \frac{1}{N-1} \sum_{j=1}^{N} (X_{aj} - \bar{X}_a)(X_{bj} - \bar{X}_b)$. It follows that $E[\bar{\mathbf{X}}] = \boldsymbol{\mu}$ and $E[\mathbf{S}] = \boldsymbol{\Sigma}$.

$\mathbf{X}_{data} = \begin{pmatrix}
X_{11} & X_{12} & \cdots & X_{1n} \\
X_{21} & X_{22} & \cdots & X_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
X_{N1} & X_{N2} & \cdots & X_{Nn}
\end{pmatrix}$  $\quad$
$\bar{\mathbf{X}} = \begin{pmatrix}
\bar{X}_1 \\
\bar{X}_2 \\
\vdots \\
\bar{X}_n
\end{pmatrix}$ $\quad$
$\mathbf{S} = \begin{pmatrix}
s_{11} & s_{12} & \cdots & s_{1n} \\
s_{21} & s_{22} & \cdots & s_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
s_{n1} & s_{n2} & \cdots & s_{nn}
\end{pmatrix}$

#### **Estimating the Correlation**

Consider $N$ data points from a bivariate distribution. Recall that the correlation between $X_1$ and $X_2$ is $\rho = E[(X_1 - \mu_1)(X_2 - \mu_2)]/(\sigma_1 \sigma_2)$.

The sample correlation (the plug-in estimator) is given by
$\hat{\rho} = \frac{\sum_{i=1}^N (X_{1i} - \bar{X}_1)(X_{2i} - \bar{X}_2)}{\sqrt{\sum_{i=1}^N (X_{1i} - \bar{X}_1)^2 \sum_{i=1}^N (X_{2i} - \bar{2}_2)^2}}$.
Alternatively, if $s_{12}$ is the sample covariance between $X_1$ and $X_2$, and $s_1$ and $s_2$ are the sample standard deviations of $X_1$ and $X_2$ respectively, then
$\hat{\rho} = \frac{s_{12}}{s_1 s_2}$.

We can construct a confidence interval for $\rho$ by applying the delta method as usual. However, it turns out that we get a more accurate confidence interval by first constructing a confidence interval for a function $\theta = f(\rho)$ and then applying the inverse function $f^{-1}$. This method is due to Fisher.

We define the Fisher transformation as
- $f(r) = \frac{1}{2} \left[ \log(1 + r) - \log(1 - r) \right]$ and let $\theta = f(\rho)$. The inverse of $f$ is $g(z) = f^{-1}(z) = \frac{e^{2z} - 1}{e^{2z} + 1}$.

#### **Approximate Confidence Interval for the Correlation**

To construct an approximate $1 - \alpha$ confidence interval for the correlation coefficient $\rho$, we use Fisher's transformation method:

Compute $\hat{\theta}$: First, compute the Fisher-transformed value of the sample correlation $r$ (often denoted as $\hat{\rho}$ or $r$):
- $\hat{\theta} = f(r) = \frac{1}{2} \left[ \log(1 + r) - \log(1 - r) \right]$

Compute the approximate standard error of $\hat{\theta}$: The approximate standard error of $\hat{\theta}$ is given by:
- $se(\hat{\theta}) = \frac{1}{\sqrt{N-3}}$, where $N$ is the sample size.

Construct an approximate $1 - \alpha$ confidence interval for $\theta$: An approximate confidence interval for $\theta = f(\rho)$ is given by:
- $(a, b) = \left( \hat{\theta} - z_{\alpha/2} \cdot se(\hat{\theta}), \hat{\theta} + z_{\alpha/2} \cdot se(\hat{\theta}) \right)$, where $z_{\alpha/2}$ is the critical value from the standard normal distribution for the desired confidence level (e.g., for a 95% CI, $z_{0.025} \approx 1.96$).

Apply the inverse transformation to get the confidence interval for $\rho$: Finally, apply the inverse Fisher transformation $g(z) = \frac{e^{2z} - 1}{e^{2z} + 1}$ to the interval $(a, b)$ to obtain the confidence interval for $\rho$: $\left[ \frac{e^{2a} - 1}{e^{2a} + 1}, \frac{e^{2b} - 1}{e^{2b} + 1} \right]$

### **Multinomial Distribution**

#### Review of Multinomial distribution
Consider drawing a ball from an urn that has $n$ balls of $k$ colors. Let $\mathbf{p} = (p_1, \dots, p_k)$ where $p_j \geq 0$ are the probabilities of drawing (with replacement) a ball of each color; $\sum_{j=1}^k p_j = 1$. Draw $n$ times and let $\mathbf{X} = (X_1, \dots, X_k)$ where $X_j$ is the number of times that color $j$ appeared; so $\sum_{j=1}^k X_j = n$. We say $\mathbf{X}$ has a Multinomial($n$, $\mathbf{p}$) distribution. The probability mass function is:
- $f(\mathbf{x}; \mathbf{p}) = \binom{n}{x_1, \dots, x_k} p_1^{x_1} \cdots p_k^{x_k}$

where the multinomial coefficient is given by $\binom{n}{x_1, \dots, x_k} = \frac{n!}{x_1! x_2! \cdots x_k!}$.

Let $\mathbf{X} \sim \text{Multinomial}(n, \mathbf{p})$. Then the marginal distribution of $X_j$ is $X_j \sim \text{Binomial}(n, p_j)$. The mean and variance of $X_j$ are
- $E[X_j] = np_j$, $V(X_j) = np_j(1 - p_j)$.

**Proof for $X_j \sim \text{Binomial}(n, p_j)$**:
Consider any specific component $X_j$. For each of the $n$ draws, the outcome is either 'color $j$' (with probability $p_j$) or 'not color $j$' (with probability $1 - p_j$). Since each draw is independent, the number of times 'color $j$' appears in $n$ draws, which is $X_j$, follows a Binomial distribution with parameters $n$ (number of trials) and $p_j$ (probability of success). Thus, $X_j \sim \text{Binomial}(n, p_j)$.

**Compute $\text{Cov}(X_i, X_j)$ for $i \neq j$**.

We know that $X_i + X_j$ represents the number of times either color $i$ or color $j$ appears in $n$ trials. The probability of drawing either color $i$ or color $j$ in a single trial is $p_i + p_j$, therefore, $X_i + X_j \sim \text{Binomial}(n, p_i + p_j)$.

From the properties of the Binomial distribution, we have $V(X_i + X_j) = n(p_i + p_j)(1 - (p_i + p_j))$. On the other hand, using the property of variance for sums of random variables $V(X_i + X_j) = V(X_i) + V(X_j) + 2\text{Cov}(X_i, X_j)$. We also know $V(X_i) = np_i(1 - p_i)$ and $V(X_j) = np_j(1 - p_j)$.

Making the substitutions we get $-2np_i p_j = 2\text{Cov}(X_i, X_j)$, and the covariance
- $\text{Cov}(X_i, X_j) = -np_i p_j$

**MLE** of $\mathbf{p}$: $\hat{\mathbf{p}} = (\hat{p}_1, \dots, \hat{p}_k)$, where $\hat{p}_j = X_j/n$.

**Proof of MLE**:
The pmf for the Multinomial distribution is $f(\mathbf{x}; \mathbf{p}) = \frac{n!}{x_1! \cdots x_k!} p_1^{x_1} \cdots p_k^{x_k}$. The log-likelihood is
$l(\mathbf{p}) = \log(p_1^{x_1} \cdots p_k^{x_k}) = \sum_{j=1}^k x_j \log p_j$.

To maximize $l(\mathbf{p})$ subject to the constraint $\sum_{j=1}^k p_j = 1$, we use the method of Lagrange multipliers,  ${L}(\mathbf{p}, \lambda) = \sum_{j=1}^k x_j \log p_j + \lambda \left( \sum_{j=1}^k p_j - 1 \right)$.

To find the maximum, we take the partial derivative with respect to each $p_j$ and $\lambda$, and set them to zero. For $p_j$ $\frac{\partial {L}}{\partial p_j} = \frac{x_j}{p_j} + \lambda = 0 \implies p_j = -\frac{x_j}{\lambda}$. For $\lambda$ $\frac{\partial {L}}{\partial \lambda} = \sum_{j=1}^k p_j - 1 = 0 \implies \sum_{j=1}^k p_j = 1$. Substitute $p_j = -\frac{x_j}{\lambda}$ into the constraint equation $\sum_{j=1}^k \left( -\frac{x_j}{\lambda} \right) = 1$. Since $\sum_{j=1}^k x_j = n$ (the total number of draws), $-\frac{n}{\lambda} = 1 \implies \lambda = -n$. Now substitute $\lambda = -n$ back into the expression for $p_j$, $p_j = -\frac{x_j}{-n} = \frac{x_j}{n}$.

- The MLE for $p_j$ is $\hat{p}_j = X_j/n$.

**The Variance of the MLE**: $\hat{\mathbf{p}} = \mathbf{X}/n$.

The direct approach is to compute the variance matrix of $\hat{\mathbf{p}}$, $V(\hat{\mathbf{p}}) = V\left(\frac{\mathbf{X}}{n}\right)$.

Using the property $V(c\mathbf{Y}) = c^2 V(\mathbf{Y})$ for a scalar $c$ and random vector $\mathbf{Y}$ (this applies component-wise for the variance matrix),
$V(\hat{\mathbf{p}}) = \frac{1}{n^2} V(\mathbf{X})$. The covariance matrix of $\mathbf{X}$ (where $X_j$ is the count for color $j$) has diagonal elements $V(X_j) = np_j(1 - p_j)$ and off-diagonal elements $\text{Cov}(X_i, X_j) = -np_i p_j$.

Therefore, the covariance matrix $V(\mathbf{X})$ and the variance of the MLE are:
- $\mathbf{\Sigma}_{\mathbf{X}} = \begin{pmatrix}
np_1(1-p_1) & -np_1p_2 & \cdots & -np_1p_k \\
-np_2p_1 & np_2(1-p_2) & \cdots & -np_2p_k \\
\vdots & \vdots & \ddots & \vdots \\
-np_kp_1 & -np_kp_2 & \cdots & np_k(1-p_k)
\end{pmatrix}$ $\quad$
$V(\hat{\mathbf{p}}) = \frac{1}{n^2} \mathbf{\Sigma}_{\mathbf{X}} = \frac{1}{n} \begin{pmatrix}
p_1(1-p_1) & -p_1p_2 & \cdots & -p_1p_k \\
-p_2p_1 & p_2(1-p_2) & \cdots & -p_2p_k \\
\vdots & \vdots & \ddots & \vdots \\
-p_kp_1 & -p_kp_2 & \cdots & p_k(1-p_k)
\end{pmatrix}$

This matrix gives the variances of the individual $\hat{p}_j$'s on the diagonal and the covariances between $\hat{p}_i$ and $\hat{p}_j$ on the off-diagonal.

### **Multivariate Normal**

Let $\mathbf{Z} = (Z_1, \dots, Z_k)$ where $Z_1, \dots, Z_k \sim N(0,1)$ are independent. The density of $\mathbf{Z}$ is $f(\mathbf{z}) = \frac{1}{(2\pi)^{k/2}} \exp\left(-\frac{1}{2} \sum_{j=1}^k z_j^2\right) = \frac{1}{(2\pi)^{k/2}} \exp\left(-\frac{1}{2} \mathbf{z}^T \mathbf{z}\right)$.

The variance matrix of $\mathbf{Z}$ is the identity matrix $\mathbf{I}$. We write $\mathbf{Z} \sim N(\mathbf{0}, \mathbf{I})$ where it is understood that $\mathbf{0}$ is a vector of $k$ zeroes. We say $\mathbf{Z}$ has a standard multivariate Normal distribution.

More generally, a vector $X$ has a multivariate Normal distribution, denoted by $X \sim N(\mu, \Sigma)$ where $\mu$ is a vector of length $k$ and $\Sigma$ is a $k \times k$ symmetric, positive definite matrix. Then $E(X) = \mu$ and $V(X) = \Sigma$. Setting $\mu = \mathbf{0}$ and $\Sigma = \mathbf{I}$ gives back the standard Normal.

The following properties hold:
1. If $\mathbf{Z} \sim N(\mathbf{0}, \mathbf{I})$ and $X=\mu +\Sigma^{1/2}\mathbf{Z}$ then $X \sim N(\mu,\Sigma)$.
2. If $X \sim N(\mu,\Sigma)$, then $\Sigma^{-1/2} (X-\mu) \sim N(\mathbf{0},\mathbf{I})$.
3. If $X \sim N(\mu,\Sigma)$ and $a$ is a vector with the same length as $X$, then $a^T X \sim N(a^T\mu,a^T\Sigma a)$.
4. Let $V =(X-\mu)^T \Sigma^{-1}(X-\mu)$. Then $V \sim \chi^2_k$.

Let $X \sim N (\mu, \Sigma)$. Then:
1. The marginal distribution of $X_a$ is $X_a \sim N(\mu_a,\Sigma_{aa})$.
2. The conditional distribution of $X_b$ given $X_a = x_a$ is $X_b|X_a = x_a \sim N(\mu_{b|a}(x_a), \Sigma_{b|a}(x_a))$ where $\mu_{b|a}(x_a)=\mu_b +\Sigma_{ba} \Sigma_{aa}^{-1}(x_a -\mu_a)$, and $\Sigma_{b|a}(x_a)=\Sigma_{bb} -\Sigma_{ba} \Sigma_{aa}^{-1} \Sigma_{ab}$

Given a random sample of size $n$ from a $N (\mu, \Sigma)$, the log-likelihood (up to a constant not depending on $\mu$ or $\Sigma$) is given by $l(\mu,\Sigma)$.

The MLE is $\hat{\mu}=\bar{X}$ and $\hat{\Sigma}=\frac{n-1}{n} S$.

##### **Proof for $E(a^T X)$ and $V(a^T X)$**

Let $X = (X_1, \dots, X_n)^T$ be a random vector with mean $\mu = E[X]$ and covariance matrix $\Sigma = V(X)$. Let $a = (a_1, \dots, a_n)^T$ be a vector of constants.

**Expectation:**
$E(a^T X) = E\left(\sum_{i=1}^n a_i X_i\right)$
By the linearity of expectation, $E\left(\sum_{i=1}^n a_i X_i\right) = \sum_{i=1}^n a_i E[X_i] = \sum_{i=1}^n a_i \mu_i = a^T \mu$.

**Variance:**
$V(a^T X) = V\left(\sum_{i=1}^n a_i X_i\right)$
We know that $V\left(\sum_{i=1}^n Y_i\right) = \sum_{i=1}^n V(Y_i) + 2 \sum_{i<j} \text{Cov}(Y_i, Y_j)$.
So, $V\left(\sum_{i=1}^n a_i X_i\right) = \sum_{i=1}^n a_i^2 V(X_i) + 2 \sum_{i<j} a_i a_j \text{Cov}(X_i, X_j)$.
This can be written in matrix form as $a^T \Sigma a$.

Specifically, if $A$ is a matrix with $k$ rows, such that $Y = AX$ is a $k$-dimensional random vector, then
$E(AX) = A E[X] = A\mu$.
$V(AX) = E[(AX - A\mu)(AX - A\mu)^T]$
$= E[A(X - \mu)(X - \mu)^T A^T]$
$= A E[(X - \mu)(X - \mu)^T] A^T$
$= A\Sigma A^T$

##### **Proof for Marginal and Conditional Distributions of Multivariate Normal**

Let $X = (X_1, \dots, X_p)^T$ be a random vector that follows a multivariate normal distribution $N(\mu, \Sigma)$, where $\mu = E[X]$ is the mean vector and $\Sigma = V(X)$ is the covariance matrix.

We partition $X$, $\mu$, and $\Sigma$ as follows:
$X = \begin{pmatrix} X_a \\ X_b \end{pmatrix}$, $\mu = \begin{pmatrix} \mu_a \\ \mu_b \end{pmatrix}$, $\Sigma = \begin{pmatrix} \Sigma_{aa} & \Sigma_{ab} \\ \Sigma_{ba} & \Sigma_{bb} \end{pmatrix}$
where $X_a$ is $k \times 1$, $X_b$ is $(p-k) \times 1$, and $\Sigma_{aa}$ is $k \times k$, etc.

**1. Marginal Distribution of $X_a$**

A fundamental property of the multivariate normal distribution is that any subvector of a multivariate normal vector is also multivariate normal. Therefore, the marginal distribution of $X_a$ is given by taking its corresponding mean vector and covariance matrix from the partitioned $\mu$ and $\Sigma$.

Thus, $X_a \sim N(\mu_a, \Sigma_{aa})$.

**2. Conditional Distribution of $X_b$ given $X_a = x_a$**

The conditional distribution of $X_b$ given $X_a = x_a$ is also normal. Its mean and covariance matrix are given by the following formulas:

*   Conditional Mean:
    $E[X_b | X_a = x_a] = \mu_{b|a}(x_a) = \mu_b + \Sigma_{ba} \Sigma_{aa}^{-1} (x_a - \mu_a)$

*   Conditional Covariance:
    $V(X_b | X_a = x_a) = \Sigma_{b|a}(x_a) = \Sigma_{bb} - \Sigma_{ba} \Sigma_{aa}^{-1} \Sigma_{ab}$

Therefore, the conditional distribution is $X_b | X_a = x_a \sim N(\mu_{b|a}(x_a), \Sigma_{b|a}(x_a))$.

**Write a custom function to generate observations from a Multinomial(n, p) distributions.**

Combinatorial Approach for Multinomial Sampling: To understand multinomial sampling from a combinatorial perspective, imagine an urn containing balls of $k$ different colors. Each color $j$ has a certain proportion, $p_j$, representing the probability of drawing a ball of that color. If we draw $n$ balls *with replacement*, the number of times each color appears ($X_1, X_2, \dots, X_k$) follows a multinomial distribution. Let's simulate this process by repeatedly drawing from the 'urn' and counting the results.

We also compare the performance with the np.random.multinomial(n,p) function.

```python
import numpy as np

def multinomial_combinatorial(n, p, nsim):
    """
    Simulates multinomial sampling using a combinatorial (urn-draw) approach.

    Args:
        n (int): The number of trials (draws from the urn).
        p (list): A list of probabilities for each category/color.
        nsim (int): The number of simulations to run.

    Returns:
        np.ndarray: An array where each row represents a simulation
                    and contains the counts for each category.
    """
    k = len(p)
    results = np.zeros((nsim, k), dtype=int)

    # Ensure probabilities sum to 1
    p = np.array(p) / np.sum(p)

    # Create an 'urn' of categories based on probabilities
    categories = np.arange(k)

    for i in range(nsim):
        # Simulate n draws from the urn
        draws = np.random.choice(categories, size=n, p=p)
        # Count occurrences of each category
        for j in range(k):
            results[i, j] = np.sum(draws == j)
    return results
```

```python
''' Multinomial(n, p) custom class  '''
def multinomial_np(n, p, nsim=1):
    cum_probs = np.cumsum(p)
    def get_observation():
        cnts = np.zeros(cum_probs.size).astype(int)
        rvs = np.random.uniform(size=n)
        for i in range(n):
            cnts[np.argmin(rvs[i] > cum_probs)] += 1
        return cnts
    return np.array([get_observation() for _ in range(nsim)])
```

**Write a function to generate observations from a Multivariate normal.**. We construct the samples of standard multivariate with mean mu and covariance matrix. To simplify it let's take Z ~ N(0, I), making $X = mu + \Sigma^1/2 Z$


```python
def multivar_normal_obs(mu, sigma, number_of_sim=1):
    mu_ = np.array(mu)
    sigma_ = np.array(sigma)
    k = mu_.shape[0]
    U, D, V = np.linalg.svd(sigma_)    # eigenvalue decomposition
    sigma_sqrt = U @np.diag(np.sqrt(D))

    def random_normals(size):
        def b_m(u1, u2):
            R = np.sqrt(-2 * (np.log(u1)))
            theta = 2 *np.pi *u2
            z0, z1 = R*np.cos(theta), R*np.sin(theta)
            return z0, z1

        def normal_gen(uniform_gen):
            while True:
                z0, z1 = b_m(next(uniform_gen), next(uniform_gen))
                yield z0
                yield z1

        def rng_(batch_size):
            while True:
                for v in np.random.uniform(size=batch_size):
                    yield v

        rns = np.empty(size)
        gen = normal_gen(rng_(min(size, 1024)))
        for i in range(size):
            rns[i] = next(gen)
        return rns

    def get_observation():
        z = random_normals(k)
        return mu_ + sigma_sqrt @ z

    return np.array([get_observation() for _ in range(number_of_sim)])
```

```python
mu = [1, 4]
sigma = [[2, 1], [1, 2]]
observations_ = multivar_normal_obs(mu, sigma, 5000)
plt.hexbin(observations_[:, 0], observations_[:, 1], cmap = plt.cm.Reds)
```

```python
sigma22 = Sigma[1, 1]
theoretical_rho = sigma12 / np.sqrt(sigma11 * sigma22)
print(f"Theoretical correlation \u03C1: {theoretical_rho:.4f}")
```

**Sample Statistics and Confidence Interval**

```python
sample_correlation = np.corrcoef(x1_samples, x2_samples)[0, 1]
print(f"Sample correlation: {sample_correlation:.4f}")
```

95% Confidence Interval for $\rho$; use Fisher's transformation to construct the confidence interval.


```python
from scipy.stats import norm

# Fisher's transformation function
def fisher_transform(r):
    # Ensure r is within (-1, 1) to avoid log of non-positive numbers
    r = np.clip(r, -0.999999, 0.999999)
    return 0.5 * (np.log(1 + r) - np.log(1 - r))

# Inverse Fisher's transformation function
def inverse_fisher_transform(z):
    return (np.exp(2 * z) - 1) / (np.exp(2 * z) + 1)

# Compute Fisher-transformed sample correlation
rho_hat = sample_correlation
theta_hat = fisher_transform(rho_hat)

# Compute standard error of theta_hat
se_theta_hat = 1 / np.sqrt(num_samples - 3)

# For a 95% CI, z_alpha/2 is approximately 1.96
z_alpha_2 = norm.ppf(0.975)

# Construct CI for theta
lower_theta = theta_hat - z_alpha_2 * se_theta_hat
upper_theta = theta_hat + z_alpha_2 * se_theta_hat

# Apply inverse transformation to get CI for rho
lower_rho = inverse_fisher_transform(lower_theta)
upper_rho = inverse_fisher_transform(upper_theta)

print(f"95% Confidence Interval for \u03C1: [{lower_rho:.4f}, {upper_rho:.4f}]")
```
