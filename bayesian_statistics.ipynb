#### Postulates of Classical (Frequentist) Inference:

Probability is defined as the **long-run relative frequency** of an event in a series of repeatable experiments. It's objective and refers to the properties of data generating processes. **Parameters are considered fixed, but unknown** constants (mean, variance, proportions). **Inference is primarily drawn from the observed sample data**. The goal is to make statements about the fixed population parameters based on the sampling distribution of estimators.

Frequentist inference heavily relies on concepts like null **hypothesis significance testing (p-values) and confidence intervals**. These methods quantify the uncertainty of estimates and the evidence against a null hypothesis, assuming repeated sampling.

**No Prior Information**. Prior beliefs or external information about the parameters are generally not incorporated into the statistical model. Inference is solely driven by the data.

#### Postulates of Bayesian Inference:

**Probability as Subjective Belief (Degree of Belief)**, is interpreted as a degree of belief regarding an event or parameter. It can be updated as new evidence becomes available.**Parameters are Random Variables**. Parameters are treated as r.v.s with their own probability distributions. This allows for expressing uncertainty about parameters directly.

Before observing any data, existing knowledge or beliefs about the parameters are expressed through a **prior probability distribution** ($P(\theta)$).

The data observed is used to construct a **likelihood function** ($P(D|\theta)$), which quantifies how probable the observed data $D$ is for different values of the parameter $\theta$.

The core of Bayesian inference is **Bayes' Theorem**, which updates the prior beliefs with the evidence from the data to produce a **posterior probability distribution** ($P(\theta|D)$): $P(\theta|D) = \frac{P(D|\theta)P(\theta)}{P(D)}$ Where $P(D)$ is the evidence or marginal likelihood, serving as a normalizing constant.

**Inference from Posterior Distribution**. All inference (point estimates, credible intervals, hypothesis testing) is derived directly from the posterior distribution, that represents the updated belief about the parameter after observing the data.

**Incorporation of All Information**. Bayesian inference naturally incorporates both prior knowledge and observed data in a coherent and principled way.

#### Bayesian Approach to Statistical Decision Making

This provides a formal framework for making decisions under uncertainty by integrating prior beliefs, observed data, and the consequences of different actions. Listed are the key components.
- **Prior Distribution ($P(\theta)$)** represents the decision-maker's initial beliefs about the unknown parameter $\theta$ (e.g., the true state of the world, the effect of a treatment) *before* observing any new data. This prior can be based on historical data, expert opinion, or subjective assessment.
- **Likelihood Function ($P(D|\theta)$)** describes how probable the observed data $D$ is for each possible value of the parameter $\theta$. It quantifies the information provided by the data regarding the parameter.
- **Posterior Distribution ($P(\theta|D)$)** is the updated belief about the parameter $\theta$ *after* observing the data $D$. It is calculated using Bayes' Theorem $P(\theta|D) = \frac{P(D|\theta)P(\theta)}{P(D)}$. The posterior distribution combines the prior beliefs with the evidence from the data, providing a complete summary of current knowledge about $\theta$.
- **Actions ($a$)** are the possible decisions or actions available to the decision-maker.
- **Loss Function ($L(\theta, a)$)** quantifies the cost or penalty incurred for taking a particular action $a$ when the true state of the world is $\theta$. A loss function measures the undesirable consequences of making a wrong decision. Conversely, a utility function measures desirable outcomes.

#### The Decision Process

The goal of Bayesian decision theory is to choose the action that minimizes the expected loss (or maximizes expected utility) with respect to the posterior distribution of $\theta$.

1.  **Calculate Expected Loss for Each Action**  which is the average loss weighted by the posterior probability of each $\theta$ $E[L(a|D)] = \int L(\theta, a) P(\theta|D) d\theta$. (For discrete $\theta$, the integral becomes a sum.)

2.  **Choose the Optimal Bayesian Action** $a^*$ that minimizes this expected loss $a^* = \operatorname{argmin}_a E[L(a|D)]$

#### Characteristics and Advantages

**Coherent and Unified Framework** giving a consistent way to combine prior knowledge with new data. **Quantifies Uncertainty** directly by producing probability distributions for parameters and predictions, allowing for a complete quantification of uncertainty. **Incorporates Prior Information** integrating prior beliefs, which can be particularly valuable when data is scarce or expensive to collect. **Optimal Decisions** lead to decisions that are optimal under the specified loss function and current state of knowledge. **Direct Interpretations** of posterior probabilities and credible intervals in probabilistic interpretations (e.g., "there is a 95% probability that the parameter is between X and Y"), unlike frequentist confidence intervals.

#### Prior and Posterior Beliefs in Bayesian Inference

At the core of Bayesian inference are the concepts of prior and posterior beliefs, which represent our evolving understanding of a parameter or hypothesis as new information becomes available.

**Prior Beliefs ($P(\theta)$)**, the prior probability distribution of our initial degree of belief about an unknown parameter $\theta$ (e.g., the true average height of a population, the effectiveness of a new drug) *before* we observe any new data. These beliefs can be based on historical data, expert opinion, previous studies, or even subjective judgments. They quantify what we know (or think we know) about $\theta$ in the absence of current evidence. The prior acts as a baseline, guiding our initial expectations and preventing us from drawing conclusions solely from limited data.

**Posterior Beliefs ($P(\theta|D)$)**, the posterior probability distribution of our updated degree of belief about the unknown parameter $\theta$ *after* observing new data $D$. It is calculated using **Bayes' Theorem**, which formally combines the prior beliefs with the evidence provided by the data's likelihood $P(\theta|D) = \frac{P(D|\theta)P(\theta)}{P(D)}$

Where:
*   $P(\theta|D)$ is the **posterior probability** (what we want to know).
*   $P(D|\theta)$ is the **likelihood function** (how well the data $D$ fits different values of $\theta$).
*   $P(\theta)$ is the **prior probability** (our initial belief).
*   $P(D)$ is the **marginal likelihood** or evidence (a normalizing constant that ensures the posterior sums to 1).

The posterior distribution is the ultimate output of Bayesian inference. All inferences, such as point estimates, credible intervals, and hypothesis tests, are derived directly from this distribution.

#### Beta Distribution as a Prior in Bayesian Statistics

The **Beta distribution** is a continuous probability distribution defined on the interval $[0, 1]$, parameterized by two positive shape parameters, $\alpha$ (alpha) and $\beta$ (beta), with pdf PDF) is given by $f(p; \alpha, \beta) = \frac{p^{\alpha-1}(1-p)^{\beta-1}}{B(\alpha, \beta)}$

Where:
*   $p$ is the variable, representing a probability or proportion, and $0 \le p \le 1$; $\alpha > 0$ and $\beta > 0$ are the shape parameters.
*   $B(\alpha, \beta)$ is the Beta function, which acts as a normalizing constant to ensure the total probability integrates to 1. It is defined as $B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}$, where $\Gamma$ is the gamma function.

#### Shape of the Beta Distribution:

The parameters $\alpha$ and $\beta$ dramatically influence the shape of the Beta distribution, allowing it to model a wide variety of prior beliefs about a probability:
*   If $\alpha = 1$ and $\beta = 1$, the distribution is **Uniform** ($U(0,1)$), representing a state of maximum uncertainty or indifference about the true probability.
*   If $\alpha > 1$ and $\beta > 1$, the distribution is **unimodal** (bell-shaped), with its peak at $p = \frac{\alpha-1}{\alpha+\beta-2}$. This indicates a prior belief that the probability is concentrated around a specific value.
*   If $\alpha < 1$ and $\beta < 1$, the distribution is **U-shaped**, with peaks at the extremes (0 and 1). This suggests a prior belief that the probability is likely to be either very low or very high.
*   If $\alpha > \beta$, the distribution is **skewed towards 1**.
*   If $\alpha < \beta$, the distribution is **skewed towards 0**.

#### Role in Bayesian Statistics as a Prior Distribution

The Beta distribution is particularly significant in Bayesian statistics because of its role as a **conjugate prior** for the Bernoulli and Binomial distributions. This means that if the likelihood function for the data comes from a Bernoulli or Binomial distribution, and the prior distribution for the success probability ($p$) is a Beta distribution, then the resulting posterior distribution will also be a Beta distribution.

**Significance of Conjugacy:**
*   **Computational Simplicity:** Conjugacy greatly simplifies the calculation of the posterior distribution. Instead of complex numerical integration, the posterior parameters can be found with simple algebraic updates of the prior parameters.
*   **Interpretability:** The parameters of the Beta distribution ($\alpha$ and $\beta$) can be intuitively interpreted as "pseudo-counts" or "prior observations." If a Beta($\alpha$, $\beta$) prior is used, $\alpha-1$ can be thought of as prior successes and $\beta-1$ as prior failures. When new data with $s$ successes and $f$ failures is observed, the posterior becomes Beta($\alpha+s$, $\beta+f$). This makes updating beliefs straightforward.
*   **Flexibility:** As demonstrated by its diverse shapes, the Beta distribution can represent a wide range of prior beliefs, from strong convictions about the value of $p$ to complete ignorance (uniform prior).

Therefore, the Beta distribution is a foundational tool for Bayesian inference involving one-dimensional probability parameters, offering both mathematical convenience and intuitive interpretability.

### Implementation

#### The Bayesian Method
We choose a prior distribution $f(\theta)$, that expresses our degrees of belief about a parameter $\theta$ before we see any data. We choose a statistical model $f(x|\theta)$ that reflects our beliefs about $x$ given $\theta$. After observing data $X_1,...,X_n$, we update our beliefs and form the posterior distribution $f(\theta|X_1, ... , X_n)$.

#### Bayesian Update Details

**The Posterior Density:**
$f(\theta|x_n) = \frac{f(x_n|\theta)f(\theta)}{\int f(x_n|\theta)f(\theta)d\theta} = \frac{L_n(\theta)f(\theta)}{\int L_n(\theta)f(\theta)d\theta}$

The denominator is a constant that does not depend on $\theta$, a **normalizing constant**. The posterior is proportional to likelihood times prior $f(\theta|x_n) \propto L_n(\theta)f(\theta)$.

**Point Estimate** is obtained for $\theta$ using the posterior mean:
$\hat{\theta} = \int \theta f(\theta|x_n)d\theta = \frac{\int \theta L_n(\theta)f(\theta)d\theta}{\int L_n(\theta)f(\theta)d\theta}$

**Bayesian Interval Estimate** obtained, defining $a$ and $b$ such that $\int_{-\infty}^{a} f(\theta|x_n)d\theta = \int_{b}^{\infty} f(\theta|x_n)d\theta = \alpha/2$.
Let $C=(a,b)$. Then, the probability that $\theta$ falls within this interval is:
$P(\theta\in C|x_n) = \int_{a}^{b} f(\theta|x_n)d\theta = 1-\alpha$. So, $C$ is a $1-\alpha$ posterior interval (also known as a credible interval).

When the prior distribution and the posterior distribution belong to the same family of distributions, we say the prior is **conjugate** to the likelihood function.

#### Functions of Parameters

For a transformed parameter $\tau = g(\theta)$, the posterior CDF for $\tau$ is given by
$H(\tau|x_n) = P(g(\theta) \le \tau) = \int_{A} f(\theta|x_n)d\theta$
where $A = \{ \theta : g(\theta) \le \tau \}$. The posterior density for $\tau$ is then found by taking the derivative of the CDF
$h(\tau|x_n) = H'(\tau|x_n)$.

#### Simulation-Based Approximation of the Posterior

The posterior distribution can often be approximated by simulation. Suppose we draw samples $\theta_1, \dots, \theta_B$ from the posterior $p(\theta|x_n)$.

*   Then, a histogram of $\theta_1, \dots, \theta_B$ approximates the posterior density $p(\theta|x_n)$.
*   An approximation to the posterior mean $\hat{\theta}_n = E(\theta|x_n)$ is given by the sample mean $B^{-1} \sum_{j=1}^{B} \theta_j$.
*   A $1-\alpha$ posterior interval can be approximated by the $(\theta_{\alpha/2}, \theta_{1-\alpha/2})$ interval, where $\theta_{\alpha/2}$ is the $\alpha/2$ sample quantile of $\theta_1, \dots, \theta_B$ and $\theta_{1-\alpha/2}$ is the $1-\alpha/2$ sample quantile.

#### Large Sample Properties for Bayes' Procedures

Under appropriate regularity conditions, we have that the posterior distribution for $\theta$ is approximately $N(\hat{\theta}, s.e.^2)$ where $\hat{\theta}$ is the Maximum Likelihood Estimator (MLE) and $s.e.^2 = 1 / (nI(\hat{\theta}))$, with $I(\hat{\theta})$ being the observed information. Hence, the posterior mean is approximately equal to the MLE, i.e., $E(\theta|x_n) \approx \hat{\theta}$.

Also, if $C_n = (\hat{\theta} - z_{\alpha/2}s.e., \hat{\theta} + z_{\alpha/2}s.e.)$ is the asymptotic frequentist $1-\alpha$ confidence interval, then $C_n$ is also an approximate $1-\alpha$ Bayesian posterior interval: $P(\theta \in C_n|X_n) \to 1-\alpha$.

There is also a **Bayesian delta method**. Let $\tau = g(\theta)$. Then the posterior distribution for $\tau$ given $X_n$ is approximately $N(\hat{\tau}, s.e.\_\tau^2)$ where $\hat{\tau} = g(\hat{\theta})$ and $s.e.\_\tau = s.e. \times |g'(\hat{\theta})|$ (where $g'(\hat{\theta})$ is the derivative of $g$ evaluated at $\hat{\theta}$).

#### Flat Priors, Improper Priors and "Noninformative" Priors

**Improper Priors:** We can adopt a flat prior $f(\theta) \propto c$ where $c > 0$ is a constant. Note that $\int f(\theta)d\theta = \infty$ so this is not a real probability density function in a proper sense. We call such a prior an **improper prior**. In general, improper priors are not a problem as long as the resulting posterior is a well-defined probability distribution.

**Flat Priors are Not Invariant:** The notion of a flat prior is not well-defined because a flat prior on a parameter does not imply a flat prior on a transformed version of a parameter.

**Jeffreys' Prior:** A common choice for a noninformative prior is **Jeffreys' prior**. It is defined as $f(\theta) \propto I(\theta)^{1/2}$ where $I(\theta)$ is the Fisher information function. This rule turns out to be transformation invariant, which is a desirable property for noninformative priors.
In a multiparameter problem, the Jeffreys' prior is defined to be $f(\theta) \propto \sqrt{\det I(\theta)}$, where $\det(A)$ denotes the determinant of a matrix A.

### Continuing with Statistical Decision Theory

#### Comparing Risk Functions

In statistical decision theory, a **risk function** quantifies the average loss associated with an estimator or a decision rule. When choosing among different estimators or decision rules, we often compare their risk functions. A common goal is to find an estimator that minimizes this risk.

Formally, the risk function $R(\theta, \delta)$ of a decision rule $\delta$ (which maps data $X$ to an action $a$) given a loss function $L(\theta, a)$ is the expected loss

$R(\theta, \delta) = E_{\theta}[L(\theta, \delta(X))] = \int L(\theta, \delta(x)) f(x|\theta) dx$, where
- $\theta$ is the true (unknown) parameter.
- $\delta(X)$ is the estimator/decision rule, which is a function of the observed data $X$.
- $L(\theta, a)$ is the loss incurred if the true parameter is $\theta$ and we take action $a$.
- $f(x|\theta)$ is the probability density (or mass) function of the data given $\theta$.

Comparing risk functions allows us to evaluate the performance of different statistical procedures across the entire parameter space. An ideal estimator would have a uniformly smaller risk than all other estimators, but such an estimator rarely exists.

#### Bayes Estimators

A **Bayes estimator** is a decision rule that minimizes the posterior expected loss.

Given a loss function $L(\theta, a)$, a prior distribution $f(\theta)$, and observed data $x$, the Bayes estimator $\delta^*(x)$ is the value $a$ that minimizes the posterior expected loss:
- $E[L(\theta, a)|X=x] = \int L(\theta, a) f(\theta|x) d\theta$

If we are estimating a parameter $\theta$ (i.e., our action $a$ is an estimate $\hat{\theta}$ of $\theta$), common Bayes estimators arise from specific loss functions:
- Squared-Error Loss ($L(\theta, \hat{\theta}) = (\theta - \hat{\theta})^2$): The Bayes estimator is the posterior mean: $\delta^*(x) = E[\theta|X=x]$.
- Absolute-Error Loss ($L(\theta, \hat{\theta}) = |\theta - \hat{\theta}|$): The Bayes estimator is the posterior median.
- 0-1 Loss ($L(\theta, \hat{\theta}) = 1$ if $\theta \ne \hat{\theta}$, $0$ otherwise): The Bayes estimator is the posterior mode (Maximum A Posteriori, MAP estimate).

Bayes estimators incorporate prior information directly, leading to estimates that are optimal given the specified loss function and the updated beliefs about the parameter.

#### Minimax Rules

A **minimax rule** (or minimax estimator) is a decision rule that minimizes the maximum possible risk. Unlike Bayes estimators, which average risk over the posterior distribution, minimax rules focus on the worst-case scenario over the entire parameter space.

The minimax risk is defined as $\inf_{\delta} \sup_{\theta} R(\theta, \delta)$

And a decision rule $\delta^*$ is minimax if $\sup_{\theta} R(\theta, \delta^*) = \inf_{\delta} \sup_{\theta} R(\theta, \delta)$.

In essence, a minimax estimator protects against the largest possible loss, ensuring that the risk does not exceed a certain level, regardless of the true value of the parameter. This approach is often considered conservative, as it prioritizes protection against the worst-case scenario, even if that scenario is unlikely.

Minimax estimators are particularly useful when there is a strong desire to control the maximum possible error, or when prior information is very vague or non-existent, making it difficult to formulate a reliable prior distribution.

#### MLE, Minimax, and Bayes

These three concepts represent different philosophies and approaches to estimation and decision-making in statistics:
- **MLE:** A frequentist concept. The MLE selects the parameter value that maximizes the likelihood of observing the given data. It does not incorporate prior beliefs directly and often has desirable asymptotic properties (consistency, asymptotic normality, efficiency).

- **Bayes Estimator:** A Bayesian concept, minimizes the expected loss with respect to the posterior distribution, combining prior beliefs with data. It explicitly incorporates prior information and a loss function.

- **Minimax Estimator:** A decision-theoretic concept that can be applied in both frequentist and Bayesian contexts (though it's often discussed in a frequentist decision-theory setting). It chooses the estimator that minimizes the maximum possible risk over the entire parameter space. It is a conservative approach, guarding against the worst-case scenario.

Key Differences and Relationships:
- **Prior Information:** MLEs do not use prior information. Bayes estimators are defined by it. Minimax estimators often assume no prior information (or a least favorable prior for finding a minimax rule).
- **Loss Function:** MLEs are not inherently tied to a loss function (though they implicitly assume a form of 0-1 loss in some contexts). Bayes and Minimax estimators are explicitly defined by a chosen loss function.
- **Optimality Criteria:** MLE aims to find the parameter that best explains the data. Bayes aims for optimality under expected posterior loss. Minimax aims for optimality under worst-case risk.
- **Connection:** In some cases, an MLE can be a Bayes estimator (e.g., MAP estimate with a uniform prior can be an MLE if the prior is flat and finite). Also, a Bayes estimator can sometimes be a minimax estimator if the prior is a "least favorable" prior that maximizes the Bayes risk.

### Admissibility (High Level)

In statistical decision theory, an estimator or decision rule is **admissible** if there is no other estimator that is uniformly better (i.e., has a risk function that is less than or equal to the admissible estimator's risk for all possible parameter values, and strictly less for at least one parameter value).

An estimator $\delta_1$ is admissible if there is no other estimator $\delta_2$ such that
- $R(\theta, \delta_2) \le R(\theta, \delta_1)$ for all $\theta \in \Theta$, and
$R(\theta_0, \delta_2) < R(\theta_0, \delta_1)$ for at least one $\theta_0 \in \Theta$.

If such a $\delta_2$ exists, then $\delta_1$ is **inadmissible**.

**Significance:**
*   **Rationality:** Inadmissible estimators are generally considered irrational choices, as a better alternative exists under all circumstances (or at least no worse, and sometimes strictly better).
*   **Bayes and Minimax:** Many Bayes estimators are admissible, especially if the prior distribution is proper (integrates to 1) and positive over the entire parameter space. Minimax estimators are often admissible as well, and sometimes a minimax estimator derived from a proper prior is admissible.

The concept of admissibility helps narrow down the set of reasonable estimators by eliminating those that are demonstrably suboptimal.
