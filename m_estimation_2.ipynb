#### **Asymptotic Variance Comparison Cauchy($\mu, 1$)**

Compare the asymptotic variance for the sample mean and sample median when $X_1, \dots, X_n \overset{i.i.d.}{\sim} \text{Cauchy}(\mu, 1)$. For $X \sim \text{Cauchy}(\mu, 1)$, we have the probability density function (PDF) and cumulative distribution function (CDF) as follows:
- $f_X(x) = \frac{1}{\pi(1 + (x - \mu)^2)}$
- $F_X(x) = \frac{1}{2} + \frac{1}{\pi} \arctan(x - \mu)$
- $F_X^{-1}(t) = \tan\left(\pi \left(t - \frac{1}{2}\right)\right) + \mu$

Notice that these are both continuous on $\mathbb{R}$.1.

**Sample Median**

Using the general result for the asymptotic distribution of the sample median, where $\sqrt{n}(\hat{\mu}_{\text{med}} - \mu) \xrightarrow{d} {N}\left(0, \frac{1}{4f(\mu)^2}\right)$, we find
- $f(\mu) = \frac{1}{\pi(1 + (\mu - \mu)^2)} = \frac{1}{\pi}$
- The asymptotic variance is $\sigma^2_{\text{med}} = \frac{1}{4 \left(\frac{1}{\pi}\right)^2} = \frac{\pi^2}{4} \approx 2.467$.

**Sample Mean**

For the sample mean $\bar{X}_n$
- $\text{AsymVar}(\bar{X}_n) = n \cdot \text{Var}(\bar{X}_n) = \text{Var}(X_i) = \infty$

**Analysis**

The Cauchy distribution is a "pathological" case in statistics because it has such heavy tails that its mean and variance are undefined (infinite).
- The Sample Mean: Because the variance is infinite, the Law of Large Numbers does not apply in the usual way. In fact, the sample mean of a Cauchy distribution has the exact same distribution as a single observation; it never "settles down" as $n$ increases.
- The Sample Median: The median remains a robust and well-defined estimator with a finite asymptotic variance, proving it is far superior to the mean for this specific distribution.

**Sample Mean and Median Distributions for Cauchy Data**

Generate 10,000 samples of size 100 from a Cauchy(0, 1) distribution, calculate the sample mean and median for each sample, and store these values to observe their empirical distributions.

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

num_simulations = 10000
sample_size = 100

cauchy_sample_means = []
cauchy_sample_medians = []

for _ in range(num_simulations):
    # Generate samples from Cauchy(0, 1) distribution
    sample = np.random.standard_cauchy(size=sample_size)

    # Calculate mean and median. Note: Mean for Cauchy distribution is undefined, but we calculate it anyway to observe its behavior
    mean_val = np.mean(sample)
    median_val = np.median(sample)

    # Append to lists
    cauchy_sample_means.append(mean_val)
    cauchy_sample_medians.append(median_val)

# Convert lists to NumPy arrays
cauchy_sample_means = np.array(cauchy_sample_means)
cauchy_sample_medians = np.array(cauchy_sample_medians)

print(f"Generated {num_simulations} sample means and medians, each from a sample size of {sample_size} for Cauchy data. \n"
f"First 5 Cauchy sample means: {cauchy_sample_means[:5]} \n"
f"First 5 Cauchy sample medians: {cauchy_sample_medians[:5]}")
```

### The Problem: Infinite Variance

The **Cauchy distribution** is characterized by an **infinite variance**, meaning the integral required to calculate it diverges due to its exceptionally heavy tails. This makes the occurrence of extreme outliers highly probable and massive in scale. Unlike a Normal distribution, which rapidly diminishes towards zero, the Cauchy distribution maintains significant probability density far from its center. This profoundly influences the choice of estimators for central tendency. For data exhibiting such heavy tails, the **sample mean** is an **unreliable estimator**, producing unstable and often invalid results as it never truly converges. In contrast, the **sample median** offers a **robust and dependable estimate** of central tendency, as its calculation is inherently less susceptible to the influence of extreme values. **M-estimators**, such as those employing Huber loss, provide a balanced approach, mimicking the mean for typical errors while adapting to act like the median when confronted with large outliers, thereby combining aspects of efficiency and robustness.

### Distribution Comparison: Mean vs. Median

When evaluating estimators, **efficiency** quantifies how concentrated an estimator's distribution is around the true parameter (lower variance implies higher efficiency), while **robustness** describes its resilience to deviations from assumptions, particularly the presence of outliers. The **sample mean** is notably **non-robust**, as even a single extreme outlier can significantly distort its value. Conversely, the **sample median** is **highly robust** because its computation relies on the rank of data points rather than their exact magnitudes. For a **Cauchy distribution**, the sample mean entirely fails because its exceptionally thick tails prevent the integral for the expected value from converging, implying the population mean does not exist. Consequently, the sample mean of Cauchy data never stabilizes or converges, rendering foundational statistical theorems like the Law of Large Numbers and the Central Limit Theorem inapplicable.

#### Infinite Variance and its Impact on the Sample Mean (Cauchy Distribution)

The **Cauchy distribution** is a unique 'heavy-tailed' probability distribution with an **undefined mean and variance**, implying a high probability of extreme outliers. This severely impacts the **sample mean** as an estimator. While the Law of Large Numbers (LLN) dictates convergence for finite variance distributions, the sample mean for Cauchy data exhibits profound **instability**, failing to stabilize or converge even with large sample sizes. Theoretically, the sample mean of *n* observations from a Cauchy distribution is itself a Cauchy distribution with the same parameters, meaning its spread does not decrease with additional data.

In stark contrast, the **sample median** for a Cauchy distribution is a remarkably **stable and robust estimator**. It effectively converges to the true location parameter (0 for Cauchy(0,1)) and is far less sensitive to extreme values. Our simulation visually confirmed this, showing cauchy_sample_medians to be much narrower and more concentrated around zero compared to cauchy_sample_means. This stability makes the median a more appropriate and reliable measure of central tendency for heavy-tailed distributions where the sample mean is problematic, as relying on the sample mean can lead to misleading conclusions due to its inherent instability and lack of convergence. This highlights the crucial role of robust estimators like the median and M-estimators when dealing with data prone to significant outliers.

#### Compare Asymptotic Variances of Laplace and Cauchy Distributions

This section compares the asymptotic variances and estimator performance for Laplace(0,1) and Cauchy(0,1) distributions.

For the **Laplace(0,1) Distribution**, which has heavier tails than a Normal distribution but still a finite variance, the **sample median** consistently demonstrates **greater efficiency** than the sample mean. Both empirical and theoretical asymptotic variances indicate the median is a more precise estimator due to its robustness against the moderately heavy tails.

The **Cauchy(0,1) Distribution**, however, is an extreme case with **infinitely heavy tails**, resulting in an undefined population mean and variance. Consequently, the **sample mean is unsuitable and highly inefficient**, failing to converge with effectively infinite variance. Our simulation vividly illustrated this, showing cauchy_sample_means as extremely widespread and unstable. In sharp contrast, cauchy_sample_medians were notably stable and concentrated around 0. The sample median for Cauchy data exhibits a **finite asymptotic variance**, establishing it as a **robust and highly efficient estimator** that successfully captures central tendency where the mean completely fails.

**Implications for Estimator Efficiency and Robustness:**

*   **Efficiency** refers to how precisely an estimator approximates the true parameter, typically inversely related to its variance.
*   **Robustness** denotes an estimator's insensitivity to deviations from underlying assumptions, such as outliers or heavy tails. The **sample mean is highly susceptible** to these issues, leading to inefficiency or undefined behavior. The **sample median is robust** as it considers only the rank of data points. **M-estimators** offer a tunable compromise, balancing efficiency and robustness.

**Conclusion on Estimator Preference:**

*   **Normal-like (thin-tailed) data:** The **sample mean** is generally the most efficient estimator.
*   **Laplace-like (moderately heavy-tailed) data:** The **sample median** or robust M-estimators are generally more efficient and robust than the sample mean.
*   **Cauchy-like (very heavy-tailed) data:** The **sample median** or other highly robust estimators are essential, as the sample mean is entirely unreliable due to the absence of a finite first moment.
