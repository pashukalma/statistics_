### Linear regression

Let $(X, Y)$ be a pair of real-valued random variables such that, there exist two real numbers $a^*$ and $b^*$, for $Y = a^* + b^* X + \epsilon$, where $\epsilon \sim N(0,1)$ and $\epsilon$ is independent of $X$. Assume that $X$ is not deterministic, and note by $\mu$ its expectation and by $v$ its variance, with $v > 0$.

Let $(X_1, Y_1), \dots, (X_n, Y_n)$ be a sample of $n$ i.i.d observed copies of $(X, Y)$. We estimate $a^*, b^*$ using this sample. We abbreviate $\rho(x,y,a,b) = [y - (a +bx)]^2$, for $x, y, a, b \in \mathbb{R}$.

**The expected value of** $\rho(x,y,a,b)$ is
- $ E[\rho(X, Y, a, b)] = (a^* - a)^2 + (b^* - b)^2 (v + \mu^2) + 1 + 2(a^* - a)(b^* - b)\mu$

To find the minimizers, we take the partial derivatives with respect to $a$ and $b$ and set them to zero.

$\frac{\partial E[\rho]}{\partial a} = -2(a^* - a) + 2(b^* - b)\mu = 0$ ; $\frac{\partial E[\rho]}{\partial b} = -2(b^* - b)(v + \mu^2) + 2(a^* - a)\mu = 0$

Solving these two equations for $a$ and $b$ yields
- $a = a^*$, $b = b^*$

Thus, the minimizers of $E[\rho(x,y,a,b)]$ are $a = a^*$ and $b = b^*$. This is an expected result, as the true parameters minimize the expected squared error.

Given $\rho(x,y,a,b) = [y - (a + bx)]^2$.

**First Partial Derivatives of $\rho(x,y,a,b)$**
- $\frac{\partial \rho}{\partial a} = -2(y - a - bx)$ ; $\frac{\partial \rho}{\partial b} = -2x(y - a - bx)$

**Evaluation at true parameters** ($a^*, b^*$)

Substitute $y = a^* + b^*x + \epsilon$ into the derivatives:
- $\frac{\partial \rho}{\partial a} \Big|_{(a^*,b^*)} = -2(a^* + b^*x + \epsilon - a^* - b^*x) = -2\epsilon$
- $\frac{\partial \rho}{\partial b} \Big|_{(a^*,b^*)} = -2x(a^* + b^*x + \epsilon - a^* - b^*x) = -2x\epsilon$

**Variance of First Partial Derivatives**
- $Var(\frac{\partial \rho}{\partial a} \Big|_{(a^*,b^*)})$: $Var(-2\epsilon) = (-2)^2 Var(\epsilon) = 4 \times 1 = 4$
- $Var(\frac{\partial \rho}{\partial b} \Big|_{(a^*,b^*)})$: $Var(-2x\epsilon) = (-2)^2 Var(X\epsilon) = 4 [E[X^2\epsilon^2] - (E[X\epsilon])^2]$

Since $X$ and $\epsilon$ are independent, $E[X\epsilon] = E[X]E[\epsilon] = \mu \times 0 = 0$. And $E[X^2\epsilon^2] = E[X^2]E[\epsilon^2] = (v + \mu^2) \times 1 = v + \mu^2$. Therefore, $Var(-2x\epsilon) = 4(v + \mu^2)$.

**Second Partial Derivatives **bold text** of $\rho(x,y,a,b)$:**
- $\frac{\partial^2 \rho}{\partial a^2} = \frac{\partial}{\partial a}(-2(y - a - bx)) = -2(-1) = 2$
- $\frac{\partial^2 \rho}{\partial b^2} = \frac{\partial}{\partial b}(-2x(y - a - bx)) = -2x(-x) = 2x^2$
- $\frac{\partial^2 \rho}{\partial a \partial b} = \frac{\partial}{\partial b}(-2(y - a - bx)) = -2(-x) = 2x$

**Expectation of Second Partial Derivatives (evaluated at true parameters):**
- **$E[\frac{\partial^2 \rho}{\partial a^2} \Big|_{(a^*,b^*)}]$** The second derivative is $2$, so $E[2] = 2$.
- **$E[\frac{\partial^2 \rho}{\partial b^2} \Big|_{(a^*,b^*)}]$** The second derivative is $2x^2$, so $E[2X^2] = 2E[X^2] = 2(v + \mu^2)$.
- **$E[\frac{\partial^2 \rho}{\partial a \partial b} \Big|_{(a^*,b^*)}]$** The second derivative is $2x$, so $E[2X] = 2E[X] = 2\mu$.

Let the true model be $Y_i = a^* + b^* X_i + \epsilon_i$, where $\epsilon_i \sim N(0,1)$ and are independent of $X_i$.

The Least Squares Estimator (LSE) of $(a^*, b^*)$ is defined as the minimizer of
- $ \rho_n(a, b) = \sum_{i=1}^n [Y_i - (a + bX_i)]^2$
with respect to $a$ and $b$.

To find the minimizers $(a_\text{hat}, b_\text{hat})$, we take the partial derivatives of $\rho_n(a, b)$ with respect to $a$ and $b$ and set them to zero.

$\frac{\partial \rho_n}{\partial a} = \sum_{i=1}^n 2[Y_i - (a + bX_i)](-1) = 0$
$\sum_{i=1}^n Y_i - na - b\sum_{i=1}^n X_i = 0 \quad \implies \quad \bar{Y} - a - b\bar{X} = 0 \quad (1)$

$\frac{\partial \rho_n}{\partial b} = \sum_{i=1}^n 2[Y_i - (a + bX_i)](-X_i) = 0$
$\sum_{i=1}^n Y_i X_i - a\sum_{i=1}^n X_i - b\sum_{i=1}^n X_i^2 = 0 \quad \implies \quad \overline{XY} - a\bar{X} - b\overline{X^2} = 0 \quad (2)$

From (1), we have $a = \bar{Y} - b\bar{X}$. Substitute this into (2):
$\overline{XY} - (\bar{Y} - b\bar{X})\bar{X} - b\overline{X^2} = 0$
$\overline{XY} - \bar{Y}\bar{X} + b\bar{X}^2 - b\overline{X^2} = 0$
$b(\overline{X^2} - \bar{X}^2) = \overline{XY} - \bar{Y}\bar{X}$
- $ b_\text{hat} = \frac{\overline{XY} - \bar{X}\bar{Y}}{\overline{X^2} - \bar{X}^2} = \frac{\text{Cov}(X,Y)}{\text{Var}(X)}$

Substitute $b_\text{hat}$ back into the expression for $a$:
- $ a_\text{hat} = \bar{Y} - b_\text{hat}\bar{X} $

**Justification for Consistency and Asymptotic Normality:**

Under standard assumptions for linear regression (e.g., Gauss-Markov assumptions), the Least Squares Estimators $(a_\text{hat}, b_\text{hat})$ possess desirable statistical properties:

1.  **Consistency**: The estimators are consistent, meaning that as the sample size $n \to \infty$, the estimators converge in probability to the true parameters:
- $a_\text{hat} \xrightarrow{P} a^*$ ; $b_\text{hat} \xrightarrow{P} b^*$. This is because $\bar{X} \xrightarrow{P} E[X] = \mu$, $\bar{Y} \xrightarrow{P} E[Y] = a^* + b^*\mu$, $\overline{X^2} \xrightarrow{P} E[X^2] = v + \mu^2$, and $\overline{XY} \xrightarrow{P} E[XY] = E[X(a^* + b^*X + \epsilon)] = a^*\mu + b^*E[X^2] = a^*\mu + b^*(v + \mu^2)$. These convergences, combined with the continuous mapping theorem, ensure the consistency of $a_\text{hat}$ and $b_\text{hat}$.

2.  **Asymptotic Normality**: The estimators are asymptotically normal, meaning that as $n \to \infty$, the distribution of the estimators, properly scaled, approaches a normal distribution
- $ \sqrt{n} \begin{pmatrix} a_\text{hat} - a^* \\ b_\text{hat} - b^* \end{pmatrix} \xrightarrow{D} N\left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \Sigma \right)$, where $\Sigma$ is the asymptotic covariance matrix.

**Asymptotic Covariance Matrix:**

The asymptotic covariance matrix $\Sigma$ for the OLS estimators in a simple linear regression model $Y_i = a^* + b^* X_i + \epsilon_i$ with $E[\epsilon_i|X_i]=0$ and $Var(\epsilon_i|X_i)=\sigma^2$ (in our case, $\sigma^2 = Var(\epsilon) = 1$) is given by:
- $ \Sigma = \frac{\sigma^2}{E[X^2] - (E[X])^2} \begin{pmatrix} E[X^2] & -E[X] \\ -E[X] & 1 \end{pmatrix} = \frac{1}{v} \begin{pmatrix} v + \mu^2 & -\mu \\ -\mu & 1 \end{pmatrix}$

Substituting $E[X] = \mu$ and $Var(X) = v$, so $E[X^2] = v + \mu^2$
- $ \Sigma = \frac{1}{v} \begin{pmatrix} v + \mu^2 & -\mu \\ -\mu & 1 \end{pmatrix} $

Note: The result for $E[\rho(X, Y, a, b)]$ minimized at $a=a^*$ and $b=b^*$ confirms that the population parameters are indeed the ones minimizing the expected squared error, which is the theoretical foundation for the LSE.
