#### **Maximum Likelihood Estimator**
Derive MLE, $\hat{\theta}$, for $\theta$ based on $n$ independent observations $X_1,...,X_n$ and the sample average $\bar{X}_n$.

Given the pdf for a single observation $X_i$, $f(x_i; \theta) = (-\ln \theta) e^{( \ln \theta) x_i}$
$L(\theta; x_1, \dots, x_n)$ is the product of the individual pdfs:
- $L(\theta; x_1, \dots, x_n) = \prod_{i=1}^n f(x_i; \theta) = \prod_{i=1}^n [(-\ln \theta) e^{( \ln \theta) x_i}] = (-\ln \theta)^n e^{( \ln \theta) \sum_{i=1}^n x_i}$

**The log-likelihood $l(\theta)$**
- $l(\theta) = \ln\left( (-\ln \theta)^n e^{( \ln \theta) \sum_{i=1}^n x_i} \right)  = \ln((-\ln \theta)^n) + \ln(e^{( \ln \theta) \sum_{i=1}^n x_i}) = n \ln(-\ln \theta) + (\ln \theta) \sum_{i=1}^n x_i$


**Differentiate the log-likelihood function**

$l(\theta)$ with respect to $\theta$ $\frac{dl}{d\theta} = \frac{d}{d\theta} [n \ln(-\ln \theta)] + \frac{d}{d\theta} [(\ln \theta) \sum_{i=1}^n x_i]$. Differentiate the first term: Using the chain rule, $\frac{d}{dx} \ln(f(x)) = \frac{f'(x)}{f(x)}$, and $\frac{d}{d\theta} (-\ln \theta) = -\frac{1}{\theta}$, $ n \cdot \frac{1}{(-\ln \theta)} \cdot \left(-\frac{1}{\theta}\right) = -\frac{n}{\theta \ln \theta} $. Differentiate the second term: Since $\sum_{i=1}^n x_i$ is a constant with respect to $\theta$, we have $ \frac{1}{\theta} \sum_{i=1}^n x_i $

Combining both terms: $\frac{dl}{d\theta} = -\frac{n}{\theta \ln \theta} + \frac{1}{\theta} \sum_{i=1}^n x_i$

**Set the derivative $\frac{dl}{d\theta}$** to zero and solve for $\theta$. This gives the **MLE, $\hat{\theta}$.** $-\frac{n}{\theta \ln \theta} + \frac{1}{\theta} \sum_{i=1}^n x_i = 0$. Multiply both sides by $\theta$ (since $\theta \neq 0$) $-\frac{n}{\ln \theta} + \sum_{i=1}^n x_i = 0$. Solving for $\theta$, we exponentiate both sides with base $e$, $\hat{\theta} = e^{\left(\frac{n}{\sum_{i=1}^n x_i}\right)}$

**MLE $\hat{\theta}$ in terms of the sample mean $\bar{X}_n = \frac{1}{n} \sum_{i=1}^n X_i$**. For MLE $\hat{\theta} = e^{\left(\frac{n}{\sum_{i=1}^n x_i}\right)}$, and $\bar{X}_n = \frac{1}{n} \sum_{i=1}^n X_i$, $\quad$  $\hat{\theta} = e^{\left(\frac{1}{\bar{X}_n}\right)}$.

Simplifying the exponent $\hat{\theta} = e^{\left(\frac{1}{\bar{X}_n}\right)}$.

#### **Asymptotic Distribution**

Asymptotic distribution of $\sqrt{n}(\hat{\theta} - \theta)$, involves calculating the Fisher Information.

**The log-likelihood function** for a single observation, $l(\theta; x)$, is given by $l(\theta; x) = \ln(-\ln \theta) + x \ln \theta$.

**The first Derivative of the Log-Likelihood**  $l'(\theta; x)$, with respect to $\theta$, we differentiate $l(\theta; x) = \ln(-\ln \theta) + x \ln \theta$:
- $l'(\theta; x) = \frac{d}{d\theta} [\ln(-\ln \theta)] + \frac{d}{d\theta} [x \ln \theta]$

Applying the chain rule for the first term and basic differentiation for the second:
- $l'(\theta; x) = \frac{1}{-\ln \theta} \cdot \frac{d}{d\theta}(-\ln \theta) + x \cdot \frac{1}{\theta}$ $ = \frac{1}{-\ln \theta} \cdot \left(-\frac{1}{\theta}\right) + \frac{x}{\theta}$ $ = \frac{1}{\theta \ln \theta} + \frac{x}{\theta}$ $ = \frac{1 + x \ln \theta}{\theta \ln \theta}$

**The second derivative of the log-likelihood**, $l''(\theta; x)$, with respect to $\theta$, we differentiate $l'(\theta; x) = \frac{1 + x \ln \theta}{\theta \ln \theta}$. We rewrite $l'(\theta; x)$ as
- $l'(\theta; x) = (\theta \ln \theta)^{-1} + x \theta^{-1}$.

Differentiate each term: First term $\frac{d}{d\theta} (\theta \ln \theta)^{-1}$. Using the chain rule and power rule, let $u = \theta \ln \theta$, then $\frac{d}{d\theta} u^{-1} = -1 \cdot u^{-2} \cdot \frac{du}{d\theta}$. $\frac{du}{d\theta} = \frac{d}{d\theta}(\theta \ln \theta) = 1 \cdot \ln \theta + \theta \cdot \frac{1}{\theta} = \ln \theta + 1$. So, $\frac{d}{d\theta} (\theta \ln \theta)^{-1} = -(\theta \ln \theta)^{-2} (\ln \theta + 1) = -\frac{\ln \theta + 1}{(\theta \ln \theta)^2}$. Second term $\frac{d}{d\theta} (x \theta^{-1}) = x \cdot (-1 \cdot \theta^{-2}) = -\frac{x}{\theta^2}$. Combining the terms $l''(\theta; x) = -\frac{\ln \theta + 1}{(\theta \ln \theta)^2} - \frac{x}{\theta^2}$
- $l''(\theta; x) = -\frac{1}{\theta^2} \left( \frac{\ln \theta + 1}{(\ln \theta)^2} + x \right)$

This second derivative is used to calculate the Fisher Information.

**Fisher Information**

For the Fisher Information $I(\theta)$,  use the $I(\theta) = -E[l''(\theta; X)]$. We have the second derivative as $l''(\theta; x) = -\frac{1}{\theta^2} \left( \frac{\ln \theta + 1}{(\ln \theta)^2} + x \right)$. The expectation with respect to $X$. Since $E[X] = -1/\ln \theta$, and we make the substitutions $E[l''(\theta; X)] = -\frac{1}{\theta^2 (\ln \theta)^2}$. The Fisher Information $I(\theta)$: $I(\theta) = -E[l''(\theta; X)] = -\left(-\frac{1}{\theta^2 (\ln \theta)^2}\right) = \frac{1}{\theta^2 (\ln \theta)^2}$

This is the Fisher Information for a single observation $I(\theta) = \frac{1}{\theta^2 (\ln \theta)^2}$

**Asymptotic Distribution** of $\sqrt{n}(\hat{\theta} - \theta)$

According to the properties of MLEs, the asymptotic distribution of $\sqrt{n}(\hat{\theta} - \theta)$ is given by $\sqrt{n}(\hat{\theta} - \theta) \xrightarrow{d} N(0, I(\theta)^{-1})$. The Fisher Information, is given by $I(\theta) = \frac{1}{\theta^2 (\ln \theta)^2}$. The inverse of the Fisher Information, $I(\theta)^{-1}$ $I(\theta)^{-1} = \left(\frac{1}{\theta^2 (\ln \theta)^2}\right)^{-1} = \theta^2 (\ln \theta)^2$.
The asymptotic distribution of $\sqrt{n}(\hat{\theta} - \theta)$ is Normal with mean 0 and variance $\theta^2 (\ln \theta)^2$
- $\sqrt{n}(\hat{\theta} - \theta) \xrightarrow{d} N(0, \theta^2 (\ln \theta)^2)$

#### Determine One-sided Confidence Interval $B$


1.  **Recall the asymptotic distribution of $\sqrt{n}(\hat{\theta} - \theta)$:**
From the CLT or the asymptotic properties of MLEs, we know that as $n \to \infty$, the sampling distribution of the MLE $\hat{\theta}$ is approximately normal.
Specifically, $\sqrt{n}(\hat{\theta} - \theta) \xrightarrow{D} N(0, I(\theta)^{-1})$, where $I(\theta)$ is the Fisher Information. Thus, the asymptotic variance of $\hat{\theta}$ is $\frac{I(\theta)^{-1}}{n}$. Let's denote this asymptotic variance as $\text{Var}_{asy}(\hat{\theta}) = \frac{\sigma^2}{n}$.

2.  **Standardize the expression $(\hat{\theta} - \theta)$:**
To relate $(\hat{\theta} - \theta)$ to a standard normal distribution, we standardize it using its asymptotic standard deviation:
$Z = \frac{\hat{\theta} - \theta}{\sqrt{\text{Var}_{asy}(\hat{\theta})}} = \frac{\hat{\theta} - \theta}{\sqrt{\sigma^2/n}} = \frac{\sqrt{n}(\hat{\theta} - \theta)}{\sigma}$.
As $n \to \infty$, $Z \xrightarrow{D} N(0, 1)$.

3.  **Use the 95% confidence level for a one-sided interval to find the appropriate critical z-value ($z_{\alpha}$):**
For a one-sided confidence interval of the form $(0, B]$ for $\theta$, we are interested in finding an upper bound $B$ such that $P(\theta \le B) = 0.95$. This corresponds to finding $z_{\alpha}$ such that $P(Z \le z_{\alpha}) = 0.95$. From standard normal tables or software, the critical z-value for a 95% upper one-sided interval is approximately $z_{0.05} = 1.645$.

4.  **Substitute the critical z-value into the standardized inequality and rearrange:**
We want to find $B$ such that $P\left(\frac{\sqrt{n}(\hat{\theta} - \theta)}{\sigma} \le z_{0.05}\right) = 0.95$.
Rearranging the inequality to isolate $\theta$:
$\frac{\sqrt{n}(\hat{\theta} - \theta)}{\sigma} \le z_{0.05}$
$\sqrt{n}(\hat{\theta} - \theta) \le z_{0.05} \sigma$
$\hat{\theta} - \theta \le \frac{z_{0.05} \sigma}{\sqrt{n}}$
$-\theta \le \frac{z_{0.05} \sigma}{\sqrt{n}} - \hat{\theta}$
$\theta \ge \hat{\theta} - \frac{z_{0.05} \sigma}{\sqrt{n}}$.
However, for an interval of the form $(0, B]$, we are looking for an upper bound. This means we are interested in a lower tail probability for Z, such that the probability of Z being *less than* a certain value is 0.95. If we use $P(Z \le z_{\alpha}) = 0.95$, we get:
$P\left(\frac{\sqrt{n}(\hat{\theta} - \theta)}{\sigma} \ge -z_{0.05}\right) = 0.95$ (Note: $P(Z \le z_{0.05}) = 0.95$ implies $P(Z \ge -z_{0.05}) = 0.95$).
Or, more directly, $P\left(\frac{\theta - \hat{\theta}}{\sqrt{\sigma^2/n}} \le z_{0.05}\right) = 0.95$.
This gives us:
$\frac{\theta - \hat{\theta}}{\sqrt{\sigma^2/n}} \le z_{0.05}$
$\theta - \hat{\theta} \le z_{0.05} \sqrt{\frac{\sigma^2}{n}}$
$\theta \le \hat{\theta} + z_{0.05} \sqrt{\frac{\sigma^2}{n}}$
So, the upper bound $B$ is $\hat{\theta} + z_{0.05} \sqrt{\frac{\sigma^2}{n}}$.

5.  **Apply the plug-in method:**
Replace the true parameter $\theta$ in the asymptotic variance $\sigma^2$ with its MLE $\hat{\theta}$. This gives the estimated asymptotic variance, $\widehat{\text{Var}}_{asy}(\hat{\theta}) = \frac{\hat{\sigma}^2}{n}$, where $\hat{\sigma}^2$ is the estimate of the variance parameter (often based on $\hat{\theta}$).

6.  **Express B in terms of n, $\hat{\theta}$, and the estimated asymptotic variance:**
Using the plug-in method, the estimated asymptotic standard deviation is $\sqrt{\widehat{\text{Var}}_{asy}(\hat{\theta})}$.
Therefore, the expression for B is:
$B = \hat{\theta} + z_{0.05} \sqrt{\widehat{\text{Var}}_{asy}(\hat{\theta})}$
    where:
    - $\hat{\theta}$ is the Maximum Likelihood Estimator of $\theta$.
    - $z_{0.05} \approx 1.645$ is the critical z-value for a 95% one-sided confidence interval (upper tail).
    - $\widehat{\text{Var}}_{asy}(\hat{\theta})$ is the estimated asymptotic variance of $\hat{\theta}$, which can be written as $\frac{\hat{\sigma}^2}{n}$, where $\hat{\sigma}^2$ is the estimated variance parameter from the Fisher Information or similar.

This expression defines the upper bound B for the one-sided confidence interval $(0, B]$ for $\theta$ at a 95% asymptotic confidence level, using the plug-in method.

### PART II: Bayesian Statistical Analysis


#### 1. Introduction to Bayesian Statistics

In contrast to the Frequentist approach, which views probability as the long-run frequency of events and parameters as fixed but unknown values, Bayesian statistics offers a different paradigm. Bayesian inference treats probability as a measure of belief or uncertainty. This allows for the incorporation of prior knowledge or beliefs about a parameter into the analysis, which are then updated with observed data to form a posterior belief.

For our specific problem, where $X_i \sim \text{Exp}(-\ln \theta)$ for $0 < \theta < 1$, the Frequentist analysis in Part I treated $\theta$ as a fixed, unknown constant. In a Bayesian framework, we would treat $\theta$ itself as a random variable, and our goal would be to infer its posterior distribution.

**Key Components:**

*   **Prior Distribution ($P(\theta)$):** This represents our initial belief or knowledge about the parameter $\theta$ before observing any data. For our exponential distribution, this would be a probability distribution over the possible values of $\theta \in (0, 1)$. It can be based on historical data, expert opinion, or theoretical considerations. The prior can be *informative* (reflecting strong prior beliefs) or *non-informative* (representing a lack of strong prior beliefs).

*   **Likelihood Function ($P(X|\theta)$):** This is the same likelihood function used in Frequentist statistics. For our problem, given $n$ i.i.d. observations $X_1, \dots, X_n$, the likelihood function for $\theta$ is based on the exponential PDF $f(x; \theta) = (-\ln \theta) e^{(\ln \theta) x}$. Specifically, the likelihood is $L(\theta; x_1, \dots, x_n) = \prod_{i=1}^n f(x_i; \theta)$. It quantifies how well a particular parameter value explains the observed data.

*   **Posterior Distribution ($P(\theta|X)$):** This is the central output of Bayesian inference. It represents our updated belief about the parameter $\theta$ *after* observing the data $X$. It is proportional to the product of the prior and the likelihood, as dictated by Bayes' Theorem.

Bayes' Theorem provides the mathematical framework for updating our beliefs: $ P(\theta|X) = \frac{P(X|\theta) P(\theta)}{P(X)} $.

#### 2. Bayes' Rule for Posterior Calculation

- $P(\theta|X)$, **Posterior Probability**, the probability of the parameter $\theta$ given the observed data $X$, for our exponential distribution's $\theta$.
- $P(X|\theta)$, the **Likelihood**, the probability of observing the data $X$ given a specific parameter $\theta$. For our exponential distribution, with $n$ i.i.d. observations $X_1, \dots, X_n$, this is given by:
$L(\theta; x_1, \dots, x_n) = \prod_{i=1}^n [(-\ln \theta) e^{( \ln \theta) x_i}] = (-\ln \theta)^n e^{( \ln \theta) \sum_{i=1}^n x_i}$
*   $P(\theta)$, the **Prior Probability**, the initial probability distribution reflecting our beliefs about $\theta \in (0, 1)$ before observing the sample data.
*   $P(X)$, the **Marginal Likelihood (or Evidence)**, the total probability of data $X$ over all possible values of $\theta$. It acts as a normalizing constant to ensure the posterior distribution integrates to 1. For our continuous parameter $\theta$, it would be $P(X) = \int_{0}^{1} P(X|\theta) P(\theta) d\theta$.

**Application in Parameter Estimation for our Exponential Problem:**

To estimate $\theta$ for our exponential distribution, we would choose a prior distribution $P(\theta)$ that represents our initial beliefs about $\theta \in (0,1)$. We would then combine this prior with the likelihood function derived above, using Bayes' Rule to obtain the posterior distribution $P(\theta|X)$. This posterior distribution encapsulates all available information about $\theta$ – both from our prior beliefs and the observed data from the exponential process. The goal is then to summarize this posterior distribution to obtain a point estimate (e.g., posterior mean, median, or MAP) or an interval estimate (credible interval) for $\theta$.

#### 3. Frequentist vs. Bayesian Perspectives

The fundamental differences between Frequentist and Bayesian statistics stem from their distinct interpretations of probability and how they treat unknown parameters.

| Feature              | Frequentist Perspective                                 | Bayesian Perspective                                    |
| :------------------- | :------------------------------------------------------ | :------------------------------------------------------ |
| **Definition of Probability** | Long-run frequency of events in repeated trials.        | Degree of belief or certainty about an event.           |
| **Treatment of Parameters** | For our problem, $\theta$ is a fixed, but unknown constant value (e.g., the true $\theta$ of the exponential distribution).           | For our problem, $\theta$ is a random variable with a probability distribution (our prior and posterior distributions for $\theta$). |
| **Inference**        | Focus on properties of estimators over repeated samples. We found the MLE $\hat{\theta} = e^{\left(\frac{1}{\bar{X}_n}\right)}$ and its asymptotic properties. | Focus on updating beliefs about parameters given data, leading to a posterior distribution for $\theta$. |
| **Prior Information** | Not directly incorporated into the statistical model.   | Essential part of the model, explicitly incorporated, e.g., choosing a prior $P(\theta)$ for our exponential parameter. |
| **Conclusions**      | P-values, confidence intervals (about the procedure). For example, the 95% confidence interval for $\theta$ (0, B] we derived.   | Posterior distributions, credible intervals (about the parameter itself). A 95% credible interval would directly state the probability that $\theta$ lies within that range. |
| **Example (Confidence Interval)** | "If we repeat this procedure many times, 95% of the intervals constructed would contain the true parameter value." | "There is a 95% probability that the true parameter value lies within this interval." |

**Key Distinctions:**

*   **Probability:** Frequentists see probability as objective, rooted in repeatable experiments. Bayesians see it as subjective, a measure of knowledge or ignorance.
*   **Parameters:** For Frequentists, $\theta$ is a single, fixed value we try to estimate. For Bayesians, $\theta$ has a distribution, reflecting our uncertainty about its true value.
*   **Prior Information:** Bayesian methods explicitly leverage prior knowledge, which can be particularly advantageous with small datasets. Frequentist methods do not directly use prior information in the same way, though it might implicitly influence study design.
*   **Interpretation:** Bayesian credible intervals are often more intuitive to interpret than Frequentist confidence intervals because they make direct probability statements about the parameter.

#### 4. Bayesian Estimator and Estimation

In Bayesian statistics, once the posterior distribution $P(\theta|X)$ is obtained, we need a way to summarize this distribution into a single point estimate for the parameter $\theta$. This summary is known as a Bayesian estimator.

A Bayesian estimator is a function of the observed data that minimizes the expected value of a loss function, where the expectation is taken with respect to the posterior distribution.

Common Types of Bayesian Estimators for our Exponential $\theta$:
1.  **Posterior Mean** (Minimum Mean Squared Error Estimator):
The most commonly used Bayesian estimator. It is the mean of the posterior distribution for $\theta$:
$ \hat{\theta}_{Mean} = E[\theta|X] = \int_{0}^{1} \theta P(\theta|X) d\theta $.
This estimator minimizes the MSE, meaning it is optimal when the cost of an estimation error increases quadratically with the size of the error.
2.  **Posterior Median** (Minimum Absolute Error Estimator):
The median of the posterior distribution for $\theta$. It minimizes the expected absolute error:
$ P(\theta \le \hat{\theta}_{Median}|X) = 0.5 $.
This estimator is more robust to outliers or skewed posterior distributions than the posterior mean.
3.  **Maximum A Posteriori (MAP) Estimator:**
The mode of the posterior distribution for $\theta$. It is the value of $\theta$ that maximizes $P(\theta|X)$:
$ \hat{\theta}_{MAP} = \operatorname{argmax}_{\theta} P(\theta|X)$.
The MAP estimator is computationally simpler to find in many cases and often coincides with the MLE if the prior is uniform (non-informative) over the range $(0,1)$ for $\theta$.

Estimation Process for our Exponential $\theta$:
1.  **Likelihood:** We have already established the likelihood function for our exponential distribution: $L(\theta; x_1, \dots, x_n) = (-\ln \theta)^n e^{( \ln \theta) \sum_{i=1}^n x_i}$.
2.  **Choose a Prior:** We would select a prior distribution $P(\theta)$ for $\theta \in (0, 1)$. For instance, a simple non-informative choice might be a Uniform(0,1) prior. More complex conjugate priors could also be considered if we transform $\theta$ into its rate parameter $\lambda = -\ln \theta$, for which a Gamma prior is conjugate.
3.  **Calculate the Posterior:** Apply Bayes' Rule: $P(\theta|X) \propto L(\theta; X)P(\theta)$. We would then need to normalize this product to get the full posterior distribution.
4.  **Derive the Estimator:** Based on our chosen loss function, we would calculate the corresponding Bayesian estimator (e.g., posterior mean, median, MAP) from the obtained posterior distribution for $\theta$.

**Advantages of Bayesian Estimators:**
- Incorporates Prior Knowledge: Allows for the formal inclusion of previous information or expert opinion about $\theta$.
- Provides Full Distribution: Yields a complete posterior distribution of $\theta$, offering a richer understanding of uncertainty about its true value.
- Direct Probability Statements: Allows for direct probability statements about $\theta$ (e.g., "the probability that $\theta$ is between 0.5 and 0.7 is 90%").
- Handles Small Samples: Can perform well with small sample sizes, especially when informative priors for $\theta$ are available.

**Disadvantages of Bayesian Estimators:**
- Choice of Prior: The choice of prior for $\theta$ can be subjective and can influence the posterior, especially with limited data.
- Computational Intensity: Calculating the posterior distribution for $\theta$, especially the normalizing constant $P(X)$, can be computationally intensive for complex models, often requiring numerical methods like Markov Chain Monte Carlo (MCMC).
- Interpretation of Priors: Communicating the justification and impact of prior choices for $\theta$ can sometimes be challenging.

#### 5. Bayesian Posterior Derivations and Estimators

To perform Bayesian inference for our exponential distribution, we need to select a prior distribution for our parameter and then combine it with the likelihood function using Bayes' Rule to obtain the posterior distribution. For simplicity and analytical tractability, we will work with the rate parameter $\lambda = -\ln \theta$ directly, as the exponential distribution's rate parameter has a well-known conjugate prior: the Gamma distribution. We later transform our results back to $\theta$.

**Recall the Likelihood Function (in terms of $\lambda$):**

For $n$ i.i.d. observations $X_1, \dots, X_n$ from an exponential distribution with rate $\lambda$, the likelihood function is
- $ L(\lambda; x_1, \dots, x_n) = \prod_{i=1}^n \lambda e^{-\lambda x_i} = \lambda^n e^{-\lambda \sum_{i=1}^n x_i} $.

**Choice of Prior Distribution for $\lambda$:**

A common choice for the prior distribution of an exponential rate parameter $\lambda$ is the **Gamma distribution**, which is a conjugate prior. A Gamma distribution with shape parameter $\alpha_0$ and rate parameter $\beta_0$ is defined as:
- $ P(\lambda) = \frac{\beta_0^{\alpha_0}}{\Gamma(\alpha_0)} \lambda^{\alpha_0-1} e^{-\beta_0 \lambda} \quad \text{for } \lambda > 0, \alpha_0 > 0, \beta_0 > 0 $

Here, $\alpha_0$ and $\beta_0$ are hyperparameters that reflect our prior beliefs about $\lambda$. For instance, a small $\alpha_0$ and $\beta_0$ (e.g., $\alpha_0=1, \beta_0=0.01$) can represent a relatively non-informative prior.

**Derivation of the Posterior Distribution for $\lambda$:**

Using Bayes' Rule, the posterior distribution $P(\lambda|X)$ is proportional to the product of the likelihood and the prior
$ P(\lambda|X) \propto L(\lambda; X) P(\lambda) $.
Substitute the likelihood and prior expressions:
$ P(\lambda|X) \propto \left( \lambda^n e^{-\lambda \sum x_i} \right) \left( \lambda^{\alpha_0-1} e^{-\beta_0 \lambda} \right) $.
Combine the terms involving $\lambda$:
$ P(\lambda|X) \propto \lambda^{n + \alpha_0 - 1} e^{-\lambda (\sum x_i + \beta_0)} $.
This functional form is recognizable as a **Gamma distribution** with updated parameters. Let's denote the new shape parameter as $\alpha_1$ and the new rate parameter as $\beta_1$:
$ \alpha_1 = \alpha_0 + n $$, $ \beta_1 = \beta_0 + \sum_{i=1}^n x_i $.

The posterior distribution for $\lambda$ is also a Gamma distribution
- $ \lambda|X \sim \text{Gamma}(\alpha_0 + n, \beta_0 + \sum_{i=1}^n x_i) $.

This demonstrates the conjugacy of the Gamma prior for the exponential likelihood. The posterior distribution summarizes our updated beliefs about $\lambda$ after observing the data.

**Bayesian Estimator for $\lambda$ (Posterior Mean):**

A common Bayesian estimator is the posterior mean. For a random variable $Y \sim \text{Gamma}(\alpha, \beta)$, the mean is $E[Y] = \frac{\alpha}{\beta}$.

Therefore, the posterior mean for $\lambda$ is:
- $ \hat{\lambda}_{Bayes} = E[\lambda|X] = \frac{\alpha_1}{\beta_1} = \frac{\alpha_0 + n}{\beta_0 + \sum_{i=1}^n x_i} $.

This estimator is a weighted average of the prior mean (if we consider $E[\lambda_{prior}] = \frac{\alpha_0}{\beta_0}$) and the data-driven estimate ($n / \sum x_i$, which is the MLE for $\lambda$). As $n$ becomes large, the data terms dominate, and the Bayesian estimator converges to the MLE.

**Translating back to $\theta$:**

Since $\lambda = -\ln \theta$, we have $\theta = e^{-\lambda}$.

While finding the exact posterior distribution for $\theta$ (which would be a transformed Gamma distribution) can be complex, we can use the derived posterior for $\lambda$ to make inferences about $\theta$. For example, a point estimate for $\theta$ could be obtained by transforming the posterior mean of $\lambda$:
- $ \hat{\theta}_{Bayes} = e^{-\hat{\lambda}_{Bayes}} = e^{-\left(\frac{\alpha_0 + n}{\beta_0 + \sum_{i=1}^n x_i}\right)} $.

This simple transformation is valid for the MAP estimator under certain conditions, but for the posterior mean, it is a point estimate of $\theta$ but not necessarily the mean of the posterior of $\theta$. More accurately, if we wanted the posterior mean of $\theta$, we would need to calculate $E[e^{-\lambda}|X]$, which involves an integral with the posterior Gamma distribution. However, for practical purposes and as an approximation, using $e^{-\hat{\lambda}_{Bayes}}$ is often sufficient.

**Exact Posterior Mean of $\theta = e^{-\lambda}$**

While $e^{-\hat{\lambda}_{Bayes}}$ provides a point estimate for $\theta$, it is not the true posterior mean $E[\theta|X]$. To find the exact posterior mean of $\theta$, we need to calculate $E[e^{-\lambda}|X]$ by integrating over the posterior distribution of $\lambda$.

Recall that the posterior distribution of $\lambda$ is $\text{Gamma}(\alpha_1, \beta_1)$, where $\alpha_1 = \alpha_0 + n$ and $\beta_1 = \beta_0 + \sum_{i=1}^n x_i$. The PDF of this Gamma distribution is:
- $ f(\lambda|X) = \frac{\beta_1^{\alpha_1}}{\Gamma(\alpha_1)} \lambda^{\alpha_1-1} e^{-\beta_1 \lambda} \quad \text{for } \lambda > 0 $.

To find the posterior mean of $\theta = e^{-\lambda}$, we compute  $ E[\theta|X] = E[e^{-\lambda}|X] = \int_{0}^{\infty} e^{-\lambda} f(\lambda|X) d\lambda $. Substitute the PDF of the Gamma distribution $ E[e^{-\lambda}|X] = \int_{0}^{\infty} e^{-\lambda} \frac{\beta_1^{\alpha_1}}{\Gamma(\alpha_1)} \lambda^{\alpha_1-1} e^{-\beta_1 \lambda} d\lambda $.
Combine the exponential terms $ E[e^{-\lambda}|X] = \frac{\beta_1^{\alpha_1}}{\Gamma(\alpha_1)} \int_{0}^{\infty} \lambda^{\alpha_1-1} e^{-(\beta_1+1) \lambda} d\lambda $.
Recognize that the integral part $\int_{0}^{\infty} \lambda^{\alpha_1-1} e^{-(\beta_1+1) \lambda} d\lambda$ is the kernel of a Gamma distribution with shape parameter $\alpha_1$ and rate parameter $(\beta_1+1)$. We know that the integral of a Gamma PDF over its support is 1, so $\int_{0}^{\infty} \frac{(\beta+1)^{\alpha}}{\Gamma(\alpha)} x^{\alpha-1} e^{-(\beta+1)x} dx = 1$. This implies that $\int_{0}^{\infty} x^{\alpha-1} e^{-(\beta+1)x} dx = \frac{\Gamma(\alpha)}{(\beta+1)^{\alpha}}$.
Applying this to our integral, with $\alpha = \alpha_1$ and $\beta = \beta_1$,
$ \int_{0}^{\infty} \lambda^{\alpha_1-1} e^{-(\beta_1+1) \lambda} d\lambda = \frac{\Gamma(\alpha_1)}{(\beta_1+1)^{\alpha_1}} $.
Substitute this back into the expression for $E[e^{-\lambda}|X]$: $ E[e^{-\lambda}|X] = \frac{\beta_1^{\alpha_1}}{\Gamma(\alpha_1)} \cdot \frac{\Gamma(\alpha_1)}{(\beta_1+1)^{\alpha_1}} $.
- $ E[e^{-\lambda}|X] = \frac{\beta_1^{\alpha_1}}{(\beta_1+1)^{\alpha_1}} = \left(\frac{\beta_1}{\beta_1+1}\right)^{\alpha_1} $.

Therefore, the exact posterior mean of $\theta$ is:
- $ \hat{\theta}_{Bayes, ExactMean} = \left(\frac{\beta_0 + \sum_{i=1}^n x_i}{\beta_0 + \sum_{i=1}^n x_i + 1}\right)^{\alpha_0 + n} $.

#### **6. MLE Asymptotic Properties for $\lambda$**

To effectively compare the Bayesian estimator with the Frequentist MLE, we should also consider the asymptotic properties of the MLE for the rate parameter $\lambda$. Recall that $\lambda = -\ln \theta$, and the MLE for $\theta$ was $\hat{\theta}_{MLE} = e^{\left(\frac{1}{\bar{X}_n}\right)}$.

First, let's find the MLE for $\lambda$. Since $\lambda = -\ln \theta$, then $\theta = e^{-\lambda}$. Substituting this into the PDF of $X$: $ f(x; \lambda) = \lambda e^{-\lambda x} \quad \text{for } x > 0 $.
The likelihood for $n$ i.i.d. observations is: $ L(\lambda; x_1, \dots, x_n) = \prod_{i=1}^n \lambda e^{-\lambda x_i} = \lambda^n e^{-\lambda \sum_{i=1}^n x_i} $.
The log-likelihood is: $ l(\lambda) = n \ln \lambda - \lambda \sum_{i=1}^n x_i $.
Differentiating with respect to $\lambda$ and setting to zero:
$ \frac{dl}{d\lambda} = \frac{n}{\lambda} - \sum_{i=1}^n x_i = 0 $.
Solving for $\lambda$, we get the MLE for $\lambda$:
$ \hat{\lambda}_{MLE} = \frac{n}{\sum_{i=1}^n x_i} = \frac{1}{\bar{X}_n} $.

Find its asymptotic properties. The Fisher Information for $\lambda$ is $I(\lambda) = -E\left[\frac{d^2 l}{d\lambda^2}\right]$. $ \frac{d^2 l}{d\lambda^2} = -\frac{n}{\lambda^2} $.
So, for a single observation, the second derivative of the log-likelihood is $-1/\lambda^2$. Therefore, the Fisher Information for one observation is: $ I(\lambda) = -E\left[-\frac{1}{\lambda^2}\right] = \frac{1}{\lambda^2} $.
According to the asymptotic properties of MLEs, $\sqrt{n}(\hat{\lambda}_{MLE} - \lambda)$ is asymptotically normally distributed:
$ \sqrt{n}(\hat{\lambda}_{MLE} - \lambda) \xrightarrow{d} N(0, I(\lambda)^{-1}) $.


Thus, the asymptotic distribution of $\sqrt{n}(\hat{\lambda}_{MLE} - \lambda)$ is $N(0, \lambda^2)$. This means that for large $n$, $\hat{\lambda}_{MLE}$ is approximately normally distributed with mean $\lambda$ and variance $\frac{\lambda^2}{n}$.

#### 7. Comparison of Estimator Properties: MLE vs. Bayesian Estimator for $\lambda$

Let's summarize and compare the properties of the Maximum Likelihood Estimator (MLE) and the Bayesian Estimator (Posterior Mean) for the rate parameter $\lambda$ of our exponential distribution.

| Property           | Maximum Likelihood Estimator (MLE) for $\lambda$                           | Bayesian Estimator (Posterior Mean) for $\lambda$                                 |
| :----------------- | :---------------------------------------------------------------------- | :-------------------------------------------------------------------------------- |
| **Form**           | $\hat{\lambda}_{MLE} = \frac{1}{\bar{X}_n}$                                   | $\hat{\lambda}_{Bayes} = \frac{\alpha_0 + n}{\beta_0 + n\bar{X}_n}$                                        |
| **Asymptotic Unbiasedness** | Asymptotically unbiased.                                                | Asymptotically unbiased.                                                          |
| **Consistency**    | Consistent (converges to true $\lambda$ as $n \to \infty$).                 | Consistent (converges to true $\lambda$ as $n \to \infty$, assuming reasonable prior). |
| **Asymptotic Normality** | $\sqrt{n}(\hat{\lambda}_{MLE} - \lambda) \xrightarrow{d} N(0, \lambda^2)$      | $\sqrt{n}(\hat{\lambda}_{Bayes} - \lambda) \xrightarrow{d} N(0, \lambda^2)$ (under regularity conditions) |
| **Asymptotic Efficiency** | Asymptotically efficient (achieves Cramér-Rao lower bound).           | Asymptotically efficient (achieves Cramér-Rao lower bound, under regularity conditions). |
| **Small Sample Behavior** | Can be biased, but generally performs well.                             | Influenced by prior; can be more robust or less variable with informative prior. |
| **Prior Information** | Does not incorporate prior beliefs directly.                            | Explicitly incorporates prior beliefs via hyperparameters $\alpha_0, \beta_0$.       |

**Comparison of Asymptotic Properties:**

For large sample sizes ($n \to \infty$), the properties of the Bayesian estimator (specifically the posterior mean or mode) tend to converge to those of the MLE, provided that the prior distribution for $\lambda$ is not overly restrictive or assigned zero probability to the true parameter value. This is a crucial result in statistical theory.

*   **Consistency:** Both the MLE and the Bayesian posterior mean for $\lambda$ are consistent estimators. As the sample size increases, both estimators will converge in probability to the true value of $\lambda$.
*   **Asymptotic Normality:** Under standard regularity conditions, both estimators are asymptotically normally distributed with the same asymptotic variance, which is the inverse of the Fisher Information ($I(\lambda)^{-1} = \lambda^2$). This means that for large datasets, the sampling distribution of both estimators will be approximately normal around the true parameter value, with variance shrinking at a rate of $1/n$.
*   **Asymptotic Efficiency:** Both are asymptotically efficient, meaning they achieve the Cramér-Rao lower bound in the limit of large samples. This implies that no other consistent and asymptotically normal estimator can have a smaller asymptotic variance.

**Key takeaway:** While the methodologies and philosophical underpinnings of Frequentist and Bayesian approaches differ significantly, their primary point estimators (like MLE and posterior mean) often exhibit similar desirable properties (consistency, asymptotic normality, efficiency) when the sample size is large. The main difference lies in their interpretation and their behavior with small sample sizes, where the prior's influence in the Bayesian approach can be substantial and potentially beneficial.

#### Impact of Prior Hyperparameters on Bayesian Estimator's Convergence

The choice of prior hyperparameters significantly impacts the Bayesian estimator's convergence, particularly in small sample sizes. Here's a breakdown:

*   **Small Sample Sizes:** When you have limited data, the prior distribution exerts a stronger influence on the posterior distribution and, consequently, on the Bayesian estimator. If you choose a very **informative prior** (e.g., small variance, strong belief in a particular range for the parameter), the posterior will heavily lean towards that prior belief, even if the observed data suggests something different. This can lead to slower convergence to the true parameter value if the prior was inaccurate. Conversely, a good informative prior can improve the estimator's performance and convergence speed if it accurately reflects the true parameter.

*   **Large Sample Sizes:** As the sample size ($n$) increases, the likelihood function (which is derived from the observed data) starts to dominate the prior distribution. The data contains more information about the true parameter, and its influence outweighs that of the initial prior beliefs. In this scenario, the posterior distribution will become increasingly concentrated around the true parameter value, regardless of the initial choice of a reasonable prior. This means that for very large sample sizes, the Bayesian estimator will generally converge to the same value as the Maximum Likelihood Estimator (MLE), and the impact of the prior hyperparameters on convergence becomes negligible.

*   **Non-informative Priors:** When little is known about the parameter *a priori*, researchers often opt for **non-informative** or **vague priors** (e.g., large variance, flat distribution over a wide range). These priors are designed to have minimal influence on the posterior. While they don't explicitly speed up convergence, they ensure that the data primarily drives the inference, and the estimator will converge to the data-driven result as $n$ grows.

*   **Speed of Convergence:** The *speed* at which the Bayesian estimator converges to the true parameter value can be influenced by the prior. An accurate informative prior can lead to faster convergence, especially in data-scarce situations. A very diffuse or inaccurate prior might lead to slower convergence if the data needs to overcome a strong, but incorrect, prior belief.

In summary, while prior hyperparameters can have a substantial effect on the Bayesian estimator in small samples by pulling the posterior towards the prior's beliefs, their influence diminishes as the sample size grows, with the data eventually dominating the inference process and leading to convergence regardless of the prior's initial settings (as long as the prior doesn't exclude the true parameter value entirely).

#### Selecting Non-Informative Priors for the Exponential Distribution

When there is little or no prior information available about the parameter of an exponential distribution, selecting a non-informative prior is crucial to ensure that the posterior distribution is primarily driven by the observed data rather than by arbitrary prior assumptions. For the exponential distribution with rate parameter $\lambda$, common choices for non-informative priors are based on the Gamma distribution (the conjugate prior) or Jeffreys' prior.

**1. 'Vague' Gamma Prior:**

The Gamma distribution, $\text{Gamma}(\alpha_0, \beta_0)$, is a conjugate prior for the exponential rate parameter $\lambda$. To make it non-informative, we choose small values for its hyperparameters $\alpha_0$ and $\beta_0$. Common choices include:

*   **$\alpha_0 = 0.001, \beta_0 = 0.001$**: This is a very common choice that results in a prior that is nearly flat over a wide range and has minimal influence, especially for moderate to large sample sizes. It often approximates a uniform prior on a logarithmic scale.
*   **$\alpha_0 = 1, \beta_0 = 0$ (improper prior, or close to 0 for practical use)**: This effectively reduces to a prior proportional to $1/\lambda$, which is a special case of Jeffreys' prior for $\lambda$. If $\beta_0$ is set to a very small positive number (e.g., $10^{-6}$), it acts as a proper but very vague prior. Note that setting $\beta_0=0$ technically makes the prior improper (does not integrate to 1), but the posterior will often still be proper.

The idea is to choose hyperparameters that spread the prior probability mass thinly across the plausible range of $\lambda$, allowing the likelihood function to shape the posterior.

**2. Jeffreys' Prior:**

Jeffreys' prior is a principled approach to constructing non-informative priors. It is derived from the Fisher Information and is invariant to reparameterization. For the exponential distribution with rate parameter $\lambda$, the Fisher Information for a single observation is $I(\lambda) = 1/\lambda^2$.

Jeffreys' prior is given by $P(\lambda) \propto \sqrt{I(\lambda)}$.
Thus, for the exponential rate parameter:
$ P(\lambda) \propto \sqrt{\frac{1}{\lambda^2}} = \frac{1}{\lambda} $.

This is an improper prior (it does not integrate to a finite value over $(0, \infty)$). However, when combined with the likelihood function, it typically results in a proper posterior distribution. This prior expresses a belief that $\ln \lambda$ is uniformly distributed, meaning that $\lambda$ is equally likely to be in any order of magnitude. It is equivalent to a Gamma prior with $\alpha_0 = 0$ and $\beta_0 = 0$, which are often approximated by small positive values in practice to ensure proper posteriors or for computational stability.

**Considerations when choosing:**

*   **Improper Priors:** While improper priors are commonly used for non-informativeness, always check that they lead to a proper posterior distribution. For the exponential-Gamma conjugate pair, Jeffreys' prior $P(\lambda) \propto 1/\lambda$ results in a proper posterior $\text{Gamma}(n, \sum x_i)$ if $n \ge 1$.
*   **Computational Stability:** In some computational Bayesian methods (like MCMC), using slightly proper but vague priors (e.g., $\text{Gamma}(0.001, 0.001)$) can be more numerically stable than strictly improper priors.
*   **Domain Knowledge:** Even when aiming for non-informativeness, it's good practice to consider the range of plausible values for $\lambda$ (or $\theta$) from domain knowledge to ensure the prior doesn't assign zero probability to regions where the true parameter might lie.

#### Implementing Jeffreys' Prior in a Bayesian Model

Implementing Jeffreys' prior, especially for the exponential distribution's rate parameter $\lambda$, involves understanding its theoretical form and how to apply it in practice. As previously discussed, for $\lambda \sim \text{Exp}(\lambda)$, Jeffreys' prior is $P(\lambda) \propto 1/\lambda$.

Here's a breakdown of how to conceptualize and implement it:

**1. Theoretical Formulation (for $\lambda$ in our exponential problem):**

*   **Jeffreys' Rule:** Jeffreys' prior is generally proportional to the square root of the Fisher Information, $P(\theta) \propto \sqrt{I(\theta)}$.
*   **Fisher Information for $\lambda$:** For the exponential distribution with PDF $f(x; \lambda) = \lambda e^{-\lambda x}$, the Fisher Information for a single observation is $I(\lambda) = 1/\lambda^2$. (This was derived in **Section 6**).
*   **Deriving Jeffreys' Prior:** Therefore, for $\lambda$:
    $$ P(\lambda) \propto \sqrt{I(\lambda)} = \sqrt{\frac{1}{\lambda^2}} = \frac{1}{\lambda} $$

**2. Nature of Jeffreys' Prior: Improper Prior**

*   **Improper:** The prior $P(\lambda) \propto 1/\lambda$ is an *improper prior* because its integral over the entire range of $\lambda$ (from $0$ to $\infty$) does not converge ($\int_0^\infty \frac{1}{\lambda} d\lambda = [\ln \lambda]_0^\infty$, which diverges). This means it does not represent a true probability distribution.
*   **Why use it?** Despite being improper, when combined with a sufficient amount of data (i.e., when the likelihood function is strong enough), it often leads to a *proper posterior distribution*. It's a popular choice for non-informativeness because it is invariant to reparameterization, meaning that if you transform the parameter (e.g., from $\lambda$ to $\ln \lambda$), the Jeffreys' prior for the transformed parameter will be consistent with the original. This is important for ensuring that your choice of parameterization does not arbitrarily influence your inference.

**3. Implementation in Bayesian Modeling (using the conjugate Gamma prior approximation):**

While $1/\lambda$ is technically improper, it corresponds to a limiting case of the conjugate Gamma prior, $\text{Gamma}(\alpha_0, \beta_0)$.

*   **Connection to Gamma Prior:** A Gamma prior with shape parameter $\alpha_0$ and rate parameter $\beta_0$ is $P(\lambda) \propto \lambda^{\alpha_0-1} e^{-\beta_0 \lambda}$. If we set $\alpha_0 = 0$ and $\beta_0 = 0$, the Gamma prior becomes $P(\lambda) \propto \lambda^{-1} e^0 = 1/\lambda$. This is precisely Jeffreys' prior.

*   **Practical Approximation:** Since we cannot use $\alpha_0 = 0$ and $\beta_0 = 0$ directly in some computational tools or for formal mathematical derivation of a proper posterior, Jeffreys' prior is often *approximated* by using very small positive values for $\alpha_0$ and $\beta_0$. For example:
    *   **$\text{Gamma}(0.001, 0.001)$**: This is a commonly used 'vague' Gamma prior that behaves very similarly to Jeffreys' prior while technically being a proper distribution, which helps with numerical stability in algorithms like MCMC.
    *   **$\text{Gamma}(1, \epsilon)$ or $\text{Gamma}(\epsilon, 1)$ with $\epsilon \to 0$**: Other approximations depending on the specific context and desired properties.

**4. Impact on the Posterior Distribution:**

When using Jeffreys' prior (or its approximation) with the exponential likelihood, the posterior distribution for $\lambda$ becomes:

*   **Likelihood:** $L(\lambda; X) \propto \lambda^n e^{-\lambda \sum x_i}$
*   **Prior (Jeffreys):** $P(\lambda) \propto \lambda^{-1}$
*   **Posterior:** $P(\lambda|X) \propto L(\lambda; X) P(\lambda) \propto (\lambda^n e^{-\lambda \sum x_i}) \cdot (\lambda^{-1}) \propto \lambda^{n-1} e^{-\lambda \sum x_i}$

This is the kernel of a Gamma distribution with shape parameter $n$ and rate parameter $\sum x_i$. Thus, the posterior is $\lambda|X \sim \text{Gamma}(n, \sum x_i)$.

*   **Proper Posterior:** Notice that despite starting with an improper prior, the posterior is proper (provided $n \ge 1$). This is a common and desirable outcome when using Jeffreys' prior.

**5. Role in Bayesian Inference:**

By using Jeffreys' prior, the Bayesian inference process becomes largely data-driven. The posterior distribution for $\lambda$ is almost entirely shaped by the observed data, with minimal influence from subjective prior beliefs. This makes it a good choice when there's a strong desire for objectivity or when genuinely little prior information is available.

### Part III Testing

#### Decision Making (Frequentist Approach)

After calculating the test statistic and the p-value, we compare the p-value to a pre-determined significance level, denoted by $\alpha$ (commonly 0.05 or 0.01).

*   If **p-value $< \alpha$**: We reject the null hypothesis ($H_0$). This means that the observed data provides sufficient evidence to conclude that the alternative hypothesis ($H_1$) is true.
*   If **p-value $\ge \alpha$**: We fail to reject the null hypothesis ($H_0$). This means that the observed data does not provide sufficient evidence to conclude that the alternative hypothesis ($H_1$) is true. It does *not* mean that the null hypothesis is true, but rather that we don't have enough evidence to disprove it.

#### Bayesian Hypothesis Testing for $\theta$

In contrast to the Frequentist approach, Bayesian hypothesis testing directly assigns probabilities to hypotheses. Instead of rejecting or failing to reject a null hypothesis, we evaluate the posterior probability of each hypothesis given the data, or we compute Bayes Factors to quantify the evidence in favor of one hypothesis over another.

**Hypotheses:**

Similar to the Frequentist approach, we start by defining hypotheses. However, in the Bayesian context, we often assign prior probabilities to these hypotheses.

*   **Null Hypothesis ($H_0$):** $\theta \in S_0$ (where $S_0$ is a specific region or value for $\theta$)
*   **Alternative Hypothesis ($H_1$):** $\theta \in S_1$ (where $S_1$ is another region, usually $S_1 = \Theta \setminus S_0$)

For our exponential parameter $\theta$, we might be interested in comparing a point null hypothesis ($H_0: \theta = \theta_0$) against a composite alternative ($H_1: \theta \neq \theta_0$). However, Bayesian testing of a point null hypothesis against a composite alternative requires careful consideration of prior distributions over the composite space and often involves using a Bayes Factor.

A more straightforward Bayesian approach for comparing two hypotheses is to compare the posterior probabilities of two competing models or parameter ranges. For instance, we could compare the probability that $\theta$ falls into one range versus another based on its posterior distribution.

#### Bayes Factor for $\theta$ (Conceptual)

The Bayes Factor ($BF_{10}$) quantifies the evidence in the data for $H_1$ relative to $H_0$. It is the ratio of the marginal likelihoods of the data under each hypothesis:

$BF_{10} = \frac{P(X|H_1)}{P(X|H_0)}$

If we have prior probabilities for the hypotheses, $P(H_0)$ and $P(H_1)$, we can update them to posterior probabilities using the Bayes Factor:

$\frac{P(H_1|X)}{P(H_0|X)} = BF_{10} \cdot \frac{P(H_1)}{P(H_0)}$

Interpretation of Bayes Factor (Kass and Raftery, 1995):
*   $1 < BF_{10} < 3$: Weak evidence for $H_1$
*   $3 < BF_{10} < 20$: Positive evidence for $H_1$
*   $BF_{10} > 20$: Strong evidence for $H_1$

Calculating the Bayes Factor, especially for comparing a point null hypothesis, can be complex as it involves integrating the likelihood over the prior distribution under each hypothesis. A common approach for comparing a point null hypothesis $H_0: \theta = \theta_0$ against a composite alternative $H_1: \theta \neq \theta_0$ involves Savage-Dickey density ratio or approximations.

#### Example Scenario (Bayesian Approach)

Instead of a point null, let's consider comparing two regions for $\theta$. For instance, we might want to know if $\theta$ is likely to be less than 0.5 or greater than or equal to 0.5, given our data.

*   **Hypothesis 1 ($H_1$):** $\theta < 0.5$
*   **Hypothesis 2 ($H_2$):** $\theta \ge 0.5$

We can use the posterior distribution for $\lambda$ (and thus for $\theta$) to calculate the posterior probability of each hypothesis. Recall that $\theta = e^{-\lambda}$. We will use the posterior from the non-informative prior for $\lambda$: $\lambda|X \sim \text{Gamma}(\alpha_1, \beta_1)$, where $\alpha_1 = \alpha_0 + n$ and $\beta_1 = \beta_0 + \sum x_i$.

To find the probability $P(\theta < 0.5 | X)$, we need to find $P(e^{-\lambda} < 0.5 | X)$, which implies $P(-\lambda < \ln(0.5) | X)$, or $P(\lambda > -\ln(0.5) | X)$.

Let $\lambda_{critical} = -\ln(0.5) = \ln(2) \approx 0.693$.

*   $P(H_1 | X) = P(\theta < 0.5 | X) = P(\lambda > \lambda_{critical} | X)$
*   $P(H_2 | X) = P(\theta \ge 0.5 | X) = P(\lambda \le \lambda_{critical} | X)$

```python
import numpy as np
import math
from scipy.stats import gamma

# Parameters for the non-informative posterior of lambda
# Re-defining values that might have been lost from the environment
alpha_0_noninf = 0.001
beta_0_noninf = 0.001
n = 50 # Sample size

# Re-simulate data to ensure sum_x is available
theta_true = 0.6
lambda_true = -math.log(theta_true)
np.random.seed(42) # for reproducibility
sample_data = np.random.exponential(scale=1/lambda_true, size=n)
sum_x = np.sum(sample_data)

alpha_1_noninf = alpha_0_noninf + n
beta_1_noninf = beta_0_noninf + sum_x

# Define the critical value for theta
theta_critical = 0.5

# Convert theta_critical to lambda_critical
# theta = exp(-lambda) => lambda = -ln(theta)
lambda_critical = -math.log(theta_critical)

# Calculate posterior probabilities
# P(theta < theta_critical | X) = P(lambda > lambda_critical | X)
p_H1_given_X = 1 - gamma.cdf(lambda_critical, a=alpha_1_noninf, scale=1/beta_1_noninf)

# P(theta >= theta_critical | X) = P(lambda <= lambda_critical | X)
p_H2_given_X = gamma.cdf(lambda_critical, a=alpha_1_noninf, scale=1/beta_1_noninf)

# Decision based on posterior probabilities
if p_H1_given_X > p_H2_given_X:
    print(f"\nDecision: The data provides more evidence for theta < {theta_critical}.")
else:
    print(f"\nDecision: The data provides more evidence for theta >= {theta_critical}.")
```

#### Summary of Bayesian Hypothesis Testing

In contrast to Frequentist methods, Bayesian hypothesis testing directly assesses the **posterior probability** of hypotheses or quantifies evidence using **Bayes Factors**. We explored an example where we calculated the posterior probabilities for two competing hypotheses about $\theta$: $H_1: \theta < 0.5$ and $H_2: \theta \ge 0.5$.

By leveraging the posterior distribution of $\lambda$ (derived from a non-informative prior) and the transformation $\theta = e^{-\lambda}$, we computed:

*   $P(\theta < 0.5 | X) = P(\lambda > \lambda_{critical} | X)$
*   $P(\theta \ge 0.5 | X) = P(\lambda \le \lambda_{critical} | X)$

Where $\lambda_{critical} = -\ln(0.5)$. This approach provides a direct measure of belief in each hypothesis given the observed data, offering a more intuitive interpretation compared to p-values and significance levels. The decision is then based on which hypothesis has a higher posterior probability.
