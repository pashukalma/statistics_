#### MLE for Censored Data

MLE for theta based on the censored data Y1,...,Yn.

**The likelihood function**
- $Y_i$ follows a Bernoulli distribution with parameter $p = \theta^\tau$, where $\tau = \frac{T_{cens}}{E[X]}$. The pmf for a single Bernoulli trial $Y_i$ is $P(Y_i=y_i) = p^{y_i} (1-p)^{1-y_i} = (\theta^\tau)^{y_i} (1-\theta^\tau)^{1-y_i}$. For $n$ i.i.d. observations $Y_1, \dots, Y_n$, the likelihood function $L(\theta)$ is the product of their individual pmfs $ L(\theta) = \prod_{i=1}^n (\theta^\tau)^{y_i} (1-\theta^\tau)^{1-y_i}$. Let $S = \sum_{i=1}^n y_i$ be the number of "successes" (i.e., observations where $Y_i=1$) and $n-S = \sum_{i=1}^n (1-y_i)$ be the number of "failures" (i.e., observations where $Y_i=0$).

The likelihood function is $ L(\theta) = (\theta^\tau)^S (1-\theta^\tau)^{n-S} $$

**The log-likelihood function**
- $l(\theta) = \ln\left( (\theta^\tau)^S (1-\theta^\tau)^{n-S} \right)$, $\quad$ $ l(\theta) = \ln((\theta^\tau)^S) + \ln((1-\theta^\tau)^{n-S})$, $\quad$ $ l(\theta) = S \ln(\theta^\tau) + (n-S) \ln(1-\theta^\tau)$

Further simplifying the first term $l(\theta) = S \tau \ln(\theta) + (n-S) \ln(1-\theta^\tau)$

**Differentiate the log-likelihood**, $l(\theta)$, with respect to $\theta$. Be careful with the chain rule when differentiating terms involving $\theta^\tau$.
- Given the log-likelihood function $l(\theta) = S \tau \ln(\theta) + (n-S) \ln(1-\theta^\tau)$, we differentiate $l(\theta)$ with respect to $\theta$ $ \frac{dl}{d\theta} = \frac{d}{d\theta} [S \tau \ln(\theta)] + \frac{d}{d\theta} [(n-S) \ln(1-\theta^\tau)]$.

The derivative of the log-likelihood is $ \frac{dl}{d\theta} = \frac{S \tau}{\theta} - \frac{(n-S) \tau \theta^{\tau-1}}{1-\theta^\tau}$

**Set the derivative $\frac{dl}{d\theta}$ equal to zero and solve for $\theta$**. This gives the MLE, $\hat{\theta}$. We solve for $\theta$, $\hat{\theta} = \left(\frac{S}{n}\right)^{1/\tau}$

**Express the MLE, $\hat{\theta}$, in terms of the number of 'successes' ($S = \sum_{i=1}^n Y_i$) and the total number of observations ($n$).**

We found the MLE to be $\hat{\theta} = \left(\frac{S}{n}\right)^{1/\tau}$, where:
- $S = \sum_{i=1}^n Y_i$ is the number of 'successes' (i.e., observations where $Y_i=1$).
- $n$ is the total number of observations.
- $\tau = \frac{T_{cens}}{E[X]}$ is a constant defined in earlier steps, representing the ratio of the censoring time to the expectation of $X_i$ (which itself depends on the unknown parameter $\theta$). However, for the purpose of deriving the MLE for $p = \theta^\tau$, $\tau$ is treated as a known constant.

This expression clearly states the MLE $\hat{\theta}$ in terms of the observed sample statistics ($S$ and $n$) and the given constant $\tau$ related to the censoring time.

#### General Conditions for MLE Consistency

Maximum Likelihood Estimators (MLEs) are generally consistent under certain regularity conditions. These conditions ensure that as the sample size $n$ approaches infinity, the MLE converges in probability to the true parameter value.

The key regularity conditions typically include:

1.  **Identically and Independently Distributed (i.i.d.) Data**: The observations $X_1, \dots, X_n$ are i.i.d. random variables.
2.  **Model Specification**: The probability distribution of the data is correctly specified (i.e., the true parameter $\theta_0$ belongs to the parameter space $\Theta$).
3.  **Identifiability**: Different parameter values lead to different probability distributions (i.e., if $\theta_1 \ne \theta_2$, then $f(x; \theta_1) \ne f(x; \theta_2)$ for some $x$).
4.  **Continuity and Differentiability**: The log-likelihood function $l(\theta; x)$ is at least twice continuously differentiable with respect to $\theta$.
5.  **Finite Fisher Information**: The Fisher Information $I(\theta)$ is finite and positive for all $\theta \in \Theta$.
6.  **Domination**: The support of the probability density function (PDF) does not depend on the parameter $\theta$, or there exist integrable functions that bound the derivatives of the log-likelihood function.

Under these regularity conditions, the MLE $\hat{\theta}_n$ is consistent, meaning it converges in probability to the true parameter $\theta_0$ as $n \to \infty$
- $\hat{\theta}_n \xrightarrow{P} \theta_0$

#### 2. The Likelihood Function, $L(\theta|y)$ (Recap)

### 3. Bayes' Rule and the Posterior Distribution, $p(\theta|y)$

Bayes' Rule combines the prior distribution with the likelihood function to produce the posterior distribution, which represents our updated beliefs about $\theta$ after observing the data.

- **Bayes' Rule**:
$ p(\theta|y) = \frac{L(\theta|y) p(\theta)}{p(y)} = \frac{L(\theta|y) p(\theta)}{\int L(\theta|y) p(\theta) d\theta} \propto L(\theta|y) p(\theta) $,
where $p(y)$ is the marginal likelihood (a normalizing constant).

- **Posterior Distribution**: The probability distribution of $\theta$ conditional on the observed data $y$. It reflects our updated beliefs.

To find the posterior distribution, we multiply our Beta prior $p(\theta; \alpha_{prior}, \beta_{prior})$ by the likelihood function $L(\theta|y)$, ignoring the normalizing constant $p(y)$ for now, as we'll determine the form of the posterior.
From Part I, the likelihood function is
$ L(\theta|y) = (\theta^\tau)^S (1-\theta^\tau)^{n-S} $.
And our Beta prior is
$ p(\theta) \propto \theta^{\alpha_{prior}-1} (1-\theta)^{\beta_{prior}-1} $.
Multiplying these two gives us the unnormalized posterior
$ p(\theta|y) \propto L(\theta|y) p(\theta) \propto (\theta^\tau)^S (1-\theta^\tau)^{n-S} \cdot \theta^{\alpha_{prior}-1} (1-\theta)^{\beta_{prior}-1} $.

```python
import sympy

# Define symbolic variables
theta_sym = sympy.symbols('theta', real=True, positive=True)
S_sym, n_sym, tau_sym = sympy.symbols('S n tau', integer=True, positive=True)
alpha_prior, beta_prior = sympy.symbols('alpha_prior beta_prior', real=True, positive=True)

# Define the Beta prior (unnormalized part for simplicity in multiplication)
prior_dist_unnormalized = theta_sym**(alpha_prior-1) * (1-theta_sym)**(beta_prior-1)

# Define the likelihood function from Part I
likelihood_func = (theta_sym**tau_sym)**S_sym * (1 - theta_sym**tau_sym)**(n_sym - S_sym)

# Calculate the unnormalized posterior by multiplying prior and likelihood
unnormalized_posterior = prior_dist_unnormalized * likelihood_func

# Simplify the expression (optional, but good for readability)
unnormalized_posterior_simplified = sympy.simplify(unnormalized_posterior)
```

#### 4. Bayesian Estimation

A Bayesian estimator is a point estimate of the parameter $\theta$ derived from its posterior distribution. Common choices for a point estimate include:

- **Posterior Mean**: $E[\theta|y] = \int \theta \cdot p(\theta|y) d\theta$. This is often used when the loss function is squared error.
- **Posterior Median**: The value $m$ such that $P(\theta \le m|y) = 0.5$.
- **Maximum A Posteriori (MAP) Estimator**: The mode of the posterior distribution, $\hat{\theta}_{MAP} = \arg\max_{\theta} p(\theta|y)$. This is similar to the MLE but incorporates the prior.

To derive the MAP estimator, we maximize the posterior distribution. Since the logarithm is a monotonically increasing function, maximizing the posterior is equivalent to maximizing its logarithm (the log-posterior). The steps involve:

1.  Take the logarithm of the unnormalized posterior distribution.
2.  Differentiate the log-posterior with respect to $\theta$.
3.  Set the derivative equal to zero and solve for $\theta$.

Recall our unnormalized posterior:
$ p(\theta|y) \propto \theta^{S\tau + \alpha_{prior} - 1} (1-\theta)^{\beta_{prior} - 1} (1-\theta^\tau)^{n - S} $.
The log-posterior (ignoring constants) is:
$ \log p(\theta|y) \propto (S\tau + \alpha_{prior} - 1) \log(\theta) + (\beta_{prior} - 1) \log(1-\theta) + (n - S) \log(1-\theta^\tau) $.
We will now attempt to differentiate this and solve for $\theta$. It is important to note that for general values of $\tau$, a closed-form solution for $\hat{\theta}_{MAP}$ might be difficult to obtain due to the presence of $\theta^\tau$ within the logarithm term.

```python
import sympy

# Define symbolic variables (re-defining to ensure they are available)
theta_sym = sympy.symbols('theta', real=True, positive=True)
S_sym, n_sym, tau_sym = sympy.symbols('S n tau', integer=True, positive=True)
alpha_prior, beta_prior = sympy.symbols('alpha_prior beta_prior', real=True, positive=True)

# Unnormalized posterior from previous step
unnormalized_posterior = theta_sym**(S_sym*tau_sym + alpha_prior - 1) * \
                         (1 - theta_sym)**(beta_prior - 1) * \
                         (1 - theta_sym**tau_sym)**(n_sym - S_sym)

# Take the logarithm of the unnormalized posterior
log_posterior = sympy.log(unnormalized_posterior)

# Differentiate the log-posterior with respect to theta
d_log_posterior_d_theta = sympy.diff(log_posterior, theta_sym)

print(f"Log-posterior: {log_posterior}\n")
print(f"Derivative of log-posterior with respect to theta: {d_log_posterior_d_theta}\n")

# Set the derivative to zero and attempt to solve for theta (MAP estimator)
# This equation is generally complex, and a simple closed-form solution may not exist for general tau.
# We will print the equation that needs to be solved.
map_equation = sympy.Eq(d_log_posterior_d_theta, 0)
print(f"Equation to solve for MAP estimator (set derivative to zero): {map_equation}\n")

# Attempt to solve (sympy.solve might not yield a simple closed form for general tau)
# map_estimator_solutions = sympy.solve(map_equation, theta_sym)
# print(f"MAP Estimator Solutions: {map_estimator_solutions}")

# Acknowledgment of complexity
print("Note: For general values of tau, solving this equation for theta_sym to find a closed-form MAP estimator is analytically very challenging, and sympy.solve may not yield a simple expression. Numerical methods would typically be employed for specific values of S, n, tau, alpha_prior, and beta_prior.")
```

#### Derivation of Fisher Information for $\lambda$ (Exponential Distribution)

To formally justify the Jeffreys prior, we need to derive the Fisher Information $I(\lambda)$ for the exponential distribution, which is the distribution of our original random variable $X$. The PDF of an exponential distribution with rate parameter $\lambda$ is $ f(x; \lambda) = \lambda e^{-\lambda x}, \quad \text{for } x \ge 0, \lambda > 0 $. We will follow these steps:
- Calculate the log-likelihood for a single observation.
- Compute the first derivative of the log-likelihood with respect to $\lambda$.
- Compute the second derivative of the log-likelihood with respect to $\lambda$.
- Calculate the negative expectation of the second derivative to obtain the Fisher Information.

Log-Likelihood for a Single Observation: Take the natural logarithm of the PDF $ \ln f(x; \lambda) = \ln(\lambda e^{-\lambda x}) = \ln(\lambda) - \lambda x $.

First Derivative of the Log-Likelihood: Differentiate the log-likelihood with respect to $\lambda$, $ \frac{d}{d\lambda} \ln f(x; \lambda) = \frac{d}{d\lambda} (\ln(\lambda) - \lambda x) = \frac{1}{\lambda} - x $.

Second Derivative of the Log-Likelihood: Differentiate the first derivative with respect to $\lambda$, $ \frac{d^2}{d\lambda^2} \ln f(x; \lambda) = \frac{d}{d\lambda} \left( \frac{1}{\lambda} - x \right) = -\frac{1}{\lambda^2} $.


Fisher Information $I(\lambda)$:
The Fisher Information for a single observation is defined as the negative expectation of the second derivative of the log-likelihood, $ I(\lambda) = -E\left[\frac{d^2}{d\lambda^2} \ln f(X; \lambda)\right] $. Since the second derivative, $-\frac{1}{\lambda^2}$, is a constant with respect to $X$ (it does not depend on $X$), its expectation is just itself,
$ E\left[-\frac{1}{\lambda^2}\right] = -\frac{1}{\lambda^2} $.
Therefore, t$I(\lambda)$ for a single observation $X$ from an exponential distribution is $ I(\lambda) = -\left(-\frac{1}{\lambda^2}\right) = \frac{1}{\lambda^2} $.

#### 5. Comparison: Frequentist (MLE) vs. Bayesian Estimators

Both MLEs and Bayesian estimators aim to estimate unknown parameters, but they do so from fundamentally different philosophical standpoints, leading to different properties and interpretations.

**Asymptotic Normality of MLE (Recap)**:
Recall from Part I, under regularity conditions, the MLE $\hat{\theta}$ is asymptotically normally distributed:
$ \sqrt{n}(\hat{\theta} - \theta) \xrightarrow{d} N(0, I(\theta)^{-1}) $,
where $I(\theta)^{-1}$ is the inverse of the Fisher Information. This implies that for large $n$, the MLE is approximately unbiased and achieves the Cramer-Rao Lower Bound, making it asymptotically efficient.

#### Properties of Bayesian Estimators

*   **Consistency**: Under broad conditions, Bayesian estimators are also consistent, meaning they converge to the true parameter value as $n \to \infty$.
*   **Efficiency**: Bayesian estimators can be more efficient than MLEs, especially with informative priors or smaller sample sizes, as they leverage additional information.
*   **Finite Sample Properties**: Unlike MLEs, which often rely on asymptotic properties for optimality, Bayesian estimators have well-defined finite-sample properties, as they directly derive from the full posterior distribution.
*   **Incorporation of Prior Information**: A key advantage is the ability to integrate prior knowledge, which can stabilize estimates, especially with sparse data.

#### Impact of Prior Hyperparameters on Bayesian Estimator's Convergence

The choice of prior hyperparameters $(\alpha_{prior}, \beta_{prior})$ significantly influences the posterior distribution and, consequently, the Bayesian estimator, especially with small sample sizes.

*   **Strong (Informative) Priors**: If the prior hyperparameters are large, the prior distribution is concentrated around a specific value, representing strong prior belief. In such cases, the data needs to be very strong to shift the posterior significantly. This can lead to slower convergence of the posterior mean (or MAP) to the true parameter if the prior is far from the true value.
*   **Weak (Non-Informative) Priors**: If the prior hyperparameters are small (e.g., $\alpha_{prior}=1, \beta_{prior}=1$ for a Uniform prior, or $\alpha_{prior}=0.5, \beta_{prior}=0.5$ for Jeffreys prior), the prior distribution is more spread out, indicating less strong prior belief. In this scenario, the data has a stronger influence, and the posterior will converge more quickly to the likelihood's mode, making the Bayesian estimator behave more like the MLE as $n \to \infty$.
*   **Asymptotic Behavior**: As the sample size $n$ grows very large, the likelihood function typically dominates the prior. Consequently, the influence of the prior hyperparameters diminishes, and the Bayesian estimator tends to converge to the same value as the MLE, regardless of the initial prior choice (provided the prior is not pathological, e.g., assigning zero probability to the true parameter value).

#### 6. Selecting Non-Informative Priors for the Exponential Distribution

When there is little to no prior information available, or when one wishes for the data to speak for itself as much as possible, non-informative priors are chosen. For the original exponential distribution with rate parameter $\lambda$, various non-informative priors can be considered, which then translate to priors on $\theta = e^{-\lambda}$.

Recall that the rate parameter $\lambda = -\ln(\theta)$. If we are interested in a non-informative prior for $\lambda$, we need to consider how this translates to $\theta$.

*   **Uniform Prior on $\lambda$**: A simple non-informative choice might be a uniform prior on $\lambda$ over some large range $[0, L]$. However, this choice can be problematic as it's not invariant to reparameterization. If $\lambda \sim U(0, L)$, then the prior for $\theta = e^{-\lambda}$ would not be uniform.

*   **Jeffreys Prior**: Jeffreys prior is a principled choice for a non-informative prior because it is invariant to reparameterization. It is derived from the Fisher Information $I(\lambda)$ as $p(\lambda) \propto \sqrt{I(\lambda)}$.
    *   For an exponential distribution, the Fisher Information for a single observation is $I(\lambda) = 1/\lambda^2$. Thus, the Jeffreys prior for $\lambda$ is $p(\lambda) \propto 1/\lambda$. This implies that a prior on $\theta$ would need to be derived using a change of variables. If $p(\lambda) = 1/\lambda$, then $p(\theta) = p(\lambda)|d\lambda/d\theta|$. Since $\lambda = -\ln(\theta)$, $d\lambda/d\theta = -1/\theta$. So, $p(\theta) \propto (1/(-\ln(\theta))) (1/\theta)$. This is not a standard Beta prior.

*   **Weakly Informative Beta Prior on $\theta$**: Given our problem's setup where $\theta \in (0,1)$, a weakly informative Beta prior, such as $\text{Beta}(1,1)$ (which is a Uniform(0,1) distribution), or $\text{Beta}(0.5, 0.5)$ (which is a Jeffreys prior for a Bernoulli probability, but not directly for our $\theta$ mapping from an exponential rate), can be used. These priors are easy to work with due to conjugacy.
    *   A **Uniform prior** on $\theta$, i.e., $p(\theta) \sim \text{Beta}(1,1)$, implies that all values of $\theta$ are equally likely *a priori*. This is often chosen for its simplicity and because it allows the data to primarily drive the posterior.
    *   **$\text{Beta}(0.5, 0.5)$** is a common choice for a Jeffreys prior for parameters like the success probability of a Bernoulli distribution. It gives more weight to values of $\theta$ near 0 and 1, reflecting uncertainty at the boundaries.

**Choosing a practical non-informative prior for $\theta$ in our censored data model:**
Given that $\theta$ is the parameter of interest in our Bernoulli likelihood $(Y_i \sim \text{Bernoulli}(\theta^\tau))$, a suitable non-informative prior for $\theta$ that maintains conjugacy and simplicity is often a **Beta distribution with small, equal parameters**, e.g., $\text{Beta}(1,1)$ (Uniform) or $\text{Beta}(0.5,0.5)$. These choices allow us to proceed with analytical derivations while minimizing the prior's influence on the posterior when data is abundant.

### PART III: Hypothesis Testing

This section will explore hypothesis testing from both Frequentist and Bayesian perspectives, applying them to our censored data model.

#### Frequentist Hypothesis Testing

In the Frequentist framework, hypothesis testing involves making decisions about population parameters based on sample data. We formulate a null hypothesis ($H_0$) and an alternative hypothesis ($H_1$), choose a test statistic, and evaluate the evidence against $H_0$ using p-values or critical regions.

**Formulating Hypotheses**

Given the context that $\theta > 1/3$ might be a condition from previous studies, we can formulate a one-sided hypothesis test for $\theta$. For example, we might be interested in whether $\theta$ is indeed greater than $1/3$.

*   **Null Hypothesis ($H_0$)**: $\theta \le 1/3$
*   **Alternative Hypothesis ($H_1$)**: $\theta > 1/3$

**Test Statistic**

For large sample sizes, we can use the MLE $\hat{\theta}$ as our test statistic. Recall that $\hat{\theta} = \left(\frac{S}{n}\right)^{1/\tau}$.

**Asymptotic Distribution of the Test Statistic**

We know that $\sqrt{n}(\hat{\theta} - \theta) \xrightarrow{d} N(0, \sigma^2)$, where $\sigma^2 = \frac{1-\theta^\tau}{\tau^2 \theta^{\tau-2}}$ is the asymptotic variance. Under the null hypothesis, specifically at the boundary $\theta = 1/3$, we can approximate the distribution of $\hat{\theta}$.
To construct a test, we can use the Z-statistic:

$ Z = \frac{\hat{\theta} - \theta_0}{\sqrt{\frac{\sigma^2}{n}}} = \frac{\hat{\theta} - \theta_0}{\sqrt{\frac{1-\theta_0^\tau}{n\tau^2 \theta_0^{\tau-2}}}} $,
where $\theta_0 = 1/3$ is the hypothesized value under the null hypothesis.

**Decision Rule**

For a given significance level $\alpha$ (e.g., $\alpha = 0.05$):

*   **Calculate the p-value**: The probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming $H_0$ is true. If $p \text{-value} < \alpha$, we reject $H_0$.
*   **Critical Region Approach**: Find the critical value $z_{\alpha}$ from the standard normal distribution such that $P(Z > z_{\alpha}) = \alpha$. If our calculated $Z > z_{\alpha}$, we reject $H_0$.

In our case, since $H_1: \theta > 1/3$ is a right-tailed test, we would reject $H_0$ if $Z > z_{\alpha}$.

#### Bayesian Hypothesis Testing

In contrast to the Frequentist approach, Bayesian hypothesis testing involves comparing the posterior probabilities of competing hypotheses, often using Bayes Factors. It directly quantifies the evidence for one hypothesis over another.

**Formulating Hypotheses and Assigning Prior Probabilities**

Similar to the Frequentist approach, we define hypotheses about $\theta$. For example:

*   **Null Hypothesis ($H_0$)**: $\theta \in (0, 1/3]$ (i.e., $\theta \le 1/3$)
*   **Alternative Hypothesis ($H_1$)**: $\theta \in (1/3, 1)$ (i.e., $\theta > 1/3$)

Before observing the data, we must assign prior probabilities to each hypothesis, $P(H_0)$ and $P(H_1)$, such that $P(H_0) + P(H_1) = 1$. These reflect our initial beliefs about which hypothesis is more likely to be true. For instance, if we have no strong prior preference, we might set $P(H_0) = P(H_1) = 0.5$.

**Calculating the Bayes Factor**

The Bayes Factor ($BF_{10}$) quantifies the evidence for $H_1$ relative to $H_0$. It is the ratio of the marginal likelihoods under each hypothesis:

$ BF_{10} = \frac{P(y|H_1)}{P(y|H_0)} = \frac{\int_{H_1} L(\theta|y) p(\theta|H_1) d\theta}{\int_{H_0} L(\theta|y) p(\theta|H_0) d\theta} $

where $p(\theta|H_0)$ and $p(\theta|H_1)$ are the prior distributions for $\theta$ restricted to the parameter spaces defined by $H_0$ and $H_1$, respectively. These are often normalized versions of our chosen prior distribution over their respective hypothesis spaces.

**Computing Posterior Probabilities of Hypotheses**

Using Bayes' Theorem, we can update our prior probabilities of the hypotheses to posterior probabilities after observing the data:

$ P(H_1|y) = \frac{P(y|H_1) P(H_1)}{P(y)} = \frac{P(y|H_1) P(H_1)}{P(y|H_0) P(H_0) + P(y|H_1) P(H_1)} $

And similarly for $P(H_0|y)$. The relationship with the Bayes Factor is:

$ \frac{P(H_1|y)}{P(H_0|y)} = BF_{10} \frac{P(H_1)}{P(H_0)} $

This shows that the posterior odds are the prior odds multiplied by the Bayes Factor.

**Making a Decision**

In Bayesian hypothesis testing, the decision is based on the posterior probabilities of the hypotheses. If $P(H_1|y)$ is sufficiently high (e.g., $P(H_1|y) > 0.95$), we might conclude that there is strong evidence for $H_1$. There are no arbitrary significance levels like $\alpha$; instead, the posterior probabilities directly provide a measure of belief in each hypothesis.
