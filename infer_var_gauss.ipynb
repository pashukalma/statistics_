#### Express Sample Variance

Express the sample variance sigma_hat^2 for a sample of n iid Gaussian random variables X1, ..., Xn with known mean mu = 0.

The sample variance, denoted as $\hat{\sigma}^2$, for a sample of $n$ i.i.d. Gaussian random variables $X_1, \dots, X_n$ with a **known population mean $\mu = 0$** is given by:
- $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (X_i - \mu)^2$

Since the population mean $\mu$ is known to be $0$, the expression simplifies to
- $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (X_i - 0)^2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2$

This simplification occurs because we are not estimating the mean from the sample; rather, we are directly using the known population mean. If the mean were unknown and estimated from the sample (e.g., by the sample mean $\bar{X}$), the denominator would typically be $n-1$ for an unbiased estimator, and the term would be $(X_i - \bar{X})^2$. However, with a known population mean, we use the sum of squared observations divided by $n$.

#### Prove Consistency

Explain why sigma_hat^2 is a consistent estimator of sigma^2, referencing relevant statistical theorems or properties.

**Definition of a Consistent Estimator**

An estimator $\hat{\theta}_n$ is said to be a consistent estimator for a parameter $\theta$ if it converges in probability to $\theta$ as the sample size $n$ approaches infinity. Mathematically, this means that for any $\epsilon > 0$, $P(|\hat{\theta}_n - \theta| \ge \epsilon) \to 0 \quad \text{as} \quad n \to \infty$

In our case, we want to show that $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2$ is a consistent estimator for $\sigma^2$. This means we need to demonstrate that $\hat{\sigma}^2$ converges in probability to $\sigma^2$ as $n \to \infty$.

One common way to prove consistency is to show that the estimator is asymptotically unbiased (i.e., $E[\hat{\theta}_n] \to \theta$ as $n \to \infty$) and that its variance approaches zero as $n \to \infty$ (i.e., $Var(\hat{\theta}_n) \to 0$ as $n \to \infty$). Alternatively, the Weak Law of Large Numbers can be directly applied.

**Expected Value of $\hat{\sigma}^2$**

Let's calculate the expected value of our estimator $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2$. Given that $X_i$ are i.i.d. Gaussian random variables with known mean $\mu = 0$ and unknown variance $\sigma^2$, we know that: $E[X_i] = 0$, & $Var(X_i) = E[X_i^2] - (E[X_i])^2 = E[X_i^2] - 0^2 = \sigma^2$

Now, let's find the expectation of $\hat{\sigma}^2$: $E[\hat{\sigma}^2] = E\left[\frac{1}{n} \sum_{i=1}^{n} X_i^2\right]$. By linearity of expectation: $E[\hat{\sigma}^2] = \frac{1}{n} \sum_{i=1}^{n} E[X_i^2]$

Since $E[X_i^2] = \sigma^2$ for all $i$, $E[\hat{\sigma}^2] = \frac{1}{n} \sum_{i=1}^{n} \sigma^2$

$E[\hat{\sigma}^2] = \frac{1}{n} (n \cdot \sigma^2) = \sigma^2$

This shows that $\hat{\sigma}^2$ is an **unbiased estimator** of $\sigma^2$. An unbiased estimator is also asymptotically unbiased.

**Variance of $\hat{\sigma}^2$**

Next, let's calculate the variance of our estimator $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2$; $Var(\hat{\sigma}^2) = Var\left(\frac{1}{n} \sum_{i=1}^{n} X_i^2\right)$

Since $X_i$ are i.i.d., $X_i^2$ are also independent (though not identically distributed in general, but here they are, as they all come from the same Gaussian distribution). So, we can pull out $\frac{1}{n^2}$ and sum the variances: $Var(\hat{\sigma}^2) = \frac{1}{n^2} \sum_{i=1}^{n} Var(X_i^2)$

Now, we need to find $Var(X_i^2)$. We know that $Var(Y) = E[Y^2] - (E[Y])^2$. So, for $Y = X_i^2$: $Var(X_i^2) = E[(X_i^2)^2] - (E[X_i^2])^2 = E[X_i^4] - (E[X_i^2])^2$

From the previous step, we know $E[X_i^2] = \sigma^2$.

For a Gaussian random variable $X$ with mean $\mu=0$ and variance $\sigma^2$, the fourth moment is $E[X^4] = 3\sigma^4$. This is a standard property of the normal distribution.

Substituting these values into the variance of $X_i^2$: $Var(X_i^2) = 3\sigma^4 - (\sigma^2)^2 = 3\sigma^4 - \sigma^4 = 2\sigma^4$

Now, substitute this back into the expression for $Var(\hat{\sigma}^2)$:

$Var(\hat{\sigma}^2) = \frac{1}{n^2} \sum_{i=1}^{n} (2\sigma^4)= \frac{1}{n^2} (n \cdot 2\sigma^4) = \frac{2\sigma^4}{n}$

This shows that the variance of the estimator $\hat{\sigma}^2$ approaches 0 as $n \to \infty$.

**Conclusion of Consistency**

We have shown:

1.  **Unbiasedness (or Asymptotic Unbiasedness)**: $E[\hat{\sigma}^2] = \sigma^2$. This means the estimator is unbiased, and therefore also asymptotically unbiased.
2.  **Variance approaches zero**: $Var(\hat{\sigma}^2) = \frac{2\sigma^4}{n}$. As $n \to \infty$, $Var(\hat{\sigma}^2) \to 0$.

An estimator is consistent if it is asymptotically unbiased and its variance tends to zero as the sample size approaches infinity. Since both conditions are met, we can conclude that $\hat{\sigma}^2$ is a **consistent estimator** for $\sigma^2$.


### Alternative Approach: Weak Law of Large Numbers (WLLN)

**Another way to prove consistency is by directly applying the WLLN.**

The WLLN states that if $Y_1, Y_2, \dots, Y_n$ are i.i.d. random variables with finite mean $E[Y_i] = \mu$ and finite variance $Var(Y_i) = \tau^2$, then the sample mean $\bar{Y}_n = \frac{1}{n} \sum_{i=1}^{n} Y_i$ converges in probability to $\mu$ as $n \to \infty$.

In our case, let $Y_i = X_i^2$. We have previously shown that:
*   $E[Y_i] = E[X_i^2] = \sigma^2$ (finite mean)
*   $Var(Y_i) = Var(X_i^2) = 2\sigma^4$ (finite variance)

Since $X_i$ are i.i.d., $Y_i = X_i^2$ are also i.i.d. Therefore, by the WLLN:

$\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2 \xrightarrow{P} E[X_i^2] = \sigma^2 \quad \text{as} \quad n \to \infty$

This means that $\hat{\sigma}^2$ converges in probability to $\sigma^2$, which is the definition of consistency. Thus, the sample variance $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2$ is a consistent estimator of $\sigma^2$.

#### Apply CLT for Asymptotic Normality

Demonstrate how the CLT implies that the vector (1/n Sum(Xi), 1/n Sum(Xi^2))' is asymptotically normal.

**Define the vector $Y_i = (X_i, X_i^2)'$ and its properties**

Given $X_i \sim N(0, \sigma^2)$ are i.i.d. Gaussian random variables. We define the vector $Y_i = (X_i, X_i^2)'$. To apply the Multivariate Central Limit Theorem, we need to find its mean vector $E[Y_i]$ and its covariance matrix $Var(Y_i)$. We compute the Mean Vector and Covariance Matrix step by step.

$E[Y_i] = \begin{pmatrix} E[X_i] \\ E[X_i^2] \end{pmatrix}$, $Var(Y_i) = E[(Y_i - E[Y_i])(Y_i - E[Y_i])'] = \begin{pmatrix} Var(X_i) & Cov(X_i, X_i^2) \\ Cov(X_i^2, X_i) & Var(X_i^2) \end{pmatrix}$

**Multivariate Central Limit Theorem (MCLT)**

The MCLT Theorem states that if $Y_1, Y_2, \dots, Y_n$ are i.i.d. random vectors with mean vector $E[Y_i] = \mu_Y$ and finite covariance matrix $Var(Y_i) = \Sigma_Y$, then as $n \to \infty$, the sample mean vector $\bar{Y}_n = \frac{1}{n} \sum_{i=1}^n Y_i$ is asymptotically normally distributed: $\sqrt{n}(\bar{Y}_n - \mu_Y) \xrightarrow{D} N(0, \Sigma_Y)$

In our case, $Y_i = (X_i, X_i^2)'$ are i.i.d. random vectors because $X_i$ are i.i.d. Gaussian random variables. Therefore, the MCLT can be directly applied to $\bar{Y}_n = (\frac{1}{n} \sum X_i, \frac{1}{n} \sum X_i^2)'$ once we determine its mean vector $\mu_Y = E[Y_i]$ and covariance matrix $\Sigma_Y = Var(Y_i)$.

**Explicitly compute the mean vector $E[Y_i]$**

Given that $X_i \sim N(0, \sigma^2)$:
1.  Expected value of $X_i$: $E[X_i] = \mu = 0$ (since the mean is given as 0).

2.  Expected value of $X_i^2$: We know that $Var(X_i) = E[X_i^2] - (E[X_i])^2$. Since $Var(X_i) = \sigma^2$ and $E[X_i] = 0$, we have $\sigma^2 = E[X_i^2] - (0)^2$, and $E[X_i^2] = \sigma^2$

Therefore, the mean vector $E[Y_i]$ is $E[Y_i] = \begin{pmatrix} E[X_i] \\\\ E[X_i^2] \end{pmatrix} = \begin{pmatrix} 0 \\\\ \sigma^2 \end{pmatrix}$

**Explicitly compute the covariance matrix $Cov(Y_i)$**

We need to compute the components of the covariance matrix $Cov(Y_i) = \begin{pmatrix} Var(X_i) & Cov(X_i, X_i^2) \\\\ Cov(X_i^2, X_i) & Var(X_i^2) \end{pmatrix}$

Given that $X_i \sim N(0, \sigma^2)$:

1.  **$Var(X_i)$**: This is directly given as $Var(X_i) = \sigma^2$.

2.  **$Cov(X_i, X_i^2)$**: $Cov(X_i, X_i^2) = E[X_i \cdot X_i^2] - E[X_i]E[X_i^2] = E[X_i^3] - E[X_i]E[X_i^2]$. For a Gaussian distribution with mean 0, all odd central moments are zero. Thus, $E[X_i^3] = 0$. Also, we know $E[X_i] = 0$. Therefore, $Cov(X_i, X_i^2) = 0 - 0 \cdot E[X_i^2] = 0$.

3.  **$Cov(X_i^2, X_i)$**:
    This is symmetric, so $Cov(X_i^2, X_i) = Cov(X_i, X_i^2) = 0$.

4.  **$Var(X_i^2)$**: $Var(X_i^2) = E[(X_i^2)^2] - (E[X_i^2])^2 = E[X_i^4] - (E[X_i^2])^2$. We know from step 3 that $E[X_i^2] = \sigma^2$. For a Gaussian distribution with mean 0, the fourth central moment is $E[X_i^4] = 3(E[X_i^2])^2 = 3(\sigma^2)^2 = 3\sigma^4$. Therefore, $Var(X_i^2) = 3\sigma^4 - (\sigma^2)^2 = 3\sigma^4 - \sigma^4 = 2\sigma^4$.

Combining these results, the covariance matrix $Var(Y_i)$ is $Var(Y_i) = \begin{pmatrix} \sigma^2 & 0 \\\\ 0 & 2\sigma^4 \end{pmatrix}$

**State the asymptotic distribution** of $\sqrt{n}(\bar{Y}_n - E[Y_i])$ according to the MCLT

Based on the MCLT Theorem and our computations, the sample mean vector $\bar{Y}_n = (\frac{1}{n} \sum X_i, \frac{1}{n} \sum X_i^2)'$ is asymptotically normally distributed.

Specifically, we have: Mean vector $E[Y_i] = \begin{pmatrix} 0 \\ \sigma^2 \end{pmatrix}$, Covariance matrix $Var(Y_i) = \begin{pmatrix} \sigma^2 & 0 \\ 0 & 2\sigma^4 \end{pmatrix}$

Therefore, according to the MCLT:
$\sqrt{n}\left(\begin{pmatrix} \frac{1}{n} \sum X_i \\ \frac{1}{n} \sum X_i^2 \end{pmatrix} - \begin{pmatrix} 0 \\ \sigma^2 \end{pmatrix}\right) \xrightarrow{D} N\left(\begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} \sigma^2 & 0 \\ 0 & 2\sigma^4 \end{pmatrix}\right)$

This demonstrates that the vector $(\frac{1}{n} \sum X_i, \frac{1}{n} \sum X_i^2)'$ is asymptotically normal, with mean vector $(0, \sigma^2)'$ and covariance matrix $\frac{1}{n} \begin{pmatrix} \sigma^2 & 0 \\ 0 & 2\sigma^4 \end{pmatrix}$.

#### Compute Asymptotic Variance of Sample Variance

Compute the asymptotic variance of sigma_hat^2, using the Delta method based on the asymptotically normal vector derived in the previous step.

**Recall the Multivariate Central Limit Theorem (CLT)**

The problem states that $(\frac{1}{n} \sum X_i, \frac{1}{n} \sum X_i^2)'$ is asymptotically normal. Let $Y_n = (\frac{1}{n} \sum X_i, \frac{1}{n} \sum X_i^2)'$. Based on the provided information, we have $\sqrt{n}(Y_n - \mu_Y) \xrightarrow{D} N(0, \Sigma_Y)$ where:
*   $\mu_Y = E[Y_n] = (E[X_i], E[X_i^2])'$. Since $X_i \sim N(0, \sigma^2)$, we have $E[X_i] = 0$ and $E[X_i^2] = Var(X_i) + (E[X_i])^2 = \sigma^2 + 0^2 = \sigma^2$. Thus, $\mu_Y = (0, \sigma^2)'$.
*   $\Sigma_Y$ is the asymptotic covariance matrix of $Y_n$. For i.i.d. Gaussian r.v.s with mean 0 and variance $\sigma^2$:
*   $Var(X_i) = \sigma^2$
*   $Var(X_i^2) = E[(X_i^2 - E[X_i^2])^2] = E[(X_i^2 - \sigma^2)^2] = E[X_i^4] - (E[X_i^2])^2$. For a normal distribution with mean 0, $E[X_i^4] = 3\sigma^4$. So, $Var(X_i^2) = 3\sigma^4 - (\sigma^2)^2 = 2\sigma^4$.
*   $Cov(X_i, X_i^2) = E[X_i \cdot X_i^2] - E[X_i]E[X_i^2] = E[X_i^3] - 0 \cdot \sigma^2 = 0$ (due to symmetry of Gaussian distribution around 0, $E[X_i^3]=0$).

Therefore, the asymptotic covariance matrix $\Sigma_Y = \begin{pmatrix} Var(X_i) & Cov(X_i, X_i^2) \\ Cov(X_i^2, X_i) & Var(X_i^2) \end{pmatrix} = \begin{pmatrix} \sigma^2 & 0 \\ 0 & 2\sigma^4 \end{pmatrix}$. This matches the provided $\Sigma_Y$.

**Define the function $g$:**

From the problem statement, $\hat{\sigma}^2 = \frac{1}{n} \sum X_i^2 - (\frac{1}{n} \sum X_i)^2$. Let $Y_n = (Y_{n,1}, Y_{n,2})'$, where $Y_{n,1} = \frac{1}{n} \sum X_i$ and $Y_{n,2} = \frac{1}{n} \sum X_i^2$.

Then, we can define the function $g(y_1, y_2)$ such that $\hat{\sigma}^2 = g(Y_n)$; $g(y_1, y_2) = y_2 - y_1^2$.

At the true parameter values, $g(\mu_Y) = g(0, \sigma^2) = \sigma^2 - 0^2 = \sigma^2$, which is the true variance we are estimating.

**Compute the gradient of $g$:**

The gradient of $g(y_1, y_2)$ is $\nabla g(y_1, y_2) = \begin{pmatrix} \frac{\partial g}{\partial y_1} \\ \frac{\partial g}{\partial y_2} \end{pmatrix}$.

*   $\frac{\partial g}{\partial y_1} = \frac{\partial}{\partial y_1} (y_2 - y_1^2) = -2y_1$ ; $\frac{\partial g}{\partial y_2} = \frac{\partial}{\partial y_2} (y_2 - y_1^2) = 1$. So, $\nabla g(y_1, y_2) = (-2y_1, 1)'$.

Now, we evaluate the gradient at $\mu_Y = (0, \sigma^2)'$: $\nabla g(\mu_Y) = \nabla g(0, \sigma^2) = (-2 \cdot 0, 1)' = (0, 1)'$.

**Apply the Delta Method:**

The Delta method states that if $\sqrt{n}(Y_n - \mu_Y) \xrightarrow{D} N(0, \Sigma_Y)$, then $\sqrt{n}(g(Y_n) - g(\mu_Y)) \xrightarrow{D} N(0, (\nabla g(\mu_Y))' \Sigma_Y (\nabla g(\mu_Y)))$.

The asymptotic variance of $\sqrt{n}(\hat{\sigma}^2 - \sigma^2)$ is given by $Var_{asy} = (\nabla g(\mu_Y))' \Sigma_Y (\nabla g(\mu_Y))$. Substituting the calculated values: $Var_{asy} = (0, 1) \begin{pmatrix} \sigma^2 & 0 \\ 0 & 2\sigma^4 \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix}$. First, multiply the matrix by the column vector:
$\begin{pmatrix} \sigma^2 & 0 \\ 0 & 2\sigma^4 \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} \sigma^2 \cdot 0 + 0 \cdot 1 \\ 0 \cdot 0 + 2\sigma^4 \cdot 1 \end{pmatrix} = \begin{pmatrix} 0 \\ 2\sigma^4 \end{pmatrix}$. Now, multiply the row vector by the resulting column vector: $(0, 1) \begin{pmatrix} 0 \\ 2\sigma^4 \end{pmatrix} = 0 \cdot 0 + 1 \cdot 2\sigma^4 = 2\sigma^4$.

So, the asymptotic variance of $\sqrt{n}(\hat{\sigma}^2 - \sigma^2)$ is $2\sigma^4$.

**Interpret the result and comparison:**

The asymptotic distribution of $\hat{\sigma}^2$ is approximately $N(\sigma^2, \frac{2\sigma^4}{n})$.

Thus, the asymptotic variance of $\hat{\sigma}^2$ is $\frac{2\sigma^4}{n}$.

**Comparison with direct calculation (from 'Prove Consistency' subtask):**

In the 'Prove Consistency' subtask, we established that $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^n X_i^2 - (\frac{1}{n} \sum_{i=1}^n X_i)^2$. Since $E[X_i] = 0$, $E[(\frac{1}{n} \sum X_i)^2]$ is of order $O(1/n)$ and vanishes as $n \to \infty$. Therefore, for large $n$, $\hat{\sigma}^2 \approx \frac{1}{n} \sum_{i=1}^n X_i^2$. This is a consistent estimator. The variance of $\frac{1}{n} \sum X_i^2$ is $\frac{1}{n^2} \sum Var(X_i^2) = \frac{1}{n^2} \cdot n \cdot 2\sigma^4 = \frac{2\sigma^4}{n}$. This matches the result obtained using the Delta method. This confirms consistency.

**Simulate Gaussian Data**

Simulate a sample of $n$ i.i.d. Gaussian random variables with a known mean ($\mu=0$) and a specified true variance ($\sigma^2$). Define $n$, $\mu$, and $\sigma^2$ as parameters.

```python
import numpy as np

# Define the parameters:
n = 1000  # Sample size
mu = 0    # Known mean (as per problem description)
sigma_sq = 5 # True variance (choose a positive value)

# Calculate standard deviation from variance
sigma = np.sqrt(sigma_sq)

# Generate a sample of n i.i.d. Gaussian random variables
X = np.random.normal(loc=mu, scale=sigma, size=n)

# Store the generated sample in a variable X (already done above)

# Print parameters and a snippet of the generated sample for verification
print(f"Parameters: n={n}, mu={mu}, sigma_sq={sigma_sq} (sigma={sigma:.2f}) \n"
f"Generated sample X (first 10 elements): {X[:10]} \n"
f"Mean of generated sample: {np.mean(X):.4f} \n"
f"Variance of generated sample: {np.var(X):.4f}")
```

**Sample Variance**

Calculate the sample variance sigma_hat^2 using the generated sample X, given the known mean mu=0. The sample variance when the population mean mu=0 is known is sigma_hat^2 = (1/n) * Sum(Xi^2).


```python
import numpy as np

# Access the generated sample X and its size n from the previous step
# n, mu, sigma_sq, sigma, X are already defined in the kernel from the previous cell.

# Calculate the sum of the squares of all elements in the X array
sum_sq_X = np.sum(X**2)

# Divide this sum by n to compute the sample variance sigma_hat_sq
sigma_hat_sq = sum_sq_X / n

print(f"Calculated sample variance (sigma_hat^2): {sigma_hat_sq:.4f}")
```

**Monte Carlo for Sample Variance Distribution**

Perform a Monte Carlo simulation by repeatedly generating Gaussian data and calculating the sample variance sigma_hat^2. Visualize the empirical distribution of these sigma_hat^2 values using a histogram or Kernel Density Estimate (KDE) plot. Compare this empirical distribution to the theoretically derived asymptotic normal distribution N(sigma^2, (2*sigma^4)/n) with appropriate labels and legends.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Set a number of Monte Carlo simulations
num_simulations = 10000

# Create an empty list to store the calculated sigma_hat_sq values
sigma_hat_sq_mc_values = []

# Access n, mu, sigma, sigma_sq from previous steps
# n = 1000 Sample size; mu = 0 Known mean
# sigma_sq = 5 True variance ; sigma = np.sqrt(sigma_sq)

# Loop num_simulations times:
for _ in range(num_simulations):
    # a. Generate a new sample X_mc of n i.i.d. Gaussian random variables
    X_mc = np.random.normal(loc=mu, scale=sigma, size=n)
    # b. Calculate the sample variance sigma_hat_sq_mc
    sigma_hat_sq_mc = np.sum(X_mc**2) / n
    # c. Append sigma_hat_sq_mc to the list
    sigma_hat_sq_mc_values.append(sigma_hat_sq_mc)

# Convert to numpy array for easier handling
sigma_hat_sq_mc_values = np.array(sigma_hat_sq_mc_values)

plt.figure(figsize=(10, 5), dpi = 60)

# Use sns.histplot() to visualize the empirical distribution
sns.histplot(sigma_hat_sq_mc_values, stat='density', bins=50, color='skyblue', label='Empirical Distribution', kde=True)

# Calculate the theoretical mean and standard deviation for the asymptotic distribution
theoretical_mean_sigma_hat_sq = sigma_sq
theoretical_variance_sigma_hat_sq = (2 * (sigma_sq**2)) / n
theoretical_std_sigma_hat_sq = np.sqrt(theoretical_variance_sigma_hat_sq)

# Generate a range of x-values
x = np.linspace(min(sigma_hat_sq_mc_values) - 0.5, max(sigma_hat_sq_mc_values) + 0.5, 500)

# Calculate the PDF values for the theoretical normal distribution
pdf_values = norm.pdf(x, loc=theoretical_mean_sigma_hat_sq, scale=theoretical_std_sigma_hat_sq)

# Plot the theoretical PDF curve
plt.plot(x, pdf_values, color='red', linestyle='--', label=f'Theoretical N({theoretical_mean_sigma_hat_sq:.2f}, {theoretical_variance_sigma_hat_sq:.4f})')

# Add appropriate labels and a title
plt.xlabel('Sample Variance (sigma_hat^2)')
plt.ylabel('Density')
plt.title('Empirical vs. Theoretical Distribution of Sample Variance (Monte Carlo)')

plt.legend() ; plt.grid(True) ; plt.show()
```

**Visualize Asymptotic Normality of Vector Components**

Visualize the asymptotic normality of the vector (1/n Sum(Xi), 1/n Sum(Xi^2))' using scatter and contour plots. Perform a Monte Carlo simulation to generate multiple samples of this vector and plot their empirical distribution. Overlay contour lines representing the theoretical bivariate normal distribution, along with the true mean vector (0, sigma^2), with proper labels and legends.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import multivariate_normal

# Access n, mu, sigma, sigma_sq from previous steps (already in kernel)
# n = 1000
# mu = 0
# sigma_sq = 5
# sigma = np.sqrt(sigma_sq)

# Set the number of Monte Carlo simulations
num_simulations_vector = 10000

# Create an empty list to store the results of each simulation
simulated_vectors = []

# Loop for the specified number of simulations:
for _ in range(num_simulations_vector):
    # a. Generate a new sample X_mc of n i.i.d. Gaussian random variables
    X_mc_vector = np.random.normal(loc=mu, scale=sigma, size=n)
    # b. Calculate the first component of the vector: 1/n Sum(Xi)
    mean_X_mc = np.sum(X_mc_vector) / n
    # c. Calculate the second component of the vector: 1/n Sum(Xi^2)
    mean_X_sq_mc = np.sum(X_mc_vector**2) / n
    # d. Append the pair (mean_X_mc, mean_X_sq_mc) to list of simulated vectors
    simulated_vectors.append((mean_X_mc, mean_X_sq_mc))

# Convert the list of simulated vectors into a NumPy array
simulated_vectors = np.array(simulated_vectors)

# Define the theoretical mean vector mu_Y_theory
mu_Y_theory = np.array([mu, sigma_sq])

# Define the theoretical covariance matrix Sigma_Y_theory
# Note: This is covariance matrix for sample mean vector, so it's Sigma_Y / n
# Where Sigma_Y = [[sigma^2, 0], [0, 2*sigma^4]] from the Delta Method
Sigma_Y_theory = np.array([[sigma_sq, 0], [0, 2 * sigma_sq**2]]) / n

# Create a scatter plot of the simulated vectors
plt.figure(figsize=(10, 5), dpi =60)
sns.scatterplot(
    x=simulated_vectors[:, 0],
    y=simulated_vectors[:, 1],
    alpha=0.5,
    label='Empirical Samples of (1/n Sum(Xi), 1/n Sum(Xi^2))'
)

# Generate a grid of points for the theoretical PDF
x_min,x_max =simulated_vectors[:,0].min()-0.5, simulated_vectors[:,0].max() +0.5
y_min,y_max =simulated_vectors[:,1].min()-0.5, simulated_vectors[:,1].max() +0.5
x_grid = np.linspace(x_min, x_max, 100)
y_grid = np.linspace(y_min, y_max, 100)
X_mesh, Y_mesh = np.meshgrid(x_grid, y_grid)
pos = np.empty(X_mesh.shape + (2,))
pos[:, :, 0] = X_mesh
pos[:, :, 1] = Y_mesh

# Calculate the Probability Density Function (PDF) values
Z = multivariate_normal.pdf(pos, mean=mu_Y_theory, cov=Sigma_Y_theory)
# Overlay contour lines of the theoretical bivariate normal PDF
plt.contour(X_mesh, Y_mesh, Z, levels=5, colors='red', linestyles='--', linewidths=1.5)
# Add a distinct marker for the true mean vector
plt.scatter(mu_Y_theory[0], mu_Y_theory[1], color='green', marker='X', s=200, label='True Mean Vector (0, sigma^2)', zorder=5)
# Add appropriate labels and a title
plt.xlabel('1/n Sum(Xi)')
plt.ylabel('1/n Sum(Xi^2)')
plt.title('Empirical vs. Theoretical Asymptotic Distribution of Vector (1/n Sum(Xi), 1/n Sum(Xi^2))')
plt.legend() ; plt.grid(True) ; plt.show()
```

**Compare Empirical and Theoretical Asymptotic Variance**

Compare the empirical variance of sigma_hat^2 from the Monte Carlo simulation (from the 'Monte Carlo for Sample Variance Distribution' step) with its theoretically derived asymptotic variance, which is (2 * sigma^4) / n. Print both values for comparison.


```python
import numpy as np

# Calculate the empirical variance of sigma_hat_sq_mc_values
# sigma_hat_sq_mc_values is available from the 'Monte Carlo for Sample Variance Distribution' step
empirical_variance_sigma_hat_sq = np.var(sigma_hat_sq_mc_values)

# Calculate the theoretical asymptotic variance
# n and sigma_sq are available from previous steps
theoretical_variance_sigma_hat_sq_comparison = (2 * sigma_sq**2) / n

# Print both values for comparison
print(f"Empirical variance of sigma_hat^2 from Monte Carlo: {empirical_variance_sigma_hat_sq:.6f}")
print(f"Theoretical asymptotic variance of sigma_hat^2: {theoretical_variance_sigma_hat_sq_comparison:.6f}")
```

**Visualize 95% Confidence Interval Coverage**

Perform a Monte Carlo simulation to generate multiple samples. For each sample, calculate the 95% asymptotic confidence interval for sigma^2. Count how many of these intervals successfully contain the true value of sigma^2 to evaluate the empirical coverage probability. Visualize the coverage using a plot (e.g., showing several intervals and indicating which cover the true value).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Access n, mu, sigma, sigma_sq from previous steps (already in kernel)
# n = 1000 ; mu = 0 ; sigma_sq = 5 ; sigma = np.sqrt(sigma_sq)

# Set the number of Monte Carlo simulations for this coverage analysis
num_simulations_coverage = 10000
# Create list to store boolean values indicating coverage
coverage_results = []
# Create lists to store the lower and upper bounds of the confidence intervals
lower_bounds, upper_bounds = [], []

# Z-value for 95% confidence interval (alpha=0.05, alpha/2=0.025 in each tail)
z_alpha_half = norm.ppf(0.975)

# Loop for the specified number of simulations:
for i in range(num_simulations_coverage):
    # a. Generate a new sample of 'n' i.i.d. Gaussian r.v.s (X_coverage)
    X_coverage = np.random.normal(loc=mu, scale=sigma, size=n)
    # b. Calculate the sample variance sigma_hat_sq_coverage
    sigma_hat_sq_coverage = np.sum(X_coverage**2) / n
    # c. Calculate the Estimated Standard Error (ASE) for sigma_hat_sq_coverage
    ASE = np.sqrt((2 * (sigma_hat_sq_coverage**2)) / n)
    # e. Calculate the lower and upper bounds of the confidence interval
    lower_bound = sigma_hat_sq_coverage - z_alpha_half * ASE
    upper_bound = sigma_hat_sq_coverage + z_alpha_half * ASE
    # f. Check if the true sigma_sq falls within this interval
    is_covered = (lower_bound <= sigma_sq <= upper_bound)
    coverage_results.append(is_covered)
    # g. Append the lower_bound and upper_bound to their respective lists
    lower_bounds.append(lower_bound)
    upper_bounds.append(upper_bound)

# Calculate the empirical coverage probability
empirical_coverage_probability = np.mean(coverage_results)

# The empirical coverage probability
print(f"Number of simulations: {num_simulations_coverage} \n"
f"Empirical coverage probability: {empirical_coverage_probability:.4f} \n"
f"Theoretical coverage probability: 0.95")

# Create a plot to visualize a subset of the confidence intervals
plt.figure(figsize=(15, 10), dpi=60)
# Limit the number of intervals to display for better visualization
num_intervals_to_plot = 100

for i in range(min(num_intervals_to_plot, num_simulations_coverage)):
    color = 'green' if coverage_results[i] else 'red'
    plt.plot([lower_bounds[i], upper_bounds[i]], [i, i], color=color, alpha=0.7)

# Draw a horizontal line representing the true sigma^2 value
plt.axvline(x=sigma_sq, color='blue', linestyle='--', label=f'True sigma^2 = {sigma_sq}')

# Create custom legend handles for covered/not covered intervals
import matplotlib.patches as mpatches
covered_patch = mpatches.Patch(color='green', label='Interval Covers True Value')
not_covered_patch = mpatches.Patch(color='red', label='Interval Does Not Cover True Value')

plt.xlabel('Estimated sigma^2 Value')
plt.ylabel('Simulation Number')
plt.title(f'95% Asymptotic Confidence Interval Coverage (First {min(num_intervals_to_plot, num_simulations_coverage)} Simulations)')
plt.legend(handles=[covered_patch, not_covered_patch, plt.gca().lines[-1]])
plt.grid(True) ; plt.show()
```
