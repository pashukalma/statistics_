### **Concave Functions**

Concavity and Matrix Definiteness: A symmetric $2 \times 2$ matrix $A$ is negative semi-definite ($A \preceq 0$), meaning $x^T A x \leq 0$ for all $x \in \mathbb{R}^2$, if and only if: $\text{tr}(A) \leq 0 \quad \text{and} \quad \det(A) \geq 0$. This is equivalent to the condition on the eigenvalues $\lambda_1, \lambda_2$ of $A$: $\text{tr}(A) = \lambda_1 + \lambda_2$, $\det(A) = \lambda_1 \lambda_2$. The matrix is negative semi-definite if and only if $\lambda_1 \leq 0$ and $\lambda_2 \leq 0$.

Application to Functions: A twice-differentiable function $f(x,y)$ is ${concave}$ if its Hessian matrix $H$ is negative semi-definite for all $(x,y)$ in its domain:
\begin{equation}
H = \begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{pmatrix}
\end{equation}

### **Objective:**
Determine whether the functions a concave, convex, or neither
- $f_4: \mathbb{R}^2 \to \mathbb{R}$: $f_4(\theta_1, \theta_2) = -\theta_1^2 + \frac{1}{2}(\theta_1 - \theta_2)^2 - 3\theta_2^2$
- $f_5(\theta_1, \theta_2) = -\theta_1^4 - \theta_2^4 - (\theta_2 - \theta_1)^3, \quad \theta_1 < \theta_2$

#### **Analysis of Concavity for f_4**

Given the function $f_4: \mathbb{R}^2 \to \mathbb{R}$: $f_4(\theta_1, \theta_2) = -\theta_1^2 + \frac{1}{2}(\theta_1 - \theta_2)^2 - 3\theta_2^2$

Condition for Concavity
- A function $f: \Omega \subseteq \mathbb{R}^d \to \mathbb{R}$ is concave if its Hessian matrix, denoted as $H(f)$ or $\nabla^2 f$, is negative semi-definite (NSD) for all points in its domain.

Calculating the Gradient ($\nabla f_4$)
- First, we compute the first partial derivatives:$\frac{\partial f_4}{\partial \theta_1} = -2\theta_1 + (\theta_1 - \theta_2) = -\theta_1 - \theta_2$. $\frac{\partial f_4}{\partial \theta_2} = -(\theta_1 - \theta_2) - 6\theta_2 = -\theta_1 - 5\theta_2$
- The gradient vector is: $\nabla f_4(\theta_1, \theta_2) = \begin{bmatrix} -\theta_1 - \theta_2 \\ -\theta_1 - 5\theta_2 \end{bmatrix}$

Calculating the Hessian Matrix ($H_{f_4}$)
- The Hessian is the matrix of second partial derivatives: $H_{f_4} = \nabla^2 f_4 = \begin{bmatrix} \frac{\partial^2 f_4}{\partial \theta_1^2} & \frac{\partial^2 f_4}{\partial \theta_1 \partial \theta_2} \\ \frac{\partial^2 f_4}{\partial \theta_2 \partial \theta_1} & \frac{\partial^2 f_4}{\partial \theta_2^2} \end{bmatrix} = \begin{bmatrix} -1 & -1 \\ -1 & -5 \end{bmatrix}$


Testing for Negative Semi-DefinitenessWe evaluate the properties of the Hessian matrix:
- Trace: $\text{tr}(H_{f_4}) = -1 + (-5) = -6$
- Determinant: $\det(H_{f_4}) = (-1)(-5) - (-1)(-1) = 5 - 1 = 4$

Since the determinant is positive ($\det > 0$) and the trace is negative ($\text{tr} < 0$), both eigenvalues of the Hessian are negative. This confirms that the Hessian matrix is negative definite. Because the Hessian is negative definite for all $(\theta_1, \theta_2) \in \mathbb{R}^2$, the function $f_4$ is strictly concave.


#### **Analysis of Concavity for f_5**

Given the function $f_5(\theta_1, \theta_2) = -\theta_1^4 - \theta_2^4 - (\theta_2 - \theta_1)^3, \quad \text{for } (\theta_1, \theta_2) \in \mathbb{R}^2, \, \theta_1 < \theta_2$

Gradient (First Derivative Vector)

- The gradient $\nabla f_5(\theta_1, \theta_2)$ obtained by taking partial derivatives with respect to $\theta_1$ and $\theta_2$:$$\nabla f_5(\theta_1, \theta_2) = \begin{bmatrix} -4\theta_1^3 + 3(\theta_2 - \theta_1)^2 \\ -4\theta_2^3 - 3(\theta_2 - \theta_1)^2 \end{bmatrix}$$

Hessian Matrix (Second Derivative Matrix)

- The Hessian matrix $H_{f_5}$ (or $D^2 f_5$) constructed from the second-order partial derivatives:$$H_{f_5} = \begin{bmatrix} -12\theta_1^2 - 6(\theta_2 - \theta_1) & 6(\theta_2 - \theta_1) \\ 6(\theta_2 - \theta_1) & -12\theta_2^2 - 6(\theta_2 - \theta_1) \end{bmatrix}$$

Trace of the Hessian
- $\text{tr}(D^2 f_5) = -12\theta_1^2 - 6(\theta_2 - \theta_1) - 12\theta_2^2 - 6(\theta_2 - \theta_1)$
$\text{tr}(D^2 f_5) = -12\theta_1^2 - 12\theta_2^2 - 12(\theta_2 - \theta_1)$

Determinant of the Hessian
- $\det(D^2 f_5) = f_{\theta_1 \theta_1} f_{\theta_2 \theta_2} - (f_{\theta_1 \theta_2})^2$
$\det(D^2 f_5) = \left[ -12\theta_1^2 - 6(\theta_2 - \theta_1) \right] \left[ -12\theta_2^2 - 6(\theta_2 - \theta_1) \right] - \left[ 6(\theta_2 - \theta_1) \right]^2$
$\det(D^2 f_5) = 144\theta_1^2\theta_2^2 + 72\theta_1^2(\theta_2 - \theta_1) + 72\theta_2^2(\theta_2 - \theta_1) + 36(\theta_2 - \theta_1)^2 - 36(\theta_2 - \theta_1)^2$
$\det(D^2 f_5) = 144\theta_1^2\theta_2^2 + 72(\theta_1^2 + \theta_2^2)(\theta_2 - \theta_1)$



Since we are given the constraint $\theta_2 > \theta_1$, we observe
- Determinant: $\det(D^2 f_5) > 0$ because all terms are positive under the given constraint.
- Trace: $\text{tr}(D^2 f_5) < 0$ because all terms are negative under the given constraint.
- Because the determinant is positive and the trace is negative (implying both eigenvalues are negative), the Hessian is negative definite. Thus, $f_5$ is strictly concave on the domain $\theta_2 > \theta_1$.

```python
import numpy as np

from sympy import solve, symbols, diff, Matrix, simplify

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#### **f_4**
Function Definition: $f_4(\theta_1, \theta_2) = -\theta_1^2 + \frac{1}{2}(\theta_1 - \theta_2)^2 - 3\theta_2^2$ $ , $ Gradient: $\nabla f_4(\theta_1, \theta_2) = \begin{bmatrix} -\theta_1 - \theta_2 \\ -\theta_1 - 5\theta_2 \end{bmatrix}$ $ , $  Hessian Matrix: $H_{f_4} = \begin{bmatrix} -1 & -1 \\ -1 & -5 \end{bmatrix}$ $ ; $  Properties: $\text{tr}(H_{f_4}) = -6, $ , $ \det(H_{f_4}) = 4$


```python
''' Computes f4(theta1, theta2) = -theta1^2 - 0.5*(theta1 - theta2)^2 - 3*theta2^2 '''
def f4_fn(theta1, theta2):
    return -theta1**2 - 0.5 * (theta1 - theta2)**2 - 3 * theta2**2

''' Computes f5(theta1, theta2) = -theta1^4 - theta2^4 - (theta2 - theta1)^3
    If apply_constraint is True, returns NaN where theta1 >= theta2. '''
def f5_fn(theta1, theta2, apply_constraint=True):
    val = -theta1**4 - theta2**4 - (theta2 - theta1)**3
    if apply_constraint:
        return np.where(theta1 < theta2, val, np.nan)
    return val

# Example usage
print(f'f4(0, 0): {f4_fn(0, 0)}, f5(0, 1): {f5_fn(0, 1)}, f5(1, 0) constrained: {f5_fn(1, 0)}')
```

```python
# Define symbols
theta1, theta2 = symbols('theta1 theta2', real=True)

# Function f4
f4 = f4_fn(theta1, theta2)

# Compute Gradient and Hessian for f4
grad_f4 = [diff(f4, theta1), diff(f4, theta2)]
hessian_f4 = Matrix([[diff(g, t) for t in (theta1, theta2)] for g in grad_f4])

tr_f4 = hessian_f4.trace()
det_f4 = hessian_f4.det()

print('Analysis of f4 \n')
print(f'Hessian:\n{hessian_f4} \n'
      f'Trace: {tr_f4} \n'
      f'Determinant: {det_f4} \n'
      f'Is Concave (Trace <= 0 and Det >= 0): {tr_f4 <= 0 and det_f4 >= 0}')
```

#### **Find critical points**

Find the critical points for the functions $f_4(\theta_1, \theta_2) = -\theta_1^2 - \frac{1}{2}(\theta_1 - \theta_2)^2 - 3\theta_2^2$ by solving for where their gradients equal zero.

```python
''' Solve for critical points where grad_f4 == 0
grad_f4 was defined as [-3.0*theta1 + 1.0*theta2, 1.0*theta1 - 7.0*theta2] '''
cp_f4 = solve(grad_f4, (theta1, theta2))

print(f'Critical Points for f4 \n'
      f'(theta1, theta2) = ({cp_f4[theta1]}, {cp_f4[theta2]})\n')
```

#### **f_5**

Function Definition: $f_5(\theta_1, \theta_2) = -\theta_1^4 - \theta_2^4 - (\theta_2 - \theta_1)^3, \quad \theta_1 < \theta_2$ $ , $ $\nabla f_5 = \begin{bmatrix} -4\theta_1^3 + 3(\theta_2 - \theta_1)^2 \\ -4\theta_2^3 - 3(\theta_2 - \theta_1)^2 \end{bmatrix}$ $ , $ $H_{f_5} = \begin{bmatrix} -12\theta_1^2 - 6(\theta_2 - \theta_1) & 6(\theta_2 - \theta_1) \\ 6(\theta_2 - \theta_1) & -12\theta_2^2 - 6(\theta_2 - \theta_1) \end{bmatrix}$ $ , $ $\text{det}(H_{f_5}) = 144\theta_1^2\theta_2^2 + 72(\theta_1^2 + \theta_2^2)(\theta_2 - \theta_1) > 0 \text{ for } \theta_2 > \theta_1$

```python
# Function f5, theta1 < theta2 as per the problem constraints
f5 = -theta1**4 - theta2**4 - (theta2 - theta1)**3
#f5 = f5_fn(theta1, theta2)

# Compute Gradient and Hessian for f5
grad_f5 = [diff(f5, theta1), diff(f5, theta2)]
hessian_f5 = Matrix([[diff(g, t) for t in (theta1, theta2)] for g in grad_f5])

tr_f5 = simplify(hessian_f5.trace())
det_f5 = simplify(hessian_f5.det())

print(f'Analysis of f5 \n'
      f'Hessian:\n{hessian_f5} \n'
      f'Simplified Trace: {tr_f5} \n'
      f'Simplified Determinant: {det_f5} \n')

# Check conditions under theta2 > theta1
print(f'Verification for f5 (given theta2 > theta1): \n'
      f'Trace is sum of negative terms: -12*theta1**2 - 12*theta2**2 - 12*(theta2 - theta1) \n'
      f'Since theta2 > theta1, (theta2 - theta1) > 0, making Trace strictly negative. \n')
```

Find the critical points for $f_5(\theta_1, \theta_2) = -\theta_1^4 - \theta_2^4 - (\theta_2 - \theta_1)^3$ by solving for where their gradients equal zero.

```python
'''  Solve grad_f5 = 0, grad_f5 = [-4*theta1**3 + 3*(theta2 - theta1)**2, -4*theta2**3 - 3*(theta2 - theta1)**2] '''
cp_f5_all = solve(grad_f5, (theta1, theta2))

# Filter for real solutions where theta2 > theta1
cp_f5_filtered = [sol for sol in cp_f5_all if all(val.is_real for val in sol) and float(sol[1]) > float(sol[0])]

print("Critical Points for f5 (theta2 > theta1)")
if not cp_f5_filtered:
    print("No real critical points found within the constraint theta2 > theta1.")
else:
    for sol in cp_f5_filtered:
        print(f"(theta1, theta2) = ({float(sol[0])}, {float(sol[1])})")

print("\nAll real solutions without constraint:")
print([sol for sol in cp_f5_all if all(val.is_real for val in sol)])
```

##### **Mathematical Explanation: Hessian, Eigenvalues, and Concavity**

For a multivariable function $f(\mathbf{\theta})$, the **Hessian matrix** $H$ is the square matrix of second-order partial derivatives. The second-order condition for concavity states that a twice-differentiable function is concave on a convex set if and only if its Hessian is **negative semi-definite (NSD)** for all points in that set.

- Negative Semi-Definiteness (NSD)
A symmetric matrix $H$ is NSD if for any non-zero vector $\mathbf{v}$, the quadratic form satisfies:
$$\mathbf{v}^T H \mathbf{v} \le 0$$
This condition ensures that the function's curvature is non-positive in every direction.

- Relationship with Eigenvalues
For symmetric matrices like the Hessian, the definiteness is directly determined by its eigenvalues ($\lambda_i$):
*   **Negative Semi-Definite**: $H$ is NSD if and only if all eigenvalues are non-positive ($\lambda_i \le 0$).
*   **Negative Definite**: $H$ is negative definite (implying strict concavity) if all eigenvalues are strictly negative ($\lambda_i < 0$).

- Geometric Interpretation of Curvature
The eigenvalues represent the function's second derivatives (curvature) along the directions of the corresponding eigenvectors. If an eigenvalue is negative, the function 'curves downward' in that direction. If all directions curve downward or are flat (NSD), the surface forms a 'cap' shape, which is the definition of concavity.

- Application to $f_4$ and $f_5$
*   **For $f_4$**: The Hessian was constant with a negative trace (-10) and a positive determinant (20). In a $2 \times 2$ matrix, since $Tr(H) = \lambda_1 + \lambda_2$ and $Det(H) = \lambda_1 \lambda_2$, these values imply $\lambda_1, \lambda_2 < 0$, confirming **strict concavity**.
*   **For $f_5$**: The Hessian is not constant. Under the constraint $\theta_1 < \theta_2$, we can analyze the trace and determinant expressions derived previously to determine the specific regions where the eigenvalues remain non-positive, identifying where the function maintains its concave property.
