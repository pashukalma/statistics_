#### **Asymptotic Variance Comparison Cauchy($\mu, 1$)**

Compare the asymptotic variance for the sample mean and sample median when $X_1, \dots, X_n \overset{i.i.d.}{\sim} \text{Cauchy}(\mu, 1)$. For $X \sim \text{Cauchy}(\mu, 1)$, we have the probability density function (PDF) and cumulative distribution function (CDF) as follows:
- $f_X(x) = \frac{1}{\pi(1 + (x - \mu)^2)}$
- $F_X(x) = \frac{1}{2} + \frac{1}{\pi} \arctan(x - \mu)$
- $F_X^{-1}(t) = \tan\left(\pi \left(t - \frac{1}{2}\right)\right) + \mu$

Notice that these are both continuous on $\mathbb{R}$.1.

**Sample Median**

Using the general result for the asymptotic distribution of the sample median, where $\sqrt{n}(\hat{\mu}_{\text{med}} - \mu) \xrightarrow{d} {N}\left(0, \frac{1}{4f(\mu)^2}\right)$, we find
- $f(\mu) = \frac{1}{\pi(1 + (\mu - \mu)^2)} = \frac{1}{\pi}$
- The asymptotic variance is $\sigma^2_{\text{med}} = \frac{1}{4 \left(\frac{1}{\pi}\right)^2} = \frac{\pi^2}{4} \approx 2.467$.

**Sample Mean**

For the sample mean $\bar{X}_n$
- $\text{AsymVar}(\bar{X}_n) = n \cdot \text{Var}(\bar{X}_n) = \text{Var}(X_i) = \infty$

**Analysis**

The Cauchy distribution is a "pathological" case in statistics because it has such heavy tails that its mean and variance are undefined (infinite).
- The Sample Mean: Because the variance is infinite, the Law of Large Numbers does not apply in the usual way. In fact, the sample mean of a Cauchy distribution has the exact same distribution as a single observation; it never "settles down" as $n$ increases.
- The Sample Median: The median remains a robust and well-defined estimator with a finite asymptotic variance, proving it is far superior to the mean for this specific distribution.

**Sample Mean and Median Distributions for Cauchy Data**

Generate 10,000 samples of size 100 from a Cauchy(0, 1) distribution, calculate the sample mean and median for each sample, and store these values to observe their empirical distributions.

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

num_simulations = 10000
sample_size = 100

cauchy_sample_means = []
cauchy_sample_medians = []

for _ in range(num_simulations):
    # Generate samples from Cauchy(0, 1) distribution
    sample = np.random.standard_cauchy(size=sample_size)

    # Calculate mean and median. Note: Mean for Cauchy distribution is undefined, but we calculate it anyway to observe its behavior
    mean_val = np.mean(sample)
    median_val = np.median(sample)

    # Append to lists
    cauchy_sample_means.append(mean_val)
    cauchy_sample_medians.append(median_val)

# Convert lists to NumPy arrays
cauchy_sample_means = np.array(cauchy_sample_means)
cauchy_sample_medians = np.array(cauchy_sample_medians)

print(f"Generated {num_simulations} sample means and medians, each from a sample size of {sample_size} for Cauchy data. \n"
f"First 5 Cauchy sample means: {cauchy_sample_means[:5]} \n"
f"First 5 Cauchy sample medians: {cauchy_sample_medians[:5]}")
```

Visualize the distributions of the sample means and medians for Cauchy data, that shows the instability of the sample mean for Cauchy distributions.


```python
plt.figure(figsize=(12, 6), dpi =60)

plt.hist(cauchy_sample_means, bins=50, density=True, alpha=0.6, color='skyblue', label='Cauchy Sample Means')
plt.hist(cauchy_sample_medians, bins=50, density=True, alpha=0.6, color='lightcoral', label='Cauchy Sample Medians')

plt.title('Distribution of Sample Means and Sample Medians (Cauchy Data)')
plt.xlabel('Value') ; plt.ylabel('Density')
plt.legend() ; plt.grid(True) ; plt.show()
```

**The Problem: Infinite Variance**

In a Cauchy distribution, the integral used to calculate variance diverges.
- Intuition: Outliers are not just possible; they are expected and massive.
- Visual Difference: While a Normal distribution curves quickly toward zero, the Cauchy distribution stays "propped up" further from the center.

The Estimator Showdown

When you try to find the "center" of a Cauchy dataset, the tool you choose changes everything:

Practical Implications
- The Mean is Dangerous: For heavy-tailed data, using the mean leads to "garbage" results and invalid hypothesis tests. It never "averages out" the noise.
- The Median is Robust: It is the reliable choice for central tendency when your data is prone to wild swings.
- The "Middle Ground": This is why statisticians use M-estimators (like Huber loss). These act like the mean for small errors but switch to acting like the median for large outliers, giving you the best of both worlds.

**Distribution Comparison: Mean vs. Median**

1. Efficiency vs. Robustness
Efficiency: A measure of how "tight" an estimator is around the true value (lower variance = higher efficiency).

Robustness: The ability of an estimator to ignore outliers. The Mean is zero-percent robust (one outlier can pull it to infinity), while the Median is highly robust because it cares about rank, not magnitude.

2. The Failure of the Mean
In a Cauchy distribution, the tails are so thick that the integral for the expected value does not converge. Because the population mean doesn't exist, the sample mean wanders aimlessly no matter how much data you collect. The Law of Large Numbers and Central Limit Theorem do not apply here.

#### Infinite Variance and its Impact on the Sample Mean (Cauchy Distribution)

**Concept of 'Infinite Variance' in the Context of the Cauchy Distribution:**
The Cauchy distribution is a peculiar type of probability distribution often referred to as a "heavy-tailed" distribution. Unlike many other distributions (e.g., Normal, Laplace), the Cauchy does not have a finite mean or variance (it means that the integral used to calculate the variance diverges). More intuitively, this implies that the probability of observing extremely large or small values (outliers) is much higher in a Cauchy compared to distributions like the Normal distribution. These extreme values are so probable that they disproportionately influence any attempt to calculate a central tendency or dispersion using methods sensitive to such extremes.

**Impact on the Sample Mean's Behavior as an Estimator:**
The sample mean, by definition, is highly sensitive to every data point. For distributions with finite variance, the LLN states that as the sample size increases, the sample mean will converge to the true population mean. Similarly, the CLT tells us about the distribution of the sample mean. However, for the Cauchy distribution:

*   **Instability:** Because of the high probability of extreme outliers, even a single large outlier can drastically pull the sample mean away from the true center of the distribution (which for Cauchy(0,1) is 0). As we saw, the distribution of cauchy_sample_means is extremely spread out, with many values far from zero, despite the underlying population having a peak at zero. This means the sample mean does not stabilize as the sample size increases; adding more data points doesn't necessarily make it a better estimate; it just adds more opportunities for extreme values to skew the result.

*   **Lack of Convergence:** Due to the infinite variance (and undefined mean), the sample mean of a Cauchy distribution does not converge to any fixed value as the sample size goes to infinity. In fact, it can be shown that the sample mean of *n* observations from a Cauchy distribution is itself a Cauchy distribution with the same parameters as the original distribution. This means the distribution of the sample mean does not become concentrated around a single point; it remains as spread out as the original data, regardless of how much data we collect.

**Contrast with the Behavior of the Sample Median for the Cauchy Distribution:**
In stark contrast to the sample mean, the sample median for a Cauchy distribution behaves much more stably:

*   **Stability and Convergence:** The sample median is a robust estimator because it is less sensitive to extreme values. It represents the middle value of the sorted data. For the Cauchy distribution, the sample median *does* converge to the true location parameter (which is 0 for Cauchy(0,1)) as the sample size increases. Our simulation visually demonstrated this: the distribution of cauchy_sample_medians was much narrower and more concentrated around zero than the cauchy_sample_means.

**Practical Implications of Using Sample Mean vs. Sample Median for Heavy-Tailed Distributions:**
The practical implications are significant:

*   **For the Sample Mean:** Using the sample mean to estimate the central tendency of a heavy-tailed distribution like Cauchy is highly problematic and can lead to misleading conclusions. Its instability means that the estimate will vary wildly from sample to sample, even for large sample sizes, and it will not converge to the true center. Any statistical inference based on the sample mean (e.g., confidence intervals, hypothesis tests) would be invalid.

*   **For the Sample Median:** The sample median is a much more appropriate and reliable estimator for the central tendency of heavy-tailed distributions. It is robust to outliers and provides a stable, converging estimate of the location parameter. When dealing with data that is suspected to come from a distribution with heavy tails or is known to contain significant outliers, the median should generally be preferred over the mean as a measure of central tendency. This highlights the importance of robust statistics, such as M-estimation with Huber loss, which offer a compromise between the efficiency of the mean for normal data and the robustness of the median for outlier-prone data.

#### Compare Asymptotic Variances of Laplace and Cauchy Distributions


**Laplace(0, 1) Distribution:**
*   **Recall Variances:** From previous calculations, for a Laplace(0, 1) distribution with a sample size of 100, we found:
    *   Empirical variance of Sample Means: 0.019730 (Theoretical: 0.020000). Empirical variance of Sample Medians: 0.011663 (Theoretical: 0.010000)
*   **Tail Characteristics and Efficiency:** The Laplace distribution is known for its **heavier tails** compared to the Normal distribution. In this case, the sample median consistently showed a smaller empirical and theoretical asymptotic variance than the sample mean. This indicates that for Laplace data, the **sample median is a more efficient estimator** than the sample mean. Its greater efficiency stems from its robustness to the heavier tails, as it focuses on the central tendency rather than being disproportionately affected by extreme values.

**Cauchy(0, 1) Distribution:**
*   **Sample Mean Behavior:** The Cauchy distribution is characterized by **extremely heavy tails**, so heavy that its population mean and variance are **undefined**. As observed in the simulation, the distribution of cauchy_sample_means is **highly unstable and extremely spread out**, with values ranging over a vast scale. This empirical observation directly reflects the theoretical fact that the sample mean for Cauchy data does not converge to a stable value (it does not satisfy the conditions for the Central Limit Theorem) and its variance is infinite. Therefore, the sample mean is a completely inappropriate and inefficient estimator for Cauchy-distributed data.
*   **Sample Median Behavior:** In stark contrast, the distribution of cauchy_sample_medians was much **more stable and concentrated around 0**, which is the true median (and mode) of the Cauchy(0, 1) distribution. For the Cauchy distribution, the sample median possesses a **finite asymptotic variance**. This makes the median a **robust and efficient estimator** for Cauchy data, successfully capturing the central tendency where the mean fails entirely. The median's ability to resist the influence of extreme values (outliers) makes it invaluable for such heavy-tailed distributions.

**Implications for Estimator Efficiency and Robustness:**
*   **Efficiency** refers to how close an estimator's distribution is to the true parameter, often quantified by its variance. A smaller variance implies higher efficiency. For Laplace data, the median was more efficient than the mean. For Cauchy data, the median was vastly more efficient, as the mean was effectively useless.
*   **Robustness** refers to an estimator's insensitivity to deviations from assumptions (like the presence of outliers or heavy tails). The sample mean is highly sensitive to outliers and heavy tails, becoming inefficient or even undefined. The sample median, by definition, is robust to such issues, as it is only concerned with the rank of the data points, not their magnitude. M-estimators, like those using Huber loss, offer a tunable compromise, providing robustness while still incorporating some information from larger residuals, balancing efficiency and robustness.

**Conclusion on Estimator Preference:**
*   **Normal-like (thin-tailed) data:** The **sample mean** is generally the most efficient estimator.
*   **Laplace-like (moderately heavy-tailed) data:** The **sample median** or robust M-estimators (e.g., Huber loss with appropriate $\delta$) are generally more efficient and robust than the sample mean.
*   **Cauchy-like (very heavy-tailed) data:** The **sample median** or other highly robust estimators (like trimmed means or specific M-estimators) are essential, as the sample mean is unreliable. The mean fails entirely due to the absence of a finite first moment.
