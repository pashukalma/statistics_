### The Monty Hall Problem: A Proof
A prize is placed at random behind one of three doors. You, the contestant, pick a door (let's assume you always pick Door 1). Monty Hall, who knows where the prize is, then opens one of the *other two* doors, always revealing an empty door. He then offers you the choice to either keep your original door or switch to the remaining unopened door. Should you stay or switch?

**Intuition vs. Reality:**

Many people intuitively believe it doesn't matter whether you stay or switch, thinking the odds become 50/50. However, the correct answer is that you should always switch. Let's prove this by carefully defining the sample space and relevant events.


**Defining the Sample Space $\Omega$**

As requested, we define the sample space $\Omega$ as the set of pairs $(\omega_1, \omega_2)$, where:
- $\omega_1 \in \{1, 2, 3\}$ represents the door behind which the prize is hidden.
- $\omega_2 \in \{1, 2, 3\}$ represents the door that Monty Hall opens.

So, $\Omega = \{ (\omega_1, \omega_2) : \omega_1, \omega_2 \in \{1,2,3\} \}$.

There are 9 possible outcomes in this initial definition of the sample space, however, we incorporate the rules of the game to determine the *valid* outcomes and their probabilities. We assume the prize is equally likely to be behind any door, so $P(\omega_1=1) = P(\omega_1=2) = P(\omega_1=3) = 1/3$.

**Rules constraining Monty's choice ($\omega_2$)**
1. You pick Door 1. (This is a fixed player action, not part of the random event $\omega_1$ or $\omega_2$).
2. Monty never opens your chosen door (Door 1).So $\omega_2 \neq 1$.
3. Monty always opens an empty door. So $\omega_2 \neq \omega_1$.

Given these rules and your fixed choice of Door 1, let's refine the possible $(\omega_1, \omega_2)$ pairs and their probabilities:

**Valid Outcomes and Probabilities**

Let's analyze the valid outcomes $(\omega_1, \omega_2)$ and their probabilities, considering the player always picks Door 1:

| Prize Door ($\omega_1$) | Monty's Options for $\omega_2$ (Empty, Not Door 1) | Valid Outcomes $(\omega_1, \omega_2)$ | $P(\omega_1)$ | $P(\omega_2 | \omega_1)$ | $P(\omega_1, \omega_2)$ |
|:---------------------:|:-----------------------------------------------------:|:--------------------------------------:|:---------------:|:-------------------------:|:--------------------------:|
| 1 (You picked prize)    | Doors 2 or 3 (both empty)                           | (1,2), (1,3)                           | 1/3             | 1/2 (Monty chooses randomly) | 1/6 (1/3 * 1/2)           |
| 2 (Prize behind Door 2) | Door 3 (only empty door not picked)                 | (2,3)                                  | 1/3             | 1 (Monty *must* open Door 3) | 1/3 (1/3 * 1)             |
| 3 (Prize behind Door 3) | Door 2 (only empty door not picked)                 | (3,2)                                  | 1/3             | 1 (Monty *must* open Door 2) | 1/3 (1/3 * 1)             |

**Summary of Valid Outcomes and Probabilities:**
*   $(\omega_1=1, \omega_2=2)$: Prize at 1, Monty opens 2. Probability = 1/3*1/2=1/6.
*   $(\omega_1=1, \omega_2=3)$: Prize at 1, Monty opens 3. Probability = 1/6.
*   $(\omega_1=2, \omega_2=3)$: Prize at 2, Monty opens 3. Probability = 1/3.
*   $(\omega_1=3, \omega_2=2)$: Prize at 3, Monty opens 2. Probability = 1/3.

Notice the sum of these probabilities is $1/6 + 1/6 + 1/3 + 1/3 = 1$. These are the only possible scenarios for the game under the given rules.


**Analyzing the Strategies: Stay vs. Switch**

Now, let's define the events for winning with each strategy, given your initial choice of Door 1.

**Strategy 1: STAY** (keep Door 1). You win if the prize is behind Door 1 (i.e., $\omega_1 = 1$).The event "Win by Staying" corresponds to the outcomes where $\omega_1=1$. This includes outcomes $(1,2)$ and $(1,3)$.
- $P(\text{Win by Staying}) = P((\omega_1=1, \omega_2=2)) + P((\omega_1=1, \omega_2=3)) = 1/6 + 1/6 = 2/6 = \mathbf{1/3}$.

**Strategy 2: SWITCH** (to the other unopened door). You win if the prize is *not* behind Door 1 (i.e., $\omega_1 \neq 1$). In this case, Monty *must* have opened the *other* empty door, leaving the prize behind the remaining unopened door.

Let's consider the scenarios:

*   **Scenario A:** Prize is behind Door 2 (i.e., $\omega_1 = 2$). Monty *must* open Door 3. You picked Door 1. If you switch, you switch from Door 1 to Door 2, and you win. This corresponds to outcome $(2,3)$.

*   **Scenario B:** Prize is behind Door 3 (i.e., $\omega_1 = 3$). Monty *must* open Door 2. You picked Door 1. If you switch, you switch from Door 1 to Door 3, and you win. This corresponds to outcome $(3,2)$.

*   **Scenario C:** Prize is behind Door 1 (i.e., $\omega_1 = 1$). Monty opens either Door 2 or Door 3. If you switch, you switch from Door 1 to an empty door (either 3 or 2), and you lose. This corresponds to outcomes $(1,2)$ and $(1,3)$.

The event "Win by Switching" corresponds to the outcomes where $\omega_1=2$ or $\omega_1=3$. This includes outcomes $(2,3)$ and $(3,2)$.

$P(\text{Win by Switching}) = P((\omega_1=2, \omega_2=3)) + P((\omega_1=3, \omega_2=2)) = 1/3 + 1/3 = 2/3$.



**Conclusion**

Comparing the probabilities:
*   $P(\text{Win by Staying}) = \mathbf{1/3}$
*   $P(\text{Win by Switching}) = \mathbf{2/3}$

Therefore, by switching doors, you double your chances of winning the prize. The intuition that it doesn't matter is incorrect because Monty Hall's action provides new information; he eliminates an empty door that is *not* your original choice.

### Coin toss

Coin toss until we get exactly two Heads. Sample space and ad the prob. that exactly k tosses are required.

**The sample space** $\Omega$ consists of sequences of coin tosses ending with HH, where HH does not appear earlier.
{HH} (2 tosses), {THH} (3 tosses), {HTHH, TTHH} (4 tosses), {TTTHH, HTTHH, THTHH} (5 tosses)

Each sequence of length $k$ has a probability of $(1/2)^k$.

**Probability that Exactly k Tosses are Required, P(X=k)**

Let $P(X=k)$ be the probability that exactly $k$ tosses are required. This probability follows a pattern related to Fibonacci numbers ($F_n$, where $F_0=0, F_1=1, \dots$).

The number of sequences of length $k$ that end in HH for the first time is $F_{k-1}$.
Given each sequence has probability $(1/2)^k$, we have:

$$P(X=k) = \frac{F_{k-1}}{2^k}\quad \text{for } k \ge 2$$

Verification: k=2 (HH): $P(X=2) = F_1/2^2 = 1/4$; k=3 (THH): $P(X=3) = F_2/2^3 = 1/8$; k=4 (TTHH, HTHH): $P(X=4) = F_3/2^4 = 2/16 = 1/8$.

**Expected Number of Tosses, E(X)**

The expected number of tosses, $E(X)$, is given by $E(X) = \sum_{k=2}^{\infty} k \cdot P(X=k) = \sum_{k=2}^{\infty} k \cdot \frac{F_{k-1}}{2^k}$

A state-based approach shows that $E(X) = \mathbf{6}$.

### Coin Flip Simulation

We will simulate a series of coin flips and observe how the proportion of heads approaches the true probability $p$ as the number of flips increases.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_coin_flips(p, n_flips):
    """Simulates coin flips and returns the cumulative proportion of heads."""
    # Generate random numbers between 0 and 1
    flips = np.random.rand(n_flips)

    # Determine if each flip is a 'head' (True if random number < p)
    heads = flips < p

    # Calculate the cumulative sum of heads
    cumulative_heads = np.cumsum(heads)

    # Calculate the proportion of heads at each step
    proportions = cumulative_heads / np.arange(1, n_flips + 1)

    return proportions
```

#### Simulation for $p = 0.25$

```python
n = 1000
p1 = 0.25
proportions1 = simulate_coin_flips(p1, n)

plt.figure(figsize=(10, 5), dpi=60)
plt.plot(range(1, n + 1), proportions1, label=f'Proportion of Heads (p={p1})')
plt.axhline(y=p1, color='r', linestyle='--', label=f'True Probability (p={p1})')
plt.xlabel('Number of Flips (n)') ; plt.ylabel('Proportion of Heads')
plt.title(f'Coin Flip Simulation: Proportion of Heads vs. Number of Flips {p1}')
plt.legend() ; plt.grid(True) ; plt.show()
```

#### Sampling with replacement

Consider a container with three types of items. We assume there are so many items we can consider removing a few insignificant and we will not change the probability of getting another item. The probability of picking the first item is 45%, 2nd is 30% and 3rd is 25%. What is the chance of getting a mix of 3 items of first type, 2 of second type and 1 of third in a random selection of 6 items?

This is known as sampling with replacement - that means that we select 6 one at a time recording their type and returning them to the pool. This is a multinomial problem with: n = 6 (total number of selections), k = 3 (number of item types) and probabilities p1 = 0.45, p2 = 0.30, p3 = 0.25. What we are looking for here is $P(X1 = 3, X2 = 2, X3 = 1)$.

The multinomial probability formula is $P(X_1=x_1, \dots, X_k=x_k) = \frac{n!}{x_1!x_2!\dots x_k!} p_1^{x_1} p_2^{x_2} \dots p_k^{x_k}$

Applying this formula, $P(X_1 = 3, X_2 = 2, X_3 = 1) = \frac{6!}{3!2!1!} \cdot (0.45)^3 \cdot (0.3)^2 \cdot (0.25)^1$. Calculating the coefficients, $\frac{6!}{3!2!1!} = \frac{6 \times 5 \times 4}{2 \times 1} = 6 \times 5 \times 2 = 60$. Calculating the probabilities, $(0.45)^3 \approx 0.091125$, $(0.3)^2 = 0.09$, $(0.25)^1 = 0.25$

Multiplying these values, the probability is approximately **0.123**.

### Poisson Distribution - Post Office

Assume the number of customers who enter the post office today is given by a Poisson distribution with a rate parameter ($\lambda$) of $1/3$. What is the probability that at least one customer enters the post office today?

Let X denote the number of customers who enter the post office today. We are looking for $P(X \ge 1)$.The Poisson probability mass function is given by $P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$.

Calculating $P(X \ge 1)$ directly would involve summing $P(X=1) + P(X=2) + \dots$ up to infinity, which is cumbersome. It's much easier to calculate the complement: $P(X \ge 1) = 1 - P(X = 0)$.

First, let's calculate $P(X=0)$ with $\lambda = 1/3$, $P(X=0) = \frac{(1/3)^0 e^{-1/3}}{0!} = \frac{1 \cdot e^{-1/3}}{1} = e^{-1/3}$. We find $P(X \ge 1)$ $P(X \ge 1) = 1 - P(X = 0) = 1 - e^{-1/3}$. Calculating the numerical value $e^{-1/3} \approx 0.72, $P(X \ge 1) = 1 - 0.72 \approx 0.283$. Therefore, the probability that at least one customer enters the post office today is approximately 0.283.

### Poisson Distribution - IKEA

A company knows from past experience that the number of a particular item sold at a particular store each day has a Poisson distribution with a rate parameter ($\lambda$) of $1/2$. What is a good approximation for the probability that this store sells more than 3 items today?

Let $X$ denote the number of items sold today. We are looking for $P(X > 3)$. It's easier to calculate the complement: $P(X > 3) = 1 - P(X \le 3)$. We need to sum the probabilities for $P(X=0)$, $P(X=1)$, $P(X=2)$, and $P(X=3)$, with $\lambda = 1/2$.

The Poisson probability mass function is given by $P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$.
- $P(X=0) = \frac{(1/2)^0 e^{-1/2}}{0!} = e^{-1/2}$, $P(X=1) = \frac{(1/2)^1 e^{-1/2}}{1!} = \frac{1}{2} e^{-1/2}$, $P(X=2) = \frac{(1/2)^2 e^{-1/2}}{2!} = \frac{1}{8} e^{-1/2}$, $P(X=3) = \frac{(1/2)^3 e^{-1/2}}{3!} = \frac{1}{48} e^{-1/2}$

So, $P(X \le 3) = e^{-1/2} + \frac{1}{2} e^{-1/2} + \frac{1}{8} e^{-1/2} + \frac{1}{48} e^{-1/2} = e^{-1/2} \left(\frac{79}{48}\right)$

Numerically, $e^{-1/2} \approx 0.61$

$P(X \le 3) \approx 0.61 \times \frac{79}{48} \approx 0.61 \times 1.65 \approx 0.9984$

Therefore, the probability that the store sells more than 3 items today is $P(X > 3) = 1 - P(X \le 3) \approx 1 - 0.9984 = 0.0016$$

**Remark about Poisson distribution:**

It might seem strange that the mean and variance are both equal to $\lambda$ for a Poisson distribution. This does not imply that the mean and variance have the same units. The mean ($\\lambda$) and the standard deviation ($\sqrt{\lambda}$) have the same units, not the mean and variance.

Consider the Poisson Probability Mass Function: $P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$. If $\lambda$ had units, then the term $e^{-\lambda}$ would involve summing terms with different units ($1 + (-\lambda) + (-\lambda)^2/2! + \dots$), which is mathematically inconsistent. Therefore, $\lambda$ must be dimensionless, and so are the mean and variance of a Poisson distribution.

### German Tank Problem (Simpler Version)

Let $N$ be the total number of tanks produced, with $N$ being the largest serial number and 1 being the smallest. We assume we record $k$ serial numbers from tanks we observe, and $m$ is the highest serial number observed among these $k$ observations.

Our goal is to estimate the unknown $N$ given our observed value of $m$ and our number of observations $k$. Our estimate for $N$ must naturally be at least as large as $m$.



#### Approach

We will approach this problem by considering the expected value of the maximum observed serial number, $M$, given $N$ and $k$. We will then reverse this relationship to use our observed $m$ to estimate $N$.

Let $M$ be the random variable representing the highest serial number observed from $k$ observations, sampled without replacement from the set $[1, 2, ..., N]$.

Probability Mass Function of M, $P(M=m)$:

To find $P(M=m)$, we need to count the number of ways to choose k distinct serial numbers from [1, ..., N] such that m is the largest. We'll fix k and N for this discussion.

1.  **Total ways to choose k distinct numbers from N:** This is given by the binomial coefficient $C(N, k) = \binom{N}{k}$.

2.  **Ways to choose k distinct numbers where m is the largest:** If m is the largest serial number observed, then m must be one of the k chosen numbers. The remaining k-1 numbers must be chosen from the set [1, 2, ..., m-1], as no number larger than m can be chosen. The number of ways to do this is $C(m-1, k-1) = \binom{m-1}{k-1}$.

Therefore, the probability that the maximum observed serial number is m is $P(M=m) = \frac{\binom{m-1}{k-1}}{\binom{N}{k}}$. This probability distribution is valid for m ranging from k to N (since we need at least k numbers to have a maximum, and m cannot exceed N). The sum of these probabilities over all possible m values must equal 1, $\sum_{m=k}^{N} \frac{\binom{m-1}{k-1}}{\binom{N}{k}} = 1$.



#### Expected Value of M

$E[M]$ is given by $E[M] = \sum_{m=k}^{N} m \cdot P(M=m) = \sum_{m=k}^{N} m \cdot \frac{\binom{m-1}{k-1}}{\binom{N}{k}}$. Let's expand the combinatorial terms, $C(m-1, k-1) = \frac{(m-1)!}{(k-1)!(m-k)!}$ and $C(N, k) = \frac{N!}{k!(N-k)!}$. So $E[M] = \sum_{m=k}^{N} m \cdot \frac{(m-1)!}{(k-1)!(m-k)!} \cdot \frac{k!(N-k)!}{N!}$. We know that $m \cdot (m-1)! = m!$, and $k! = k \cdot (k-1)!$. Substituting these $E[M] = \sum_{m=k}^{N} \frac{m!}{(k-1)!(m-k)!} \cdot \frac{k(k-1)!(N-k)!}{N!}$.
$E[M] = \frac{k(N-k)!}{N!} \sum_{m=k}^{N} \frac{m!}{(m-k)!}$

This simplification is challenging directly. A more elegant way to sum this series involves a combinatorial identity related to $\sum_{j=r}^{n} \binom{j}{r} = \binom{n+1}{r+1}$.

Using another identity $m \binom{m-1}{k-1} = k \binom{m}{k}$: (This is because $m \frac{(m-1)!}{(k-1)!(m-k)!} = k \frac{m!}{k!(m-k)!}$)

$E[M] = \frac{1}{\binom{N}{k}} \sum_{m=k}^{N} k \binom{m}{k} = \frac{k}{\binom{N}{k}} \sum_{m=k}^{N} \binom{m}{k}$

Now, applying the identity $\sum_{j=r}^{n} \binom{j}{r} = \binom{n+1}{r+1}$ (with $j=m$, $r=k$, $n=N$) $E[M] = \frac{k}{\binom{N}{k}} \binom{N+1}{k+1}$.

Expanding the binomial coefficient $\binom{N+1}{k+1} = \frac{(N+1)!}{(k+1)!(N+1-(k+1))!} = \frac{(N+1)!}{(k+1)!(N-k)!}$
- $E[M] = \frac{k(N+1)}{k+1}$



#### Estimator for N

We set the observed maximum m equal to the expected value $E[M]$ to derive an estimator for N: $m \approx \frac{k(N+1)}{k+1}$

Solving for N: $N_{estimator} = m + \frac{m}{k} - 1$ (This is one common form)

The prompt provides an alternative form from the intermediate algebraic step $E[M] = \frac{k}{k+1} (N+1)$ leading to an estimator: $N = m \frac{k+1}{k} - 1$.

This can be rewritten as $N_{estimator} = m + \frac{m}{k} - 1$.

Both forms are identical and represent the same estimator. This estimator is guaranteed to be at least m (since m/k - 1 can be positive or negative but the sum provides a sensible estimate for N >= m).

As k (the number of observations) increases, the factor (k+1)/k approaches 1, meaning the inflation applied to m becomes smaller, which is intuitive as more observations should lead to a more precise estimate closer to the true maximum.
