#### Multivariate - Covariance Matrix

The covariance matrix of a random vector $X = (X^{(1)}, \dots, X^{(d)})$ in $\mathbb{R}^d$ is given by $\Sigma = \text{Cov}(X) = E[(X -E[X])(X -E[X])^T]$ (matrix of size $d \times d$)

The term on the $i$-th row and $j$-th column is given by $\Sigma_{ij} = E[(X_i -E[X_i])(X_j -E[X_j])] = \text{Cov}(X_i, X_j)$

In particular, on the diagonal we have $\Sigma_{ii} = \text{Cov}(X_i, X_i) = \text{Var}(X_i)$.

For $X \in \mathbb{R}$, $\text{Var}(aX +b) = a^2 \text{Var}(X)$

For $X \in \mathbb{R}^d$, and $A$ & $B$ are matrices,  $\text{Cov}(AX +B) = \text{Cov}(AX) = A \text{Cov}(X) A^T = A \Sigma A^T$

#### Multivariate Gaussian distribution

If $(X, Y)^T$ is a Gaussian vector that PDF depends on five parameters $E[X]$, $\text{Var}(X)$, $E[Y]$, $\text{Var}(Y)$, $\text{Cov}(X, Y)$

In general, a Gaussian vector $X$ in $\mathbb{R}^d$ is completely determined by its expected value $E[X] = \mu \in \mathbb{R}^d$ and Covariance matrix $\Sigma$.

We denote $X \sim N_d(\mu, \Sigma)$

Its PDF in $\mathbb{R}^d$ is given by $\frac{1}{(2\pi \det(\Sigma))^{d/2}} \exp\left(-\frac{1}{2} (x -\mu)^T \Sigma^{-1} (x -\mu)\right)$

#### Multivariate CLT

The CLT is generalized to averages, or random vectors or vectors of averages.

Let $X_1, \dots, X_n$ in $\mathbb{R}^d$ be independent copies of $X$, such that $E[X] = \mu$
- $\text{Cov}(X) = \Sigma$, then $\sqrt{n} (\bar{X}_n -\mu) \xrightarrow{d} N_d(0, \Sigma)$ for $n \to \infty$

Equivalently,
- $\sqrt{n} \Sigma^{-1/2} (\bar{X}_n - \mu) \xrightarrow{d} N_d(0, Id)$ for $n \to \infty$

#### Linear Transformation of Multivariate Gaussian Random Variables

If $X \sim N_d(\mu, \Sigma)$ is a $d$-dimensional Gaussian random vector, and $A$ is an $m \times d$ matrix, and $b$ is an $m$-dimensional vector, then the linear transformation $Y = AX + b$ is also a Gaussian random vector.

The distribution of $Y$ is given by:
$Y \sim N_m(A\mu + b, A\Sigma A^T)$

This means:
*   The new mean vector is $A\mu + b$.
*   The new covariance matrix is $A\Sigma A^T$.

**Example: 2-dimensional Gaussian Random Variable**

Let $X = \begin{pmatrix} X_1 \\ X_2 \end{pmatrix}$ be a 2-dimensional Gaussian random vector with mean $\mu = \begin{pmatrix} \mu_1 \\ \mu_2 \end{pmatrix}$ and covariance matrix $\Sigma = \begin{pmatrix} \sigma_1^2 & \rho\sigma_1\sigma_2 \\ \rho\sigma_1\sigma_2 & \sigma_2^2 \end{pmatrix}$.

Consider a linear transformation $Y = AX + b$, where $A = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix}$ and $b = \begin{pmatrix} b_1 \\ b_2 \end{pmatrix}$.

Then $Y$ is also a 2-dimensional Gaussian random vector with:

Mean: $E[Y] = A\mu + b = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix} \begin{pmatrix} \mu_1 \\ \mu_2 \end{pmatrix} + \begin{pmatrix} b_1 \\ b_2 \end{pmatrix}$

Covariance: $\text{Cov}(Y) = A\Sigma A^T = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix} \begin{pmatrix} \sigma_1^2 & \rho\sigma_1\sigma_2 \\ \rho\sigma_1\sigma_2 & \sigma_2^2 \end{pmatrix} \begin{pmatrix} a_{11} & a_{21} \\ a_{12} & a_{22} \end{pmatrix}$

```python
import numpy as np

# Define Original 2D Gaussian Distribution (X)
# Mean vector for X
mu_X = np.array([1, 2])

# Covariance matrix for X (must be symmetric and positive semi-definite)
Sigma_X = np.array([[1.0, 0.5], [0.5, 2.0] ])

# Number of samples to generate
n_samples = 100000 # Increased samples for better approximation

# Generate samples from the original multivariate Gaussian distribution
X_samples = np.random.multivariate_normal(mu_X, Sigma_X, n_samples)

# Define Linear Transformation: Y = A @ X + b
# Transformation matrix (A)
A = np.array([[2.0, 1.0], [0.5, 3.0] ])

# Translation vector (b)
b = np.array([5.0, -1.0])

# Apply the Transformation
# Apply the transformation to each sample X_i: Y_i = A @ X_i + b
# (Using broadcasting for b for efficient calculation)
Y_samples = (A @ X_samples.T).T + b

# Calculate Theoretical Properties for Transformed Distribution (Y)
# Theoretical Mean of Y: E[Y] = A @ E[X] + b
mu_Y_theoretical = A @ mu_X + b

# Theoretical Covariance of Y: Cov(Y) = A @ Cov(X) @ A.T
Sigma_Y_theoretical = A @ Sigma_X @ A.T

# Calculate Sample Properties for Transformed Distribution (Y)
mu_Y_sample = np.mean(Y_samples, axis=0)
Sigma_Y_sample = np.cov(Y_samples, rowvar=False)
```

#### Proof: Linear Transformation for Independent Zero-Mean Gaussian Components

Let $X$ be a $d$-dimensional Gaussian random vector with mean $E[X] = \mu$ and covariance matrix $\text{Cov}(X) = \Sigma$. We are given that $\Sigma$ is symmetric and positive definite.

We want to find a matrix $A$ and a vector $B$ such that $Y = A(X+B)$ has the following properties:
1.  $Y$ is a Gaussian random vector.
2.  $E[Y] = \mathbf{0}$ (a vector of zeros).
3.  The components of $Y$ are independent.

**Ensure Zero Mean**

First, let's address the mean of the transformed random variable $Y$. The expected value of $Y$ is $E[Y] = E[A(X+B)] = A \cdot E[X+B] = A \cdot (E[X] + B) = A(\mu + B)$.

We want $E[Y] = \mathbf{0}$. This means we need $A(\mu + B) = \mathbf{0}$. A simple way to achieve this is to choose $B = -\mu$. This shifts the original r.v. so its mean becomes zero $E[X+B] = E[X-\mu] = E[X] - \mu = \mu - \mu = \mathbf{0}$. So, with $B = -\mu$, our transformed variable becomes $Y = A(X-\mu)$, and its mean is $E[Y] = A \cdot \mathbf{0} = \mathbf{0}$.

**Ensure Independent Components (Diagonal Covariance)**

Now, let's consider the covariance of $Y = A(X-\mu)$. The covariance of a linear transformation $CZ$ is $C \cdot \text{Cov}(Z) \cdot C^T$.

Here, $Z = (X-\mu)$, and $\text{Cov}(X-\mu) = E[(X-\mu)(X-\mu)^T] = \text{Cov}(X) = \Sigma$.

So, the covariance of $Y$ is $\text{Cov}(Y) = A \cdot \text{Cov}(X-\mu) \cdot A^T = A \Sigma A^T$.

For the components of $Y$ to be independent, its covariance matrix $A \Sigma A^T$ must be a diagonal matrix. Furthermore, to make them standard Gaussian (mean 0, variance 1), we want $A \Sigma A^T = I_d$, where $I_d$ is the $d \times d$ identity matrix. Since $\Sigma$ is a symmetric positive definite matrix, it has a Cholesky decomposition. This means there exists a unique lower triangular matrix $L$ with positive diagonal entries such that $\Sigma = LL^T$.

Now, let's choose $A = L^{-1}$. Since $L$ is a lower triangular matrix with non-zero diagonal entries (because $\Sigma$ is positive definite), its inverse $L^{-1}$ exists.

Substitute this choice of $A$ into the covariance expression $\text{Cov}(Y) = A \Sigma A^T = (L^{-1})(LL^T)(L^{-1})^T$.

We know that $(L^{-1})^T = (L^T)^{-1}$. So, $\text{Cov}(Y) = L^{-1} L L^T (L^T)^{-1} = (L^{-1} L)(L^T (L^T)^{-1}) = I_d I_d = I_d$.

**Concluding the Proof**

**Gaussianity:** Since $X$ is a Gaussian random variable and $Y = A(X+B)$ is a linear transformation of $X$, $Y$ is also a Gaussian random variable. This is a fundamental property of multivariate Gaussian distributions.

**Zero Mean:** By choosing $B = -\mu$, we ensured $E[Y] = \mathbf{0}$.

**Independence:** By choosing $A = L^{-1}$ (where $L$ is from the Cholesky decomposition of $\Sigma$), we found that $\text{Cov}(Y) = I_d$. A covariance matrix equal to the identity matrix implies that the components of $Y$ are uncorrelated. For Gaussian random variables, uncorrelatedness implies independence.

Thus, by choosing $B = -\mu$ and $A = L^{-1}$ (where $\Sigma = LL^T$), we construct a Gaussian random variable $Y = A(X-\mu)$ whose components are independent and each have a mean of 0 (and a variance of 1, i.e., standard normal). Q.E.D.

#### Sample Covariance

For $(X_1, Y_1), (X_2, Y_2), \dots,(X_n, Y_n) \sim \text{iid} (X, Y)$, with $E[X]= \mu_X$, $E[Y]= \mu_Y$, $E[XY]= \mu_{XY}$ (each r.v pair $(X_i, Y_i)$ has same distribution as random pair $(X,Y)$)

Estimating the $\text{Cov}(X, Y)$ is a useful practice as non-zero covariance implies statistical dependence of $X$ and $Y$. We use one way to obtain an unbiased estimator for $\text{Cov}(X, Y)$.
We consider estimator of the covariance $S_{XY} = \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})$

$E\left[\frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})\right]$ is an unbiased estimator of $\text{Cov}(X, Y)$ (if dividing by $n-1$).



Expansion for $E\left[\left(\sum_{i=1}^{n} X_i\right)\left(\sum_{j=1}^{n} Y_j\right)\right]$

Let $S_X = \sum_{i=1}^{n} X_i$ and $S_Y = \sum_{j=1}^{n} Y_j$.
- $E[S_X S_Y] = E\left[\left(\sum_{i=1}^{n} X_i\right)\left(\sum_{j=1}^{n} Y_j\right)\right]$.

This can be expanded as $E\left[\sum_{i=1}^{n} \sum_{j=1}^{n} X_i Y_j\right]$. By linearity of expectation, we can move the expectation inside the sums $= \sum_{i=1}^{n} \sum_{j=1}^{n} E[X_i Y_j]$

Now, we split the sum into two cases, when $i = j$ and when $i \neq j$.
- $\sum_{i=1}^{n} E[X_i Y_i] + \sum_{i=1}^{n} \sum_{j \neq i}^{n} E[X_i Y_j]$

Assuming $X_k, Y_k$ are i.i.d. pairs, and $X_i, Y_i$ are independent of $X_j, Y_j$ for $i \neq j$. We know that $E[X_i Y_i] = \text{Cov}(X_i, Y_i) + E[X_i]E[Y_i]$. Since all pairs are i.i.d., $\text{Cov}(X_i, Y_i) = \text{Cov}(X, Y)$ and $E[X_i] = E[X] = \mu_X$, $E[Y_i] = E[Y] = \mu_Y$.

So, $E[X_i Y_i] = \text{Cov}(X, Y) + \mu_X \mu_Y$.

For $i \neq j$, since $(X_i, Y_i)$ is independent of $(X_j, Y_j)$, we have $E[X_i Y_j] = E[X_i]E[Y_j] = \mu_X \mu_Y$.

Substitute these back into the expanded sum:
- $\sum_{i=1}^{n} (\text{Cov}(X, Y) + \mu_X \mu_Y) + \sum_{i=1}^{n} \sum_{j \neq i}^{n} (\mu_X \mu_Y)$

The first sum has $n$ terms $= n (\text{Cov}(X, Y) + \mu_X \mu_Y)$. The second sum has $n(n-1)$ terms (there are $n$ choices for $i$, and for each $i$, there are $n-1$ choices for $j \neq i$) $= n (\text{Cov}(X, Y) + \mu_X \mu_Y) + n(n-1) (\mu_X \mu_Y) = n \text{Cov}(X, Y) + n \mu_X \mu_Y + (n^2 - n) \mu_X \mu_Y = n \text{Cov}(X, Y) + n \mu_X \mu_Y + n^2 \mu_X \mu_Y - n \mu_X \mu_Y = n \text{Cov}(X, Y) + n^2 \mu_X \mu_Y$

So, the correct expansion is:
- $E\left[\left(\sum_{i=1}^{n} X_i\right)\left(\sum_{j=1}^{n} Y_j\right)\right] = n \text{Cov}(X, Y) + n^2 E[X]E[Y]$

#### Expectation of Sample Covariance $E[S_{XY}]$

Let the sample covariance be defined as $S_{XY} = \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})$.

First, expand the term $(X_i - \bar{X})(Y_i - \bar{Y})$:

$(X_i - \bar{X})(Y_i - \bar{Y}) = X_i Y_i - X_i \bar{Y} - \bar{X} Y_i + \bar{X} \bar{Y}$

Now, sum over all $i$ and divide by $n$:

$S_{XY} = \frac{1}{n}\sum_{i=1}^{n}(X_i Y_i - X_i \bar{Y} - \bar{X} Y_i + \bar{X} \bar{Y}) = \frac{1}{n}\left(\sum_{i=1}^{n} X_i Y_i - \bar{Y}\sum_{i=1}^{n} X_i - \bar{X}\sum_{i=1}^{n} Y_i + \sum_{i=1}^{n} \bar{X} \bar{Y}\right)$

Recall that $\sum X_i = n\bar{X}$ and $\sum Y_i = n\bar{Y}$, and $\sum \bar{X} \bar{Y} = n\bar{X}\bar{Y}$ (since $\bar{X}$ and $\bar{Y}$ are constants with respect to the sum over $i$):

$S_{XY} = \frac{1}{n}\left(\sum_{i=1}^{n} X_i Y_i - \bar{Y}(n\bar{X}) - \bar{X}(n\bar{Y}) + n\bar{X}\bar{Y}\right)= \frac{1}{n}\left(\sum_{i=1}^{n} X_i Y_i - n\bar{X}\bar{Y} - n\bar{X}\bar{Y} + n\bar{X}\bar{Y}\right) = \frac{1}{n}\left(\sum_{i=1}^{n} X_i Y_i - n\bar{X}\bar{Y}\right) = \left(\frac{1}{n}\sum_{i=1}^{n} X_i Y_i\right) - \bar{X}\bar{Y}$

Now, let's take the expectation of $S_{XY}$ $E[S_{XY}] = E\left[\frac{1}{n}\sum_{i=1}^{n} X_i Y_i\right] - E[\bar{X}\bar{Y}]$

For the first term, by linearity of expectation:
$E\left[\frac{1}{n}\sum_{i=1}^{n} X_i Y_i\right] = \frac{1}{n}\sum_{i=1}^{n} E[X_i Y_i]$

Since $X_i, Y_i$ are i.i.d. pairs, $E[X_i Y_i] = E[XY]$ for all $i$. We know that $\text{Cov}(X, Y) = E[XY] - E[X]E[Y]$, so $E[XY] = \text{Cov}(X, Y) + E[X]E[Y]$.

Thus, $E\left[\frac{1}{n}\sum_{i=1}^{n} X_i Y_i\right] = \frac{1}{n} \cdot n (\text{Cov}(X, Y) + \mu_X \mu_Y) = \text{Cov}(X, Y) + \mu_X \mu_Y$

For the second term, $E[\bar{X}\bar{Y}]$:

$E[\bar{X}\bar{Y}] = E\left[\left(\frac{1}{n}\sum_{i=1}^{n} X_i\right)\left(\frac{1}{n}\sum_{j=1}^{n} Y_j\right)\right] = \frac{1}{n^2} E\left[\left(\sum_{i=1}^{n} X_i\right)\left(\sum_{j=1}^{n} Y_j\right)\right]$

From our previous derivation, we have $E\left[\left(\sum_{i=1}^{n} X_i\right)\left(\sum_{j=1}^{n} Y_j\right)\right] = n \text{Cov}(X, Y) + n^2 \mu_X \mu_Y$.

Substituting this into the expression for $E[\bar{X}\bar{Y}]$:

$E[\bar{X}\bar{Y}] = \frac{1}{n^2} (n \text{Cov}(X, Y) + n^2 \mu_X \mu_Y) = \frac{1}{n} \text{Cov}(X, Y) + \mu_X \mu_Y$

Finally, substitute both terms back into $E[S_{XY}]$:

$E[S_{XY}] = (\text{Cov}(X, Y) + \mu_X \mu_Y) - \left(\frac{1}{n} \text{Cov}(X, Y) + \mu_X \mu_Y\right) = \text{Cov}(X, Y) + \mu_X \mu_Y - \frac{1}{n} \text{Cov}(X, Y) - \mu_X \mu_Y = \text{Cov}(X, Y) - \frac{1}{n} \text{Cov}(X, Y) = \left(1 - \frac{1}{n}\right) \text{Cov}(X, Y) = \frac{n-1}{n} \text{Cov}(X, Y)$

This result shows that $S_{XY}$ (the sample covariance dividing by $n$) is a biased estimator for $\text{Cov}(X, Y)$. To obtain an unbiased estimator, one typically uses $\frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})$.


#### Discussion on Scaling Factor for Sample Covariance ($S_{XY}$)

From our derivation, we found that the expectation of $S_{XY} = \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})$ is:

$E[S_{XY}] = \frac{n-1}{n} \text{Cov}(X, Y)$

This result indicates that $S_{XY}$, when scaled by $\frac{1}{n}$, is a **biased estimator** for the true covariance $\text{Cov}(X, Y)$. Specifically, it systematically *underestimates* the true population covariance.

To obtain an **unbiased estimator** for $\text{Cov}(X, Y)$, we need to adjust the scaling factor. If we define a modified sample covariance, often denoted $S'_{XY}$, as:

$S'_{XY} = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})$

Then its expectation is:

$E[S'_{XY}] = E\left[\frac{n}{n-1} S_{XY}\right] = \frac{n}{n-1} E[S_{XY}] = \frac{n}{n-1} \cdot \frac{n-1}{n} \text{Cov}(X, Y) = \text{Cov}(X, Y)$

Therefore:
*   The scaling factor of $\frac{1}{n}$ results in a **biased estimator** for the covariance.
*   The scaling factor of $\frac{1}{n-1}$ (Bessel's correction) results in an **unbiased estimator** for the covariance.
